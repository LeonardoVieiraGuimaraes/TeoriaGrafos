- c) Um conjunto não vazio de vértices e um conjunto de arestas, onde cada aresta conecta dois vértices.
- d) Um conjunto de arestas conectadas em sequência, formando um ciclo completo.

### Questão de Múltipla Escolha: Subgrafos

Considere o grafo principal abaixo e os quatro grafos apresentados (a, b, c, d). Apenas um deles é um subgrafo válido do grafo principal. Identifique qual é o subgrafo correto.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar o grafo principal
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

# Desenhar o grafo principal
plt.figure(figsize=(10, 5))
plt.subplot(1, 5, 1)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Grafo Principal')

# Criar e desenhar os subgrafos (a, b, c, d)
subgrafos = {
    'a': [(1, 2), (2, 3), (3, 3)],
    'b': [(1, 2), (2, 4)],
    'c': [(2, 3), (2, 4)],
    'd': [(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)]
}
for i, (label, edges) in enumerate(subgrafos.items(), start=2):
    SG = nx.Graph()
    SG.add_edges_from(edges)
    plt.subplot(1, 5, i)
    nx.draw(SG, with_labels=True, node_color='lightgreen', edge_color='black', node_size=500, font_size=10)
    plt.title(f'Subgrafo {label}')

plt.tight_layout()
plt.show()
```

Qual dos grafos apresentados (a, b, c, d) é um subgrafo válido do grafo principal?

- a) Subgrafo a
- b) Subgrafo b
- c) Subgrafo c
- d) Subgrafo d

**Resposta correta:** c) Subgrafo c

**Por que o subgrafo c é válido?**

Um subgrafo é válido se todos os seus vértices e arestas pertencem ao grafo principal. No caso do subgrafo `c`, temos:

- **Vértices do subgrafo c:** {1, 3, 4}
- **Arestas do subgrafo c:** {(1, 3), (3, 4)}

Comparando com o grafo principal:
- **Vértices do grafo principal:** {1, 2, 3, 4}
- **Arestas do grafo principal:** {(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)}

Todos os vértices e arestas do subgrafo `c` estão contidos no grafo principal, o que o torna um subgrafo válido.

6. **Grafo Bipartido**

7. **Grafo Bipartido Completo**  
   Um grafo bipartido onde todos os vértices de um conjunto estão conectados a todos os vértices do outro conjunto.

8. **Grafo Cubo (Hipercubo)**  
   Um grafo onde os vértices representam palavras binárias de k bits, conectados se diferirem em exatamente um bit.


### Associe cada tipo de grafo descrito abaixo com o desenho correspondente

```python
import networkx as nx
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(20, 10))

# Grafo Regular
G_regular = nx.cycle_graph(4)
nx.draw(G_regular, ax=axes[0, 0], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[0, 0].set_title('Grafo A')

# Grafo Nulo
G_nulo = nx.empty_graph(4)
nx.draw(G_nulo, ax=axes[0, 1], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[0, 1].set_title('Grafo B')

# Grafo Completo
G_completo = nx.complete_graph(4)
nx.draw(G_completo, ax=axes[0, 2], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[0, 2].set_title('Grafo C')

# Grafo Ciclo
G_ciclo = nx.cycle_graph(5)
nx.draw(G_ciclo, ax=axes[0, 3], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[0, 3].set_title('Grafo D')

# Árvore
G_arvore = nx.balanced_tree(2, 2)
nx.draw(G_arvore, ax=axes[1, 0], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[1, 0].set_title('E')

# Grafo Bipartido
G_bipartido = nx.Graph()
V1 = [1, 2, 3]
V2 = ['A', 'B', 'C']
G_bipartido.add_nodes_from(V1, bipartite=0)
G_bipartido.add_nodes_from(V2, bipartite=1)
G_bipartido.add_edges_from([(1, 'A'), (2, 'B'), (3, 'C')])
pos = nx.drawing.layout.bipartite_layout(G_bipartido, V1)
nx.draw(G_bipartido, pos, ax=axes[1, 1], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[1, 1].set_title('Grafo F')

# Grafo Bipartido Completo
G_bipartido_completo = nx.complete_bipartite_graph(3, 3)
pos = nx.drawing.layout.bipartite_layout(G_bipartido_completo, range(3))
nx.draw(G_bipartido_completo, pos, ax=axes[1, 2], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[1, 2].set_title('Grafo G')

# Grafo Cubo (Hipercubo)
G_cubo = nx.hypercube_graph(3)
nx.draw(G_cubo, ax=axes[1, 3], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500)
axes[1, 3].set_title('Grafo H')

plt.tight_layout()
plt.show()
```
