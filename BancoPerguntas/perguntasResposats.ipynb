### Questão de Múltipla Escolha: Definição de Grafo

Qual das alternativas a seguir define corretamente um grafo?

- a) Uma estrutura composta apenas por vértices sem arestas conectando-os.
- b) Um conjunto vazio de vértices e um conjunto vazio de arestas.
- c) Um conjunto não vazio de vértices e um conjunto de arestas, onde cada aresta conecta dois vértices.
- d) Um conjunto de arestas conectadas em sequência, formando um ciclo completo.

**Resposta correta:** c) Um grafo é definido como um conjunto não vazio de vértices (V) e um conjunto (E) de arestas, onde cada aresta conecta dois vértices.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo simples
G = nx.Graph()

# Adicionar vértices
G.add_nodes_from([1, 2, 3, 4])

# Adicionar arestas
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Desenhar o grafo
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Exemplo de Grafo')
plt.show()
```

### Questão de Múltipla Escolha: Definição de Grafo Direcionado

Qual das alternativas a seguir define corretamente um grafo direcionado (ou dígrafo)?

- a) Um grafo onde as arestas não possuem direção e conectam dois vértices em ambas as direções.
- b) Um grafo onde as arestas possuem direção, conectando um vértice origem a um vértice destino.
- c) Um grafo onde as arestas formam ciclos completos entre todos os vértices.
- d) Um grafo onde não existem arestas conectando os vértices.

**Resposta correta:** b) Um grafo direcionado é definido por um conjunto não vazio de vértices (V) e um conjunto de arestas (E), onde cada aresta é um par ordenado de vértices (u, v), sendo u o vértice origem e v o vértice destino.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo direcionado
DG = nx.DiGraph()

# Adicionar vértices
DG.add_nodes_from([1, 2, 3, 4])

# Adicionar arestas direcionadas
DG.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Desenhar o grafo direcionado
nx.draw(DG, with_labels=True, node_color='lightgreen', edge_color='black', node_size=500, font_size=10, arrowsize=20)
plt.title('Exemplo de Grafo Direcionado')
plt.show()
```

### Questão de Correspondência: Tipos de Grafos

Associe os tipos de grafos às suas respectivas definições:

1. Grafo simples
2. Multigrafo
3. Laço
4. Arestas paralelas

a) Um grafo que não contém laços nem arestas paralelas.
b) Um grafo que contém pelo menos um laço ou um conjunto de arestas paralelas.
c) Uma aresta que conecta um vértice a ele mesmo.
d) Conjuntos de duas ou mais arestas que conectam o mesmo par de vértices.

**Resposta correta:**
1-a, 2-b, 3-c, 4-d

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um multigrafo
MG = nx.MultiGraph()

# Adicionar vértices
MG.add_nodes_from([1, 2, 3])

# Adicionar arestas (incluindo laços e arestas paralelas)
MG.add_edge(1, 1)  # Laço
MG.add_edge(1, 2)
MG.add_edge(1, 2)  # Aresta paralela
MG.add_edge(2, 3)

# Desenhar o multigrafo
pos = nx.spring_layout(MG)
nx.draw(MG, pos, with_labels=True, node_color='lightcoral', edge_color='black', node_size=500, font_size=10)
plt.title('Exemplo de Multigrafo com Laço e Arestas Paralelas')
plt.show()
```

### Questão de Correspondência: Conceitos de Grafos

Associe os conceitos de grafos às suas respectivas definições:

1. Ordem
2. Tamanho
3. Densidade
4. Grau de vértice
5. Sequência de graus

a) Representa a quantidade de vértices do grafo.
b) Representa a quantidade de arestas do grafo.
c) Proporção entre a quantidade de arestas e vértices do grafo.
d) Quantidade de arestas incidentes a um vértice, com laços contados duas vezes.
e) Lista dos graus de todos os vértices em ordem não-decrescente.

**Resposta correta:**
1-a, 2-b, 3-c, 4-d, 5-e

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo simples
G = nx.Graph()

# Adicionar vértices e arestas
G.add_nodes_from([1, 2, 3, 4, 5])
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5), (5, 1), (1, 3)])

# Calcular ordem, tamanho e sequência de graus
ordem = G.number_of_nodes()
tamanho = G.number_of_edges()
sequencia_graus = sorted([deg for _, deg in G.degree()])

# Exibir informações
print(f'Ordem do grafo: {ordem}')
print(f'Tamanho do grafo: {tamanho}')
print(f'Sequência de graus: {sequencia_graus}')

# Desenhar o grafo
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Exemplo de Grafo para Conceitos de Ordem e Tamanho')
plt.show()
```

### Questão de Múltipla Escolha: Subgrafos

Considere o grafo principal abaixo e os quatro grafos apresentados (a, b, c, d). Apenas um deles é um subgrafo válido do grafo principal. Identifique qual é o subgrafo correto.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar o grafo principal
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

# Desenhar o grafo principal
plt.figure(figsize=(10, 5))
plt.subplot(1, 5, 1)
nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Grafo Principal')

# Criar e desenhar os subgrafos (a, b, c, d)
subgrafos = {
    'a': [(1, 2), (2, 3), (3, 3)],
    'b': [(1, 2), (2, 4)],
    'c': [(2, 3), (2, 4)],
    'd': [(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)]
}
for i, (label, edges) in enumerate(subgrafos.items(), start=2):
    SG = nx.Graph()
    SG.add_edges_from(edges)
    plt.subplot(1, 5, i)
    nx.draw(SG, with_labels=True, node_color='lightgreen', edge_color='black', node_size=500, font_size=10)
    plt.title(f'Subgrafo {label}')

plt.tight_layout()
plt.show()
```

Qual dos grafos apresentados (a, b, c, d) é um subgrafo válido do grafo principal?

- a) Subgrafo a
- b) Subgrafo b
- c) Subgrafo c
- d) Subgrafo d

**Resposta correta:** c) Subgrafo c

**Por que o subgrafo c é válido?**

Um subgrafo é válido se todos os seus vértices e arestas pertencem ao grafo principal. No caso do subgrafo `c`, temos:

- **Vértices do subgrafo c:** {1, 3, 4}
- **Arestas do subgrafo c:** {(1, 3), (3, 4)}

Comparando com o grafo principal:
- **Vértices do grafo principal:** {1, 2, 3, 4}
- **Arestas do grafo principal:** {(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)}

Todos os vértices e arestas do subgrafo `c` estão contidos no grafo principal, o que o torna um subgrafo válido.

### Questão de Múltipla Escolha: Isomorfismo de Grafos

Considere os grafos abaixo, representados como 𝐺₁ e 𝐺₂. Ambos possuem as seguintes características:

- Número de vértices: 4
- Número de arestas: 5
- Graus dos vértices: [2, 2, 3, 3]

Com base nessas informações, responda: Os grafos 𝐺₁ e 𝐺₂ são isomorfos?

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criar o grafo G1
G1 = nx.Graph()
G1.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'), ('a', 'c')])

# Criar o grafo G2
G2 = nx.Graph()
G2.add_edges_from([('w', 'x'), ('x', 'y'), ('y', 'z'), ('z', 'w'), ('w', 'y')])

# Desenhar os grafos
fig, axes = plt.subplots(1, 2, figsize=(10, 5))
nx.draw(G1, ax=axes[0], with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
axes[0].set_title('Grafo G₁')

nx.draw(G2, ax=axes[1], with_labels=True, node_color='lightgreen', edge_color='gray', node_size=500, font_size=10)
axes[1].set_title('Grafo G₂')

plt.tight_layout()
plt.show()
```

a) Sim, porque possuem o mesmo número de vértices, arestas e graus dos vértices.

b) Não, porque possuem diferentes números de vértices.

c) Não, porque possuem diferentes números de arestas.

d) Não, porque possuem diferentes graus dos vértices.
