# Questão: Caracterização de Árvores em Grafos

Uma empresa de telecomunicações deseja instalar cabos de fibra óptica para conectar 6 cidades. O objetivo é minimizar o custo total da instalação, garantindo que todas as cidades estejam conectadas diretamente ou indiretamente, sem formar ciclos.

O grafo abaixo representa as cidades (vértices) e os custos de instalação entre elas (arestas com pesos):

1. Cidade A - Cidade B: 4
2. Cidade A - Cidade C: 3
3. Cidade B - Cidade C: 2
4. Cidade B - Cidade D: 6
5. Cidade C - Cidade D: 5
6. Cidade C - Cidade E: 7
7. Cidade D - Cidade E: 8
8. Cidade D - Cidade F: 9
9. Cidade E - Cidade F: 6

Qual é o custo total mínimo para conectar todas as cidades sem formar ciclos?

### Opções:
a) 16  
b) 18  
c) 20  
d) 22

```python
# Criando o grafo com os custos de instalação
import networkx as nx

# Definindo o grafo
G = nx.Graph()
edges = [
    ("A", "B", 4),
    ("A", "C", 3),
    ("B", "C", 2),
    ("B", "D", 6),
    ("C", "D", 5),
    ("C", "E", 7),
    ("D", "E", 8),
    ("D", "F", 9),
    ("E", "F", 6),
]
G.add_weighted_edges_from(edges)

# Encontrando a árvore geradora mínima
mst = nx.minimum_spanning_tree(G)

# Calculando o custo total mínimo
total_cost = sum(data["weight"] for _, _, data in mst.edges(data=True))
print(f"Custo total mínimo: {total_cost}")

# Visualizando a árvore geradora mínima
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)  # Layout para o grafo
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10)
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=8
)
plt.title("Grafo com Custos de Instalação")
plt.show()

# Visualizando apenas a árvore geradora mínima
plt.figure(figsize=(10, 6))
nx.draw(mst, pos, with_labels=True, node_color="lightgreen", node_size=500, font_size=10)
nx.draw_networkx_edge_labels(
    mst, pos, edge_labels=nx.get_edge_attributes(mst, "weight"), font_size=8
)
plt.title("Árvore Geradora Mínima")
plt.show()
```

### Resposta Correta:
b) 18

A árvore geradora mínima conecta todas as cidades com o menor custo total, que é 18.

# Questão: Caracterização de Árvores em Grafos

Uma empresa de logística deseja construir uma rede de estradas para conectar 5 cidades de forma eficiente. A rede deve ser construída de forma que:
1. Todas as cidades estejam conectadas diretamente ou indiretamente.
2. Não existam ciclos na rede.
3. A remoção de qualquer estrada desconecte a rede.

Com base nas condições acima e nas definições de árvores, qual das opções abaixo representa corretamente o número de estradas necessárias para conectar as 5 cidades?

### Opções:
a) 3  
b) 4  
c) 5  
d) 6  
e) Nenhuma das anteriores

```python
# Criando o grafo para representar a rede de estradas
import networkx as nx
import matplotlib.pyplot as plt

# Definindo o grafo
G = nx.Graph()
edges = [
    ("Cidade 1", "Cidade 2"),
    ("Cidade 2", "Cidade 3"),
    ("Cidade 3", "Cidade 4"),
    ("Cidade 4", "Cidade 5"),
]
G.add_edges_from(edges)

# Verificando as propriedades da árvore
n_vertices = len(G.nodes)
n_arestas = len(G.edges)

print(f"Número de vértices (cidades): {n_vertices}")
print(f"Número de arestas (estradas): {n_arestas}")
print(f"O grafo é uma árvore? {nx.is_tree(G)}")

# Visualizando o grafo com layout ajustado
pos = nx.circular_layout(G)  # Usando layout circular para melhor visualização
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10, edge_color="gray")
plt.title("Rede de Estradas (Árvore)")
plt.show()
```

### Resposta Correta:
b) 4

### Feedback:
- **a) 3**: Incorreto. Para conectar 5 cidades, são necessárias pelo menos 4 estradas, pois uma árvore com $n$ vértices possui exatamente $n - 1$ arestas.
- **b) 4**: Correto. De acordo com o Teorema 1, uma árvore com $n$ vértices possui exatamente $n - 1$ arestas. Para 5 cidades ($n = 5$), são necessárias $5 - 1 = 4$ estradas.
- **c) 5**: Incorreto. Se houver 5 estradas, o grafo terá um ciclo, o que viola a definição de uma árvore.
- **d) 6**: Incorreto. Com 6 estradas, o grafo terá múltiplos ciclos, o que não é permitido em uma árvore.
- **e) Nenhuma das anteriores**: Incorreto. A resposta correta é 4, conforme explicado acima.

- **c) A rede elétrica é composta por múltiplos componentes desconexos, cada um sendo minimamente conexo e sem ciclos**: Correto. Cada componente é uma árvore, e a união desses componentes forma a rede.

# Questão: Busca em Largura (BFS) no Contexto de Transporte Público

Uma cidade está planejando otimizar seu sistema de transporte público. A cidade é representada como um grafo, onde os vértices são os bairros e as arestas representam as rotas de ônibus entre os bairros. A prefeitura deseja encontrar o menor número de paradas de ônibus necessárias para ir de um bairro inicial até um bairro de destino.

Considere o grafo abaixo, onde os bairros e as rotas são representados como:

1. Bairro A - Bairro B  
2. Bairro A - Bairro C  
3. Bairro B - Bairro D  
4. Bairro C - Bairro D  
5. Bairro C - Bairro E  
6. Bairro D - Bairro F  
7. Bairro E - Bairro F  

Qual é o menor número de paradas de ônibus necessárias para ir do Bairro A ao Bairro F?

### Opções:
a) 2  
b) 3  
c) 4  
d) 5  
e) Nenhuma das anteriores

```python
# Criando o grafo para representar os bairros e as rotas de ônibus
import networkx as nx

# Definindo o grafo
G = nx.Graph()
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("C", "D"),
    ("C", "E"),
    ("D", "F"),
    ("E", "F"),
]
G.add_edges_from(edges)

# Implementando a busca em largura (BFS) para encontrar o menor caminho
start = "A"
end = "F"
shortest_path_length = nx.shortest_path_length(G, source=start, target=end)

print(f"O menor número de paradas de ônibus de {start} para {end} é: {shortest_path_length}")

# Visualizando o grafo
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10, edge_color="gray")
plt.title("Grafo Representando o Sistema de Transporte Público")
plt.show()
```

### Resposta Correta:
b) 3

### Feedback:
- **a) 2**: Incorreto. O menor caminho de A para F passa por 3 paradas: A → C → D → F.
- **b) 3**: Correto. A busca em largura encontra o menor caminho em termos de número de arestas, que é 3.
- **c) 4**: Incorreto. O menor caminho é mais curto do que 4 paradas.
- **d) 5**: Incorreto. O menor caminho é significativamente mais curto do que 5 paradas.
- **e) Nenhuma das anteriores**: Incorreto. A resposta correta é 3.

# Questão: Busca em Largura (BFS) e Vetor de Roteamento

Considere o grafo abaixo, onde os armazéns e as rotas são representados como:

1. Armazém A - Armazém B  
2. Armazém A - Armazém C  
3. Armazém B - Armazém D  
4. Armazém C - Armazém D  
5. Armazém C - Armazém E  
6. Armazém D - Armazém F  
7. Armazém E - Armazém F  

Qual dos vetores de roteamento abaixo, gerado a partir do Armazém A, está correto?

### Opções:

#### Opção a)
| $v$   | A   | B   | C   | D   | E   | F   |
|-------|-----|-----|-----|-----|-----|-----|

#### Opção b)
| $v$   | A   | B   | C   | D   | E   | F   |
|-------|-----|-----|-----|-----|-----|-----|
| $p(v)$ | nil | A   | B   | C   | D   | E   |

#### Opção c)
| $v$   | A   | B   | C   | D   | E   | F   |
|-------|-----|-----|-----|-----|-----|-----|
| $p(v)$ | nil | A   | A   | C   | D   | E   |

#### Opção d)
| $v$   | A   | B   | C   | D   | E   | F   |
|-------|-----|-----|-----|-----|-----|-----|
| $p(v)$ | nil | C   | A   | B   | C   | D   |

#### Opção e)

```python
# Criando o grafo para representar os armazéns e as rotas
import networkx as nx

# Definindo o grafo
G = nx.Graph()
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("C", "D"),
    ("C", "E"),
    ("D", "F"),
    ("E", "F"),
]
G.add_edges_from(edges)

# Implementando a busca em largura (BFS) para calcular o vetor de roteamento com predecessores
start = "A"
predecessors = nx.bfs_predecessors(G, source=start)
routing_vector = {start: "nil"}  # O vértice raiz tem predecessor 'nil'
routing_vector.update({v: u for u, v in predecessors})

print(f"Vetor de roteamento com predecessores a partir do armazém {start}: {routing_vector}")

# Visualizando o grafo
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10, edge_color="gray")
plt.title("Grafo Representando o Sistema de Logística")
plt.show()
```

# Questão: Árvore Geradora Mínima com o Algoritmo de Prim

Uma empresa deseja instalar uma rede de comunicação entre seus escritórios localizados em diferentes cidades. As cidades são representadas como vértices de um grafo, e os custos de instalação entre as cidades são representados como pesos das arestas. O objetivo é conectar todas as cidades com o menor custo possível, utilizando o algoritmo de Prim para encontrar a Árvore Geradora Mínima (AGM).

O grafo é definido da seguinte forma:

- Vértices: A, B, C, D, S (S é a cidade inicial escolhida como raiz).
- Arestas e seus custos:
  1. S - A: 2  
  2. S - C: 4  
  3. S - D: 7  
  4. A - B: 6  
  5. C - B: 5  

Utilizando o algoritmo de Prim, qual é o resultado final da Árvore Geradora Mínima (AGM), incluindo as chaves e os predecessores de cada vértice?

### Opções

#### a)
| Vértice         | S    | A    | C    | D    | B    |
|-----------------|------|------|------|------|------|
| **Chave**       | 0    | 2    | 4    | 7    | 5    |
| **Predecessor** | raiz | S    | S    | S    | C    |

#### b)
| Vértice         | S    | A    | C    | D    | B    |
|-----------------|------|------|------|------|------|
| **Chave**       | 0    | 6    | 4    | 7    | 5    |
| **Predecessor** | raiz | B    | S    | S    | C    |

#### c)
| Vértice         | S    | A    | C    | D    | B    |
|-----------------|------|------|------|------|------|
| **Chave**       | 0    | 2    | 4    | 7    | 6    |
| **Predecessor** | raiz | S    | S    | S    | A    |

#### d)
| Vértice         | S    | A    | C    | D    | B    |
|-----------------|------|------|------|------|------|
| **Chave**       | 0    | 2    | 4    | 7    | 5    |
| **Predecessor** | raiz | S    | S    | S    | S    |

Nenhuma das anteriores

```python
# Implementando o algoritmo de Prim para encontrar a Árvore Geradora Mínima (AGM)
import networkx as nx
import heapq

# Criando o grafo
G = nx.Graph()
edges = [
    ("S", "A", 2),
    ("S", "C", 4),
    ("S", "D", 7),
    ("A", "B", 6),
    ("C", "B", 5),
]
G.add_weighted_edges_from(edges)

# Inicializando as estruturas para o algoritmo de Prim
start = "S"
mst = []  # Lista para armazenar as arestas da AGM
visited = set()
min_heap = [(0, start, None)]  # (peso, nó atual, predecessor)
chaves = {v: float("inf") for v in G.nodes}
predecessores = {v: None for v in G.nodes}
chaves[start] = 0

# Executando o algoritmo de Prim
while min_heap:
    peso, atual, predecessor = heapq.heappop(min_heap)
    if atual not in visited:
        visited.add(atual)
        if predecessor:
            mst.append((predecessor, atual, peso))
        predecessores[atual] = predecessor
        chaves[atual] = peso
        for vizinho, dados in G[atual].items():
            if vizinho not in visited:
                heapq.heappush(min_heap, (dados["weight"], vizinho, atual))

# Exibindo o resultado final
print("Resultado Final:")
for v in G.nodes:
    predecessor = predecessores[v] if predecessores[v] else "raiz"
    print(f"{v}: chave = {chaves[v]}, predecessor = {predecessor}")

# Visualizando o grafo original
import matplotlib.pyplot as plt

pos = nx.spring_layout(G)
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10)
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=8
)
plt.title("Grafo Original")
plt.show()

# Visualizando a Árvore Geradora Mínima
mst_graph = nx.Graph()
mst_graph.add_weighted_edges_from([(u, v, w) for u, v, w in mst])
plt.figure(figsize=(10, 6))
nx.draw(mst_graph, pos, with_labels=True, node_color="lightgreen", node_size=500, font_size=10)
nx.draw_networkx_edge_labels(
    mst_graph, pos, edge_labels=nx.get_edge_attributes(mst_graph, "weight"), font_size=8
)
plt.title("Árvore Geradora Mínima (AGM)")
plt.show()
```

### Resposta Correta:

A: chave = 2, predecessor = S  
B: chave = 5, predecessor = C  
C: chave = 4, predecessor = S  
D: chave = 7, predecessor = S  
S: chave = 0, predecessor = raiz

- **c)** Incorreto. O predecessor de B está incorreto.

# Questão: Trajetos Hamiltonianos e Eulerianos

Considere o grafo \( G(V, A) \) abaixo, onde os vértices representam cidades e as arestas representam estradas entre elas. O objetivo é analisar as propriedades do grafo em relação aos trajetos Hamiltonianos e Eulerianos.

### Grafo \( G(V, A) \):
- Vértices: \( V = \{A, B, C, D, E, F\} \)
- Arestas e seus pesos:
  1. A - B: 5  
  2. A - C: 7  
  3. B - C: 4  
  4. B - D: 6  
  5. C - D: 3  
  6. C - E: 8  
  7. D - E: 2  
  8. D - F: 9  
  9. E - F: 6  

Com base nas propriedades do grafo, qual das afirmações abaixo é correta?

### Opções:
a) No Grafo \( G(V, A) \), não é possível ter um trajeto Euleriano de “A” até “F”.  
b) No Grafo \( G(V, A) \), não é possível ter um trajeto Hamiltoniano de “A” até “F”.  
c) É possível construir um trajeto Euleriano entre “A” e “F”, com tamanho total igual a 28.  
d) É possível construir um trajeto Hamiltoniano entre “A” e “F”, com tamanho total igual a 30.  
e) O Grafo \( G(V, A) \) não permite um trajeto Hamiltoniano nem Euleriano.

```python
# Verificando as propriedades do grafo para trajetos Hamiltonianos e Eulerianos
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.Graph()
edges = [
    ("A", "B", 5),
    ("A", "C", 7),
    ("B", "C", 4),
    ("B", "D", 6),
    ("C", "D", 3),
    ("C", "E", 8),
    ("D", "E", 2),
    ("D", "F", 9),
    ("E", "F", 6),
]
G.add_weighted_edges_from(edges)

# Verificando se o grafo possui um trajeto Euleriano
is_eulerian = nx.is_eulerian(G)

# Verificando se o grafo possui um trajeto Hamiltoniano
# (Um trajeto Hamiltoniano visita todos os vértices exatamente uma vez)
def has_hamiltonian_path(graph):
    from itertools import permutations
    nodes = list(graph.nodes)
    for perm in permutations(nodes):
        if all(graph.has_edge(perm[i], perm[i + 1]) for i in range(len(perm) - 1)):
            return True
    return False

has_hamiltonian = has_hamiltonian_path(G)

# Exibindo os resultados
print(f"O grafo possui trajeto Euleriano? {is_eulerian}")
print(f"O grafo possui trajeto Hamiltoniano? {has_hamiltonian}")

# Visualizando o grafo
pos = nx.spring_layout(G)
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=500, font_size=10)
nx.draw_networkx_edge_labels(
    G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=8
)
plt.title("Grafo Representando as Cidades e Estradas")
plt.show()
```

### Resposta Correta:

# Questão: Algoritmo de Dijkstra para Caminhos Mínimos

Uma empresa de entregas deseja calcular o menor custo de transporte entre o depósito principal (vértice A) e os demais centros de distribuição em uma cidade. O grafo abaixo representa os centros (vértices) e as estradas (arestas com pesos):

- Vértices: A, B, C, D, E
- Arestas e seus custos:
  1. A – B: 2  
  2. A – C: 5  
  3. B – C: 1  
  4. B – D: 2  
  5. C – D: 3  
  6. C – E: 1  
  7. D – E: 2  

Utilizando o algoritmo de Dijkstra a partir do vértice A, qual das tabelas abaixo representa corretamente os menores custos (`dv`) e predecessores (`pv`) para cada vértice ao final do algoritmo?

### Opções:

#### a)
| Vértice |  A  |  B  |  C  |  D  |  E  |
|---------|-----|-----|-----|-----|-----|
|   dv    |  0  |  2  |  3  |  4  |  4  |
|   pv    | nil |  A  |  B  |  B  |  C  |

#### b)
| Vértice |  A  |  B  |  C  |  D  |  E  |
|---------|-----|-----|-----|-----|-----|
|   dv    |  0  |  2  |  5  |  4  |  6  |
|   pv    | nil |  A  |  A  |  B  |  D  |

#### c)
| Vértice |  A  |  B  |  C  |  D  |  E  |
|---------|-----|-----|-----|-----|-----|
|   dv    |  0  |  2  |  3  |  5  |  4  |
|   pv    | nil |  A  |  B  |  C  |  C  |

#### d)
| Vértice |  A  |  B  |  C  |  D  |  E  |
|---------|-----|-----|-----|-----|-----|
|   dv    |  0  |  2  |  3  |  5  |  5  |
|   pv    | nil |  A  |  B  |  C  |  D  |

#### e)
Nenhuma das anteriores

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo com os centros de distribuição e estradas (arestas com pesos)
edges = [
    ("A", "B", 2),
    ("A", "C", 5),
    ("B", "C", 1),
    ("B", "D", 2),
    ("C", "D", 3),
    ("C", "E", 1),
    ("D", "E", 2),
]
G = nx.Graph()
G.add_weighted_edges_from(edges)

# Executando o algoritmo de Dijkstra a partir do vértice A
dv, pv = nx.single_source_dijkstra(G, source="A")

# Exibindo os menores custos (dv) e predecessores (pv)
print("Menores custos (dv) a partir de A:")
for v in G.nodes:
    print(f"{v}: {dv[v]}")

print("\nPredecessores (pv) no menor caminho a partir de A:")
for v in G.nodes:
    if v == "A":
        print(f"{v}: nil")
    else:
        # O predecessor é o penúltimo vértice no caminho de A até v
        caminho = pv[v]
        predecessor = caminho[-2] if len(caminho) > 1 else "nil"
        print(f"{v}: {predecessor}")

# Visualizando o grafo
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=700, font_size=14)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=12)
plt.title("Grafo dos Centros de Distribuição")
plt.show()
```

### Questão: Matriz de Custos de um Grafo Valorado

Uma empresa de energia está planejando instalar cabos entre cinco cidades: A, B, C, D e E. O custo de instalação entre cada par de cidades está representado nas arestas do grafo abaixo:

- Arestas e custos:
    - A–B: 4
    - A–C: 7
    - B–C: 2
    - B–D: 5
    - C–D: 3
    - C–E: 6
    - D–E: 1

Qual das alternativas abaixo representa corretamente a matriz de custos (matriz de adjacência valorada) desse grafo?

#### a)
\[
\begin{bmatrix}
0 & 4 & 7 & \infty & \infty \\
4 & 0 & 2 & 5 & \infty \\
7 & 2 & 0 & 3 & 6 \\
\infty & 5 & 3 & 0 & 1 \\
\infty & \infty & 6 & 1 & 0 \\
\end{bmatrix}
\]

#### b)
\[
\begin{bmatrix}
0 & 4 & 2 & 5 & \infty \\
4 & 0 & 7 & 3 & 6 \\
2 & 7 & 0 & 1 & \infty \\
5 & 3 & 1 & 0 & \infty \\
\infty & 6 & \infty & \infty & 0 \\
\end{bmatrix}
\]

#### c)
\[
\begin{bmatrix}
0 & 4 & 7 & 5 & \infty \\
4 & 0 & 2 & 3 & 6 \\
7 & 2 & 0 & 1 & 3 \\
5 & 3 & 1 & 0 & 6 \\
\infty & 6 & 3 & 6 & 0 \\
\end{bmatrix}
\]

#### d)
\[
\begin{bmatrix}
0 & 4 & 7 & 5 & 6 \\
4 & 0 & 2 & 3 & 1 \\
7 & 2 & 0 & 1 & 3 \\
5 & 3 & 1 & 0 & 6 \\
6 & 1 & 3 & 6 & 0 \\
\end{bmatrix}
\]

#### e)
Nenhuma das anteriores

```python
import numpy as np

labels = ["A", "B", "C", "D", "E"]
edges = [
    ("A", "B", 4),
    ("A", "C", 7),
    ("B", "C", 2),
    ("B", "D", 5),
    ("C", "D", 3),
    ("C", "E", 6),
    ("D", "E", 1),
]

# Inicializa a matriz com infinito (ou um valor alto)
W = np.full((5, 5), np.inf)
for u, v, w in edges:
    i, j = labels.index(u), labels.index(v)
    W[i, j] = w
    W[j, i] = w  # Grafo não direcionado

print("Matriz de custos (W):")
print(W)
```
