Uma empresa de energia deseja conectar 5 cidades com o menor custo possível. As cidades são representadas como vértices de um grafo, e os custos de instalação entre as cidades são representados como pesos das arestas:

- Cidades: {A, B, C, D, E}
- Custos de instalação:
  1. A - B: 6  
  2. A - C: 4  
  3. B - C: 5  
  4. B - D: 7  
  5. C - D: 3  
  6. C - E: 8  
  7. D - E: 2  

Qual é o custo total mínimo para conectar todas as cidades sem formar ciclos?

### Opções:
a) 14  
b) 15  
c) 16  
d) 17  
e) Nenhuma das anteriores

```python
# filepath: d:\GitHub\NewtonPaiva\TeoriaGrafos\BancoPerguntas\QuestoesContextualizadas.ipynb
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo
G = nx.Graph()
edges = [
    ("A", "B", 6),
    ("A", "C", 4),
    ("B", "C", 5),
    ("B", "D", 7),
    ("C", "D", 3),
    ("C", "E", 8),
    ("D", "E", 2),
]
G.add_weighted_edges_from(edges)

# Calculando a árvore geradora mínima
mst = nx.minimum_spanning_tree(G)
total_cost = sum(data["weight"] for _, _, data in mst.edges(data=True))

# Visualizando o grafo original
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=700, font_size=14)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=12)
plt.title("Grafo Original com Custos de Instalação")
plt.show()

# Visualizando a árvore geradora mínima
plt.figure(figsize=(8, 6))
nx.draw(mst, pos, with_labels=True, node_color="lightgreen", node_size=700, font_size=14)
nx.draw_networkx_edge_labels(mst, pos, edge_labels=nx.get_edge_attributes(mst, "weight"), font_size=12)
plt.title(f"Árvore Geradora Mínima (Custo Total = {total_cost})")
plt.show()

# Exibindo o custo total mínimo
print(f"Custo total mínimo para conectar todas as cidades: {total_cost}")
```

### Resposta Correta:
a) 14

- **b) 15**: Incorreto. Este valor é maior que o custo total mínimo calculado.
- **c) 16**: Incorreto. Este valor é maior que o custo total mínimo calculado.
- **d) 17**: Incorreto. Este valor é maior que o custo total mínimo calculado.
- **e) Nenhuma das anteriores**: Incorreto. O custo total mínimo é 14, que está listado na opção **a**.

## Questão 2: Busca em Largura no Contexto de Transporte Público

Uma cidade deseja otimizar seu sistema de transporte público. A cidade é representada como um grafo, onde os vértices são os bairros e as arestas representam as rotas de ônibus entre os bairros:

- Bairros: {A, B, C, D, E, F}
- Rotas:
  1. A - B  
  2. A - C  
  3. B - D  
  4. C - D  
  5. C - E  
  6. D - F  
  7. E - F  

Qual é o menor número de paradas de ônibus necessárias para ir do Bairro A ao Bairro F?

### Opções:
a) 2  
b) 3  
c) 4  
d) 5  
e) Nenhuma das anteriores

```python
# filepath: d:\GitHub\NewtonPaiva\TeoriaGrafos\BancoPerguntas\QuestoesContextualizadas.ipynb
# Criando o grafo
G = nx.Graph()
edges = [
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("C", "D"),
    ("C", "E"),
    ("D", "F"),
    ("E", "F"),
]
G.add_edges_from(edges)

# Calculando o menor caminho
shortest_path_length = nx.shortest_path_length(G, source="A", target="F")

# Visualizando o grafo
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=700, font_size=14)
plt.title("Grafo Representando o Sistema de Transporte Público")
plt.show()
```

### Resposta Correta:
b) 3

### Feedback:
- **a) 2**: Incorreto. O menor caminho de A para F passa por 3 paradas: A → C → D → F.
- **b) 3**: Correto. A busca em largura encontra o menor caminho em termos de número de arestas, que é 3.
- **c) 4**: Incorreto. O menor caminho é mais curto do que 4 paradas.
- **d) 5**: Incorreto. O menor caminho é significativamente mais curto do que 5 paradas.
- **e) Nenhuma das anteriores**: Incorreto. A resposta correta é a opção **b**.

## Questão 3: Algoritmo de Dijkstra para Caminhos Mínimos

Uma empresa de entregas deseja calcular o menor custo de transporte entre o depósito principal (vértice A) e os demais centros de distribuição:

- Centros: {A, B, C, D, E}
- Custos:
  1. A - B: 3  
  2. A - C: 6  
  3. B - C: 2  
  4. B - D: 4  
  5. C - D: 1  
  6. C - E: 5  
  7. D - E: 2  

Qual é o menor custo para ir do depósito principal (A) ao centro E?

### Opções:
a) 7  
b) 8  
c) 9  
d) 10  
e) Nenhuma das anteriores

```python
# filepath: d:\GitHub\NewtonPaiva\TeoriaGrafos\BancoPerguntas\QuestoesContextualizadas.ipynb
# Criando o grafo
G = nx.Graph()
edges = [
    ("A", "B", 3),
    ("A", "C", 6),
    ("B", "C", 2),
    ("B", "D", 4),
    ("C", "D", 1),
    ("C", "E", 5),
    ("D", "E", 2),
]
G.add_weighted_edges_from(edges)

# Calculando o menor caminho usando o algoritmo de Dijkstra
shortest_path_length, shortest_path = nx.single_source_dijkstra(G, source="A", target="E")

# Exibindo o menor custo e o caminho correspondente
print(f"Menor custo de A para E: {shortest_path_length}")
print(f"Caminho correspondente: {' → '.join(shortest_path)}")

# Visualizando o grafo
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=700, font_size=14)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=12)
plt.title("Grafo Representando os Centros de Distribuição")
plt.show()
```

### Resposta Correta:
a) 7

### Feedback:
- **a) 7**: Correto. O menor custo de A para E é 7, passando pelas arestas A → B → D → E.
- **b) 8**: Incorreto. Este valor é maior que o menor custo calculado.
- **c) 9**: Incorreto. Este valor é maior que o menor custo calculado.
- **d) 10**: Incorreto. Este valor é maior que o menor custo calculado.
- **e) Nenhuma das anteriores**: Incorreto. A resposta correta é a opção **a**.

## Questão 4: Análise de Trajetos Hamiltonianos e Eulerianos

Considere o grafo abaixo, onde os vértices representam cidades e as arestas representam estradas entre elas. O objetivo é analisar as propriedades do grafo em relação aos trajetos Hamiltonianos e Eulerianos.

### Grafo:
- **Vértices (Cidades):** {A, B, C, D, E, F}
- **Arestas (Estradas e seus pesos):**
  1. A - B: 4  
  2. A - C: 6  
  3. B - C: 3  
  4. B - D: 5  
  5. C - D: 2  
  6. C - E: 7  
  7. D - E: 4  
  8. D - F: 6  
  9. E - F: 5  

### Pergunta:
Com base nas propriedades do grafo, analise as condições para trajetos Hamiltonianos e Eulerianos e escolha a alternativa correta:

### Opções:
a) O grafo possui um trajeto Euleriano, pois todos os vértices possuem grau par.  
b) O grafo possui um trajeto Hamiltoniano, pois é possível visitar todos os vértices exatamente uma vez.  
c) O grafo possui tanto um trajeto Euleriano quanto Hamiltoniano, pois atende às condições de ambos.  
d) O grafo não possui trajeto Euleriano, mas possui um trajeto Hamiltoniano.  
e) Nenhuma das anteriores

```python
# filepath: d:\GitHub\NewtonPaiva\TeoriaGrafos\BancoPerguntas\QuestoesContextualizadas.ipynb
# Criando o grafo
G = nx.Graph()
edges = [
    ("A", "B", 4),
    ("A", "C", 6),
    ("B", "C", 3),
    ("B", "D", 5),
    ("C", "D", 2),
    ("C", "E", 7),
    ("D", "E", 4),
    ("D", "F", 6),
    ("E", "F", 5),
]
G.add_weighted_edges_from(edges)

# Verificando propriedades
is_eulerian = nx.is_eulerian(G)
has_hamiltonian = nx.is_connected(G) and all(len(G.edges(v)) >= len(G.nodes) - 1 for v in G.nodes)

# Exibindo os resultados
print(f"O grafo possui trajeto Euleriano? {'Sim' if is_eulerian else 'Não'}")
print(f"O grafo possui trajeto Hamiltoniano? {'Sim' if has_hamiltonian else 'Não'}")

# Explicação detalhada
if not is_eulerian:
    print("O grafo não possui trajeto Euleriano porque existem vértices com grau ímpar, o que viola a condição necessária para um trajeto Euleriano.")
else:
    print("O grafo possui trajeto Euleriano porque todos os vértices possuem grau par.")

if has_hamiltonian:
    print("O grafo possui trajeto Hamiltoniano porque é conexo e permite visitar todos os vértices exatamente uma vez.")
else:
    print("O grafo não possui trajeto Hamiltoniano porque não atende às condições necessárias para tal trajeto.")

# Visualizando o grafo
pos = nx.spring_layout(G, seed=42)
plt.figure(figsize=(8, 6))
nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=700, font_size=14)
nx.draw_networkx_edge_labels(G, pos, edge_labels=nx.get_edge_attributes(G, "weight"), font_size=12)
plt.title("Grafo Representando as Cidades e Estradas")
plt.show()
```

### Resposta Correta:
b) O grafo possui um trajeto Hamiltoniano.

### Feedback:
- **a)** Incorreto. O grafo não possui um trajeto Euleriano, pois existem vértices com grau ímpar.
- **b)** Correto. O grafo possui um trajeto Hamiltoniano, pois é conexo e permite visitar todos os vértices exatamente uma vez.
- **c)** Incorreto. O grafo não possui um trajeto Euleriano.
- **d)** Incorreto. O grafo possui um trajeto Hamiltoniano.
- **e)** Incorreto. A resposta correta é a opção **b**.
