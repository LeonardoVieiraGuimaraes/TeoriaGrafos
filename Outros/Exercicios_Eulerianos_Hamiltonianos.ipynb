## 1. Verifique se os grafos abaixo s√£o eulerianos e/ou hamiltonianos. Justifique sua resposta encontrando ciclos eulerianos e/ou hamiltonianos nos grafos.

### Exemplo (Grafo 1):

- **V√©rtices:** a, b, c, d, e
- **Arestas:** (a,b), (b,c), (c,d), (d,e), (e,a), (b,e), (c,e)

#### Solu√ß√£o:

- **Grafo Euleriano?** N√£o, pois h√° mais de dois v√©rtices de grau √≠mpar.
- **Grafo Hamiltoniano?** Sim, pois existe um ciclo que passa por todos os v√©rtices uma √∫nica vez.

**Ciclo Hamiltoniano:** a ‚Üí b ‚Üí c ‚Üí d ‚Üí e ‚Üí a

**Ciclo Euleriano:** N√£o existe.


```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([('a','b'), ('b','c'), ('c','d'), ('d','e'), ('e','a'), ('b','e'), ('c','e')])
plt.figure(figsize=(5,4))
nx.draw(G, with_labels=True, node_color="#b2dfdb", node_size=800, font_weight='bold')
plt.title("Grafo 1")
plt.show()

# Verifica√ß√£o autom√°tica para o Grafo 1 (Exemplo)
grafo1 = G  # j√° criado anteriormente

# Euleriano: todos os v√©rtices t√™m grau par
def is_eulerian(g):
    return nx.is_eulerian(g)

# Hamiltoniano: n√£o h√° fun√ß√£o pronta no networkx, mas podemos tentar todas as permuta√ß√µes para grafos pequenos
from itertools import permutations

def is_hamiltonian(g):
    nodes = list(g.nodes)
    n = len(nodes)
    for path in permutations(nodes):
        if all(g.has_edge(path[i], path[i+1]) for i in range(n-1)) and g.has_edge(path[-1], path[0]):
            return True
    return False

print("Grafo 1 √© euleriano?", is_eulerian(grafo1))
print("Grafo 1 √© hamiltoniano?", is_hamiltonian(grafo1))
```
