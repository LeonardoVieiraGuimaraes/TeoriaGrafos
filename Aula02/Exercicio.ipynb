# Exercícios de Teoria dos Grafos

## 1. **O que é um grafo? Responda com suas próprias palavras.**

Um **grafo** é uma estrutura matemática composta por **vértices** (ou nós) e **arestas** (ou arcos) que conectam pares de vértices. Grafos são usados para modelar relações entre objetos, onde os vértices representam os objetos e as arestas representam as conexões entre eles.

---

## 2. **Para que serve um grafo?**

Grafos são usados para:
- Modelar redes (redes sociais, redes de computadores, redes de transporte).
- Resolver problemas de caminhos mínimos (como no GPS).
- Representar relações em bancos de dados.
- Analisar conectividade e fluxo em sistemas.
- Estudar estruturas em química, biologia e outras ciências.

## 4. **Classifique os grafos da questão anterior como simples ou multigrafos.**

- **Simples**: Grafos sem arestas múltiplas ou laços.
  - Exemplo: (c), (d), (e).
- **Multigrafos**: Grafos com arestas múltiplas ou laços.
  - Exemplo: (a), (b), (f).

---

## 5. **Quais são as diferenças entre grafos simples e multigrafos?**

- **Grafo Simples**:
  - Não possui arestas múltiplas (mais de uma aresta conectando o mesmo par de vértices).
  - Não possui laços (arestas que conectam um vértice a ele mesmo).

- **Multigrafo**:
  - Pode ter arestas múltiplas.
  - Pode ter laços.

## 21. **O que é um grafo completo?**

Um **grafo completo** é um grafo simples onde cada par de vértices distintos está conectado por uma aresta. Denotado por \( K_n \), onde \( n \) é o número de vértices.

---

## 22. **O que é um grafo bipartido?**

Um **grafo bipartido** é um grafo cujos vértices podem ser divididos em dois conjuntos disjuntos \( U \) e \( V \), de modo que toda aresta conecta um vértice de \( U \) a um vértice de \( V \).

---

## 23. **O que é um grafo bipartido completo?**

Um **grafo bipartido completo** é um grafo bipartido onde todo vértice de \( U \) está conectado a todo vértice de \( V \). Denotado por \( K_{r,s} \), onde \( r \) e \( s \) são os tamanhos de \( U \) e \( V \), respectivamente.

---

## 24. **Sob que circunstâncias devemos usar um grafo dirigido ao invés de um grafo não dirigido?**

Use um **grafo dirigido** quando as relações entre os vértices têm direção (por exemplo, ruas de mão única, dependências em tarefas).

---

## 26. **Quantas arestas tem um grafo nulo \( N_n \)?**

Um grafo nulo tem **0 arestas**.

---

## 27. **Quantas arestas tem um grafo ciclo \( C_n \)?**

Um grafo ciclo tem **\( n \) arestas**.

---

## 28. **Quantas arestas tem um grafo bipartido completo \( K_{r,s} \)?**

Um grafo bipartido completo tem **\( r \times s \) arestas**.

---

## 29. **Quantas arestas tem um grafo completo \( K_n \)?**

Um grafo completo tem **\( \frac{n(n-1)}{2} \) arestas**.

---

## 30. **Qual é o grafo simples mais denso que existe? Justifique sua resposta.**

O **grafo completo \( K_n \)** é o mais denso, pois contém o número máximo de arestas possível para um grafo simples com \( n \) vértices.
