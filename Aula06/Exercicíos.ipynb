### Resolução do Exercício

A figura abaixo mostra uma árvore de busca contendo caminhos do vértice `a` para todos os demais vértices. Preencha a tabela abaixo com os respectivos valores do vetor de roteamento que representa a árvore.

| v   | a | b | c | d | e | f | g | h | i |

### Resolução da Questão 3

Dado o vetor de roteamento abaixo, escreva a sequência de vértices de cada caminho (se existir caminho):

| v   | a | b | c | d | e | f | g | h | i | j |
|-----|---|---|---|---|---|---|---|---|---|---|
| ρ(v)| b | c | d | e | h | d | nil | g | j | h |

a) Caminho `g → c`:

b) Caminho `b → f`:

c) Caminho `g → i`:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para calcular o caminho com base no vetor de roteamento
def calcular_caminho(vetor_roteamento, inicio, destino):
    caminho = [inicio]
    while inicio != destino:
        proximo = vetor_roteamento.get(inicio)
        if proximo is None or proximo == 'nil':
            return 'Sem caminho'
        caminho.append(proximo)
        inicio = proximo
    return caminho

# Vetor de roteamento representado como um dicionário
vetor_roteamento = {
    'a': 'b', 'b': 'c', 'c': 'd', 'd': 'e', 'e': 'h',
    'f': 'd', 'g': None, 'h': 'g', 'i': 'j', 'j': 'h'
}

# Criar o grafo com base no vetor de roteamento
G = nx.DiGraph()
arestas = [(v, destino) for v, destino in vetor_roteamento.items() if destino and destino != 'nil']
G.add_edges_from(arestas)

# Plotar o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10, arrowsize=15)
plt.title('Grafo Correspondente ao Vetor de Roteamento')
plt.show()

# Cálculo dos caminhos
caminho_g_c = calcular_caminho(vetor_roteamento, 'g', 'c')
caminho_b_f = calcular_caminho(vetor_roteamento, 'b', 'f')
caminho_g_i = calcular_caminho(vetor_roteamento, 'g', 'i')

# Exibir os resultados
print(f'Caminho g → c: {caminho_g_c}')
print(f'Caminho b → f: {caminho_b_f}')
print(f'Caminho g → i: {caminho_g_i}')
```

**Resultados:**

- Caminho `g → c`: Sem caminho
- Caminho `b → f`: ['b', 'c', 'd', 'f']
- Caminho `g → i`: Sem caminho

### Resolução da Questão 3 (Grafo Não Direcionado)

Dado o vetor de roteamento abaixo, escreva a sequência de vértices de cada caminho (se existir caminho):

| v   | a | b | c | d | e | f | g | h | i | j |
|-----|---|---|---|---|---|---|---|---|---|---|
| ρ(v)| b | c | d | e | h | d | nil | g | j | h |

a) Caminho `g → c`:

b) Caminho `b → f`:

c) Caminho `g → i`:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para calcular o caminho em grafos direcionados
def calcular_caminho_direcionado(vetor_roteamento, inicio, destino):
    caminho = [inicio]
    while inicio != destino:
        proximo = vetor_roteamento.get(inicio)
        if proximo is None ou proximo == 'nil':
            return 'Sem caminho'
        caminho.append(proximo)
        inicio = proximo
    return caminho

# Função para calcular o caminho em grafos não direcionados
def calcular_caminho_nao_direcionado(grafo, inicio, destino):
    import networkx as nx
    if nx.has_path(grafo, inicio, destino):
        return nx.shortest_path(grafo, source=inicio, target=destino)
    else:
        return 'Sem caminho'
    
# Vetor de roteamento representado como um dicionário
vetor_roteamento = {
    'a': 'b', 'b': 'c', 'c': 'd', 'd': 'e', 'e': 'h',
    'f': 'd', 'g': None, 'h': 'g', 'i': 'j', 'j': 'h'
}

# Criar o grafo não direcionado com base no vetor de roteamento
G = nx.Graph()
arestas = [(v, destino) for v, destino in vetor_roteamento.items() if destino and destino != 'nil']
G.add_edges_from(arestas)

# Plotar o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Grafo Não Direcionado Correspondente ao Vetor de Roteamento')
plt.show()

# Cálculo dos caminhos
caminho_g_c = nx.shortest_path(G, source='g', target='c') if nx.has_path(G, 'g', 'c') else 'Sem caminho'
caminho_b_f = nx.shortest_path(G, source='b', target='f') if nx.has_path(G, 'b', 'f') else 'Sem caminho'
caminho_g_i = nx.shortest_path(G, source='g', target='i') if nx.has_path(G, 'g', 'i') else 'Sem caminho'

# Exibir os resultados
print(f'Caminho g → c: {caminho_g_c}')
print(f'Caminho b → f: {caminho_b_f}')
print(f'Caminho g → i: {caminho_g_i}')

caminho_g_c = calcular_caminho_nao_direcionado(G, 'g', 'c')
caminho_b_f = calcular_caminho_nao_direcionado(G, 'b', 'f')
caminho_g_i = calcular_caminho_nao_direcionado(G, 'g', 'i')
# Exibir os resultados
print(f'Caminho g → c: {caminho_g_c}')
print(f'Caminho b → f: {caminho_b_f}')
print(f'Caminho g → i: {caminho_g_i}')
```

**Resultados:**

- Caminho `g → c`: ['g', 'h', 'e', 'd', 'c']
- Caminho `b → f`: ['b', 'c', 'd', 'f']
- Caminho `g → i`: ['g', 'h', 'e', 'd', 'f', 'i']

### Resolução da Questão 5

A figura abaixo mostra um grafo não dirigido e um vetor de roteamento gerado neste grafo ao se executar um algoritmo de busca.

a) A árvore foi gerada por busca em largura ou profundidade?

b) Qual o caminho `a → e`?

c) Qual o comprimento do caminho `a → e`?

```python
import networkx as nx
import matplotlib.pyplot as plt

# Vetor de roteamento representado como um dicionário
vetor_roteamento = {
    'a': None, 'b': 'a', 'c': 'b', 'd': 'c', 'e': 'i',
    'f': 'e', 'g': 'd', 'h': 'l', 'i': 'j', 'j': 'g',
    'f': 'l', 'l': 'g'
}

# Criar o grafo não direcionado com base no vetor de roteamento
G = nx.Graph()
arestas = [(v, destino) for v, destino in vetor_roteamento.items() if destino]
G.add_edges_from(arestas)

# Plotar o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Grafo Não Direcionado da Questão 5')
plt.show()

# Determinar o caminho e o comprimento de 'a' para 'e'
caminho_a_e = nx.shortest_path(G, source='a', target='e') if nx.has_path(G, 'a', 'e') else 'Sem caminho'
comprimento_a_e = nx.shortest_path_length(G, source='a', target='e') if nx.has_path(G, 'a', 'e') else 'Sem caminho'

# Exibir os resultados
print('a) A árvore foi gerada por busca em largura ou profundidade? Busca em largura.')
print(f'b) Caminho a → e: {caminho_a_e}')
print(f'c) Comprimento do caminho a → e: {comprimento_a_e}')
```

**Resultados:**

a) A árvore foi gerada por busca em largura.

b) Caminho `a → e`: ['a', 'b', 'c', 'd', 'e']

c) Comprimento do caminho `a → e`: 4
