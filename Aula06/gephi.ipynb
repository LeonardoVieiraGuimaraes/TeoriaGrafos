# Exemplo de Grafo para Gephi
Este exemplo demonstra como criar um grafo simples usando a biblioteca `networkx` e exportá-lo para o formato `.gexf`, que pode ser importado no Gephi.

## Estudo de Caso: Rede de Colaboração Acadêmica

Imagine que você está analisando uma rede de colaboração acadêmica entre pesquisadores. Cada nó no grafo representa um pesquisador, e cada aresta representa uma colaboração em um artigo científico.

### Objetivos:
- Identificar os pesquisadores mais influentes na rede.
- Visualizar comunidades de colaboração.
- Exportar os dados para o Gephi para análise avançada.

### Tarefas:
1. Criar um grafo representando a rede de colaboração.
2. Adicionar atributos aos nós, como o nome do pesquisador.
3. Adicionar pesos às arestas, representando o número de colaborações.
4. Exportar o grafo para o formato `.gexf` e analisá-lo no Gephi.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criação do grafo
G = nx.Graph()

# Adicionando nós
G.add_node(1, label='Node 1')
G.add_node(2, label='Node 2')
G.add_node(3, label='Node 3')

# Adicionando arestas
G.add_edge(1, 2, weight=1.0)
G.add_edge(2, 3, weight=2.0)
G.add_edge(3, 1, weight=3.0)

# Plotando o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
plt.show()

# Exportando para o formato GEXF
nx.write_gexf(G, 'grafo_exemplo.gexf')
print('Grafo exportado para o arquivo grafo_exemplo.gexf')
```

## Passo a Passo para Análise no Gephi

1. **Importar o Arquivo GEXF**:
   - Abra o Gephi.
   - No menu principal, clique em `File > Open` e selecione o arquivo `grafo_exemplo.gexf` exportado anteriormente.

2. **Visualizar o Grafo**:
   - Após a importação, o grafo será exibido na aba `Overview`.
   - Use as ferramentas de navegação para ajustar o zoom e posicionar o grafo.

3. **Calcular Métricas de Centralidade**:
   - Vá para a aba `Statistics`.
   - Execute métricas como `Degree`, `Betweenness Centrality` ou `Closeness Centrality` para identificar os nós mais influentes.

4. **Detectar Comunidades**:
   - Na aba `Statistics`, execute o algoritmo de `Modularity` para identificar comunidades.
   - Atribua cores aos nós com base nas comunidades detectadas.

5. **Personalizar a Visualização**:
   - Na aba `Appearance`, ajuste as cores, tamanhos e rótulos dos nós e arestas.
   - Use layouts como `ForceAtlas` ou `Yifan Hu` para reorganizar o grafo.

6. **Exportar os Resultados**:
   - Após finalizar a análise, exporte o grafo processado clicando em `File > Export`.
   - Escolha o formato desejado, como PDF, PNG ou um novo arquivo GEXF.

Com esses passos, você poderá explorar e analisar a rede de colaboração acadêmica de forma avançada no Gephi.

## Estudo de Caso Simples: Grafo Básico

Neste exemplo, criaremos um grafo básico com poucos nós e arestas. Em seguida, calcularemos:
- O número de vértices (nós).
- O número de arestas.
- O grau de cada nó.

Este estudo de caso é ideal para entender os conceitos fundamentais de grafos.

```python
import networkx as nx

# Criando um grafo simples
G_simple = nx.Graph()

# Adicionando nós com rótulos em letras
G_simple.add_nodes_from(['A', 'B', 'C', 'D', 'E'])

# Adicionando arestas
G_simple.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E'), ('E', 'A')])

# Número de vértices e arestas
num_vertices = G_simple.number_of_nodes()
num_arestas = G_simple.number_of_edges()

# Graus de cada nó
graus = dict(G_simple.degree())

print(f'Número de vértices: {num_vertices}')
print(f'Número de arestas: {num_arestas}')
print(f'Graus dos nós: {graus}')
```

```python
# Imprimindo os dados do grafo para copiar
print('Nós:', list(G_simple.nodes()))
print('Arestas:', list(G_simple.edges()))

# Plotando o grafo simples
import matplotlib.pyplot as plt
pos_simple = nx.spring_layout(G_simple)
nx.draw(G_simple, pos_simple, with_labels=True, node_color='lightgreen', edge_color='blue', node_size=1500, font_size=12)
plt.show()
```

## Estudo de Caso: Grafo Complexo com Letras

Neste exemplo, criaremos um grafo mais complexo com nós rotulados por letras e múltiplas conexões. Em seguida, calcularemos:
- O número de vértices (nós).
- O número de arestas.
- O grau de cada nó.

### Cenário:
Imagine que você está analisando uma rede de comunicação entre departamentos de uma empresa. Cada nó representa um departamento, e cada aresta representa uma colaboração entre eles.

### Como fazer isso no Gephi:
1. **Importar o Arquivo GEXF**:
   - Exporte o grafo para o formato `.gexf` e abra no Gephi.

2. **Calcular Métricas**:
   - Vá para a aba `Statistics` e execute `Degree` para calcular o grau de cada nó.

3. **Visualizar Resultados**:
   - Use a aba `Data Laboratory` para visualizar o número de vértices e arestas.
   - Na aba `Overview`, ajuste o tamanho dos nós com base no grau para destacar os mais conectados.

```python
import networkx as nx

# Criando um grafo complexo
G_complex = nx.Graph()

# Adicionando nós com rótulos em letras
G_complex.add_nodes_from(['A', 'B', 'C', 'D', 'E', 'F', 'G', 'H'])

# Adicionando arestas
G_complex.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'),
    ('D', 'E'), ('E', 'F'), ('F', 'G'), ('G', 'H'), ('H', 'A')
])

# Número de vértices e arestas
num_vertices = G_complex.number_of_nodes()
num_arestas = G_complex.number_of_edges()

# Graus de cada nó
graus = dict(G_complex.degree())

print(f'Número de vértices: {num_vertices}')
print(f'Número de arestas: {num_arestas}')
print(f'Graus dos nós: {graus}')

# Exportando para o formato GEXF
nx.write_gexf(G_complex, 'grafo_complexo.gexf')
print('Grafo exportado para o arquivo grafo_complexo.gexf')
```

```python
# Imprimindo os dados do grafo para copiar
print('Nós:', list(G_complex.nodes()))
print('Arestas:', list(G_complex.edges()))

# Plotando o grafo complexo
import matplotlib.pyplot as plt
pos_complex = nx.spring_layout(G_complex)
nx.draw(G_complex, pos_complex, with_labels=True, node_color='lightblue', edge_color='gray', node_size=1500, font_size=12)
plt.show()
```

## Observação Importante sobre Estatísticas Dinâmicas

No Gephi, estatísticas dinâmicas, como evolução temporal de métricas, só podem ser calculadas se o grafo for dinâmico.

### O que é um Grafo Dinâmico?
Um grafo dinâmico é aquele que possui informações temporais associadas aos nós e arestas, como intervalos de tempo em que as conexões existem.

### Como Criar um Grafo Dinâmico?
Para criar um grafo dinâmico, é necessário adicionar atributos temporais aos nós e arestas durante a construção do grafo. No formato `.gexf`, isso pode ser feito utilizando as propriedades de tempo (`start` e `end`).

Se o grafo não for dinâmico, apenas estatísticas estáticas, como grau, modularidade e centralidade, estarão disponíveis.

## Exemplo de Grafo Dinâmico

Neste exemplo, criaremos um grafo dinâmico com informações temporais associadas aos nós e arestas. Isso permitirá a análise de estatísticas dinâmicas no Gephi.

### Como Criar um Grafo Dinâmico no Gephi:
1. **Adicionar Informações Temporais**:
   - Cada nó e aresta deve conter atributos de tempo, como `start` e `end`.

2. **Importar no Gephi**:
   - Exporte o grafo no formato `.gexf` e abra no Gephi.

3. **Analisar Estatísticas Dinâmicas**:
   - Use a aba `Timeline` para ajustar o intervalo de tempo e visualizar as mudanças na rede.

Com isso, será possível explorar a evolução temporal da rede.

```python
import networkx as nx

# Criando um grafo dinâmico
G_dynamic = nx.Graph()

# Adicionando nós com atributos temporais
G_dynamic.add_node('A', start=0, end=5)
G_dynamic.add_node('B', start=1, end=6)
G_dynamic.add_node('C', start=2, end=7)
G_dynamic.add_node('D', start=3, end=8)

# Adicionando arestas com atributos temporais
G_dynamic.add_edge('A', 'B', start=1, end=4)
G_dynamic.add_edge('B', 'C', start=2, end=5)
G_dynamic.add_edge('C', 'D', start=3, end=6)
G_dynamic.add_edge('D', 'A', start=4, end=7)

# Número de vértices e arestas
num_vertices = G_dynamic.number_of_nodes()
num_arestas = G_dynamic.number_of_edges()

# Graus de cada nó
graus = dict(G_dynamic.degree())

print(f'Número de vértices: {num_vertices}')
print(f'Número de arestas: {num_arestas}')
print(f'Graus dos nós: {graus}')

# Exportando para o formato GEXF com informações dinâmicas
nx.write_gexf(G_dynamic, 'grafo_dinamico.gexf')
print('Grafo dinâmico exportado para o arquivo grafo_dinamico.gexf')
```

## O que são Informações Temporais em um Grafo Dinâmico?

As informações temporais em um grafo dinâmico representam os intervalos de tempo em que os nós e arestas estão ativos. Isso permite analisar como a estrutura do grafo muda ao longo do tempo.

### Exemplos de Informações Temporais:
- **Nós**: Um nó pode ter atributos `start` e `end` que indicam o período em que ele está presente na rede.
- **Arestas**: Uma aresta pode ter atributos `start` e `end` que indicam o período em que a conexão entre dois nós existe.

### Aplicações:
- **Redes Sociais**: Analisar como as conexões entre pessoas mudam ao longo do tempo.
- **Redes de Comunicação**: Estudar a evolução de interações entre dispositivos ou sistemas.
- **Redes de Colaboração**: Identificar períodos de colaboração ativa entre pesquisadores ou equipes.

### Como Usar no Gephi:
1. Certifique-se de que os atributos `start` e `end` estão presentes nos nós e arestas ao exportar o grafo.
2. No Gephi, use a aba `Timeline` para ajustar o intervalo de tempo e visualizar as mudanças na rede.

Com essas informações, é possível realizar análises dinâmicas e entender a evolução temporal da rede.
