## üí° Defini√ß√µes Matem√°ticas

- **Grafo:** $G = (V, E)$, onde $V$ √© o conjunto de v√©rtices e $E$ o conjunto de arestas.
- **Caminho Hamiltoniano:** Um caminho simples em $G$ que passa por todos os v√©rtices exatamente uma vez. Formalmente, uma sequ√™ncia de v√©rtices $(v_1, v_2, ..., v_n)$ tal que $\forall i, (v_i, v_{i+1}) \in E$ e $\{v_1, ..., v_n\} = V$.
- **Ciclo Hamiltoniano:** Um caminho hamiltoniano que √© um ciclo, ou seja, al√©m de passar por todos os v√©rtices uma √∫nica vez, retorna ao v√©rtice inicial: $(v_1, v_2, ..., v_n, v_1)$.

**Observa√ß√£o:**  
Nem todo grafo possui caminho ou ciclo hamiltoniano. Determinar se um grafo √© hamiltoniano √© um problema dif√≠cil (NP-completo).

## üï∏Ô∏è Exemplo Visual

Considere o grafo abaixo:

- **V√©rtices:** A, B, C, D, E
- **Arestas:** (A, B), (A, C), (B, C), (B, D), (C, D), (C, E), (D, E)

Vamos visualizar esse grafo:

````python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.Graph()
G.add_edges_from([
    ('A', 'B'), ('A', 'C'), ('B', 'C'),
    ('B', 'D'), ('C', 'D'), ('C', 'E'), ('D', 'E')
])

pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', font_weight='bold')
plt.title("Exemplo de Grafo para Caminho/Ciclo Hamiltoniano")
plt.show()
print("Posi√ß√µes dos v√©rtices:", pos)
````

## üîé Como Identificar Caminhos e Ciclos Hamiltonianos

- **Caminho Hamiltoniano:** Exemplo neste grafo: A ‚Üí B ‚Üí D ‚Üí E ‚Üí C
- **Ciclo Hamiltoniano:** Exemplo: A ‚Üí B ‚Üí D ‚Üí E ‚Üí C ‚Üí A

**Dificuldade:**  
N√£o existe um algoritmo eficiente conhecido para encontrar caminhos/ciclos hamiltonianos em grafos arbitr√°rios. O problema √© NP-completo.

## üßÆ Propriedades e Teoremas

- **Teorema de Dirac:** Se $G$ √© um grafo simples com $n \geq 3$ v√©rtices e grau m√≠nimo de cada v√©rtice $\geq n/2$, ent√£o $G$ √© hamiltoniano.
- **Teorema de Ore:** Se para todo par de v√©rtices n√£o adjacentes $u, v$ em $G$, $deg(u) + deg(v) \geq n$, ent√£o $G$ √© hamiltoniano.

**Aten√ß√£o:**  
Essas condi√ß√µes s√£o suficientes, mas n√£o necess√°rias. Ou seja, existem grafos hamiltonianos que n√£o satisfazem esses teoremas.

## üìù Pseudoc√≥digo Gen√©rico para Busca de Caminho/Ciclo Hamiltoniano

```plaintext
Hamiltoniano(Grafo G, Caminho atual)
    se Caminho atual cont√©m todos os v√©rtices:
        se existe aresta do √∫ltimo v√©rtice ao primeiro:
            retorna "Ciclo Hamiltoniano encontrado"
        sen√£o:
            retorna "Caminho Hamiltoniano encontrado"
    para cada vizinho v do √∫ltimo v√©rtice em Caminho atual:
        se v n√£o est√° em Caminho atual:
            chama Hamiltoniano(G, Caminho atual + v)
```

## üèÜ Exerc√≠cio Pr√°tico: Problema do Caixeiro Viajante

Considere o seguinte grafo ponderado (dist√¢ncias entre cidades):

- **V√©rtices:** A, B, C, D
- **Arestas e pesos:**  
  (A, B): 10  
  (A, C): 15  
  (A, D): 20  
  (B, C): 35  
  (B, D): 25  
  (C, D): 30  

**Tarefa:**  
1. Liste todos os ciclos hamiltonianos poss√≠veis e calcule o custo total de cada um.
2. Qual √© o ciclo de menor custo?  
3. Implemente em Python uma busca exaustiva para encontrar o ciclo hamiltoniano de menor custo neste grafo.

````python
# Implementa√ß√£o do TSP (busca exaustiva) para o exerc√≠cio proposto

import itertools

# Definindo o grafo ponderado
G_tsp = nx.Graph()
arestas = [
    ('A', 'B', 10), ('A', 'C', 15), ('A', 'D', 20),
    ('B', 'C', 35), ('B', 'D', 25),
    ('C', 'D', 30)
]
G_tsp.add_weighted_edges_from(arestas)

# Gerando todos os ciclos hamiltonianos poss√≠veis (come√ßando em 'A')
vertices = list(G_tsp.nodes())
start = 'A'
outros = [v for v in vertices if v != start]
melhor_ciclo = None
menor_custo = float('inf')

print("Ciclos Hamiltonianos e seus custos:")

for perm in itertools.permutations(outros):
    ciclo = [start] + list(perm) + [start]
    custo = 0
    valido = True
    for i in range(len(ciclo)-1):
        u, v = ciclo[i], ciclo[i+1]
        if G_tsp.has_edge(u, v):
            custo += G_tsp[u][v]['weight']
        else:
            valido = False
            break
    if valido:
        print(f"{' ‚Üí '.join(ciclo)} | Custo: {custo}")
        # Plotando o ciclo atual
        plt.figure(figsize=(4,3))
        pos_tsp = nx.spring_layout(G_tsp, seed=1)
        edge_colors = []
        for edge in G_tsp.edges():
            if all(e in zip(ciclo, ciclo[1:]) or e[::-1] in zip(ciclo, ciclo[1:]) for e in [edge]):
                edge_colors.append('red')
            else:
                edge_colors.append('gray')
        nx.draw(G_tsp, pos_tsp, with_labels=True, node_color='lightyellow', edge_color=edge_colors, width=2)
        nx.draw_networkx_edge_labels(G_tsp, pos_tsp, edge_labels=nx.get_edge_attributes(G_tsp, 'weight'))
        plt.title(f"Ciclo: {' ‚Üí '.join(ciclo)} | Custo: {custo}")
        plt.show()
        print("Posi√ß√µes dos v√©rtices:", pos_tsp)
        if custo < menor_custo:
            menor_custo = custo
            melhor_ciclo = ciclo

print("\nMelhor ciclo encontrado:", " ‚Üí ".join(melhor_ciclo), f"| Custo: {menor_custo}")
````
