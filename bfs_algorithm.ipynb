## Estudo de Caso Real: Rede de Transporte Público


Vamos considerar um estudo de caso real onde utilizamos a Teoria dos Grafos para modelar e analisar uma rede de transporte público. Suponha que queremos encontrar o caminho mais curto entre duas estações de metrô em uma cidade.

## Criação do Grafo

Vamos criar um grafo que representa a rede de transporte público. Cada nó representa uma estação de metrô e cada aresta representa uma conexão direta entre duas estações.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criação do grafo
G = nx.Graph()

# Adicionando nós e arestas (estações e conexões)
G.add_edges_from([
    ("Estação A", "Estação B"),
    ("Estação A", "Estação C"),
    ("Estação B", "Estação D"),
    ("Estação C", "Estação D"),
    ("Estação D", "Estação E"),
    ("Estação E", "Estação F"),
    ("Estação F", "Estação G"),
    ("Estação G", "Estação H"),
    ("Estação H", "Estação I"),
    ("Estação I", "Estação J"),
    ("Estação A", "Estação k"),
    ("Estação B", "Estação L"),
    ("Estação C", "Estação M"),
    ("Estação D", "Estação N"),
    ("Estação E", "Estação O"),
    ("Estação F", "Estação P"),
    ("Estação G", "Estação Q"),
    ("Estação H", "Estação R"),
     ("Estação L", "Estação J"),
    
 
 
])

# Desenhando o grafo
plt.figure(figsize=(10, 8))
nx.draw(G, with_labels=True, node_size=2000, node_color="lightblue", font_size=10, font_weight="bold")
plt.show()
```

## Implementação do Algoritmo BFS

Vamos implementar o algoritmo de Busca em Largura (BFS) para encontrar o caminho mais curto entre duas estações de metrô.

BFS significa Breadth-First Search (Busca em Largura, em português). É um algoritmo de busca utilizado para percorrer ou buscar elementos em estruturas de dados como grafos ou árvores. O algoritmo começa em um nó raiz e explora todos os seus vizinhos de nível mais próximo antes de mover-se para os nós de nível seguinte.

```python
from collections import deque

def bfs(graph, start, goal):
    visited = set()
    queue = deque([(start, [start])])
    
    while queue:
        (vertex, path) = queue.popleft()
        for next in set(graph[vertex]) - set(path):
            if next == goal:
                return path + [next]
            else:
                queue.append((next, path + [next]))
    return None

# Executando o BFS para encontrar o caminho mais curto entre "Estação A" e "Estação J"
start_station = "Estação A"
goal_station = "Estação J"
shortest_path = bfs(G, start_station, goal_station)
print(f"Caminho mais curto entre {start_station} e {goal_station}: {shortest_path}")
```
