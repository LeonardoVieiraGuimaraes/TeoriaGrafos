# Lista de Adjacência

A **lista de adjacência** é uma estrutura de dados usada para representar grafos, onde cada vértice está associado a uma lista de seus vértices adjacentes. Essa representação é eficiente para grafos **esparsos** (com poucas arestas em relação ao número de vértices).

---

## 1. Definição

Dado um grafo $ G = (V, E) $, a lista de adjacência é um conjunto de listas, onde cada vértice $ v n V $ tem uma lista contendo todos os vértices $ u $ tais que $ v, u n E $ (ou $ (v, u) n E $, no caso de dígrafos).

### Representação:
- Para grafos **não direcionados**, cada aresta $ u, v $ é representada duas vezes: uma na lista de $ u $ e outra na lista de $ v $.
- Para grafos **direcionados**, cada arco $ (u, v) $ é representado apenas na lista de $ u $.

---

## 2. Exemplo de Lista de Adjacência

### Grafo Não Direcionado:
Considere o grafo $ G $ com:
- **Vértices**: $ V = 1, 2, 3 $.
- **Arestas**: $ E = 1, 2, 2, 3 $.

A lista de adjacência é:
```markdown
1: [2]
2: [1, 3]
3: [2]

````python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph_with_labels_outside_arestas(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    edge_labels = {e: f'e{idx}' for idx, e in enumerate(G.edges(), 1)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(title)
    plt.show()

# Definindo o grafo não direcionado
G = nx.Graph()
G.add_edges_from([(2, 1), (3, 2), (4, 3), (1, 4), (3, 1)])

# Plotando o grafo usando a função personalizada
draw_graph_with_labels_outside_arestas(G, "Grafo Não Direcionado")

# Obtendo a lista de adjacência
adj_list = {node: list(G[node]) for node in G.nodes()}
print("Lista de Adjacência:")
for node, edges in adj_list.items():
    print(f"{node}: {edges}")

# Função para exibir a lista de adjacência de forma legível
def print_adj_list(adj_list):
    for node, edges in adj_list.items():
        print(f"{node}: ", end="")
        for edge in edges:
            print(f"{edge} ", end="")
        print()

# Exibindo a lista de adjacência de forma legível
print("\nLista de Adjacência:")
print_adj_list(adj_list)

# Plotando a lista de adjacência
def plot_adj_list(adj_list):
    fig, ax = plt.subplots()
    ax.axis('off')
    table_data = []
    max_edges = max(len(edges) for edges in adj_list.values())
    for node, edges in adj_list.items():
        row = [f"{node}"] + [f"{edge}" for edge in edges]
        row += [""] * (max_edges - len(edges))  # Preencher com valores vazios
        table_data.append(row)
    col_labels = ["Vértice"] + [f"Adjacência {i+1}" for i in range(max_edges)]
    table = ax.table(cellText=table_data, colLabels=col_labels, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1.2, 1.2)
    plt.title("Lista de Adjacência")
    plt.show()

# Plotando a lista de adjacência
plot_adj_list(adj_list)
````

# Lista de Incidência

A **lista de incidência** é uma estrutura de dados usada para representar grafos, onde cada vértice está associado a uma lista de arestas incidentes a ele. Esta representação é útil para algumas operações específicas em grafos.

---

## 1. Definição

Dado um grafo $ G = (V, E) $, a lista de incidência é um conjunto de listas, onde cada vértice $ v \in V $ tem uma lista contendo todas as arestas $ e \in E $ tais que $ v $ é um dos extremos de $ e $.

### Representação:
- Para grafos **não direcionados**, cada aresta $ \{u, v\} $ é representada duas vezes: uma na lista de $ u $ e outra na lista de $ v $.
- Para grafos **direcionados**, cada arco $ (u, v) $ é representado apenas na lista de $ v $.

---

## 2. Exemplo de Lista de Incidência

### Grafo Direcionado:
Considere o grafo $ G $ com:
- **Vértices**: $ V = \{1, 2, 3, 4\} $.
- **Arestas**: $ E = \{(1, 2), (2, 3), (3, 4), (4, 1), (3, 1)\} $.

A lista de incidência é:

```markdown
1: [(4, 1)]
2: [(1, 2)]
3: [(2, 3), (4, 3)]
4: [(3, 4)]

````python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph_with_labels_outside_arestas(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    edge_labels = {e: f'e{idx}' for idx, e in enumerate(G.edges(), 1)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(title)
    plt.show()

# Definindo o grafo direcionado
G = nx.DiGraph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (3, 1)])

# Plotando o grafo usando a função personalizada
draw_graph_with_labels_outside_arestas(G, "Grafo Direcionado")

# Obtendo a lista de incidência
incidence_list = {node: list(G.in_edges(node)) for node in G.nodes()}
print("Lista de Incidência:")
for node, edges in incidence_list.items():
    print(f"{node}: {edges}")

# Função para exibir a lista de incidência de forma legível
def print_incidence_list(incidence_list):
    for node, edges in incidence_list.items():
        print(f"{node}: ", end="")
        for edge in edges:
            print(f"{edge} ", end="")
        print()

# Exibindo a lista de incidência de forma legível
print("\nLista de Incidência:")
print_incidence_list(incidence_list)

# Plotando a lista de incidência
def plot_incidence_list(incidence_list):
    fig, ax = plt.subplots()
    ax.axis('off')
    table_data = []
    max_edges = max(len(edges) for edges in incidence_list.values())
    for node, edges in incidence_list.items():
        row = [f"{node}"] + [f"{edge}" for edge in edges]
        row += [""] * (max_edges - len(edges))  # Preencher com valores vazios
        table_data.append(row)
    col_labels = ["Vértice"] + [f"Incidência {i+1}" for i in range(max_edges)]
    table = ax.table(cellText=table_data, colLabels=col_labels, loc='center', cellLoc='center')
    table.auto_set_font_size(False)
    table.set_fontsize(12)
    table.scale(1.2, 1.2)
    plt.title("Lista de Incidência")
    plt.show()

# Plotando a lista de incidência
plot_incidence_list(incidence_list)
````
