# Teoria dos Grafos
## Aula sobre Estrutura de Grafos

Nesta aula, vamos explorar os conceitos básicos da teoria dos grafos, incluindo tipos de grafos e como representá-los visualmente usando a biblioteca NetworkX do Python.

```python
# Importar a biblioteca NetworkX e Matplotlib
import networkx as nx
import matplotlib.pyplot as plt
```

### Conceito de Grafo
Um grafo é uma estrutura composta por vértices (ou nós) e arestas que conectam esses vértices. Vamos criar um grafo simples como exemplo.

**Definição Matemática:** Um grafo  G  é definido como um par  G = (V, E) , onde  V  é um conjunto de vértices e  E  é um conjunto de arestas que conectam pares de vértices.

```python
# Criar um grafo simples
G = nx.Graph()
G.add_edge('A', 'B')
G.add_edge('B', 'C')
G.add_edge('C', 'A')

# Plotar o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='black', font_color='white')
plt.show()
```

### Tipos de Grafos
Existem vários tipos de grafos, incluindo grafos direcionados, grafos ponderados, e grafos completos. Vamos ver exemplos de cada um.

#### Grafo Direcionado
Um grafo direcionado é um grafo onde as arestas têm uma direção. Vamos criar um exemplo de grafo direcionado.

**Definição Matemática:** Um grafo direcionado  G  é um par  G = (V, E) , onde  V  é um conjunto de vértices e  E  é um conjunto de arestas direcionadas, ou seja, pares ordenados de vértices.

```python
# Criar um grafo direcionado
DG = nx.DiGraph()
DG.add_edge('A', 'B')
DG.add_edge('B', 'C')
DG.add_edge('C', 'A')

# Plotar o grafo direcionado
pos = nx.spring_layout(DG)
nx.draw(DG, pos, with_labels=True, node_color='black', font_color='white', arrows=True)
plt.show()
```

#### Grafo Ponderado
Um grafo ponderado é um grafo onde as arestas têm pesos. Vamos criar um exemplo de grafo ponderado.

**Definição Matemática:** Um grafo ponderado  G  é um grafo onde cada aresta  e n E  tem um peso associado  w(e) .

```python
# Criar um grafo ponderado
WG = nx.Graph()
WG.add_edge('A', 'B', weight=4)
WG.add_edge('B', 'C', weight=2)
WG.add_edge('C', 'A', weight=5)

# Plotar o grafo ponderado
pos = nx.spring_layout(WG)
edges = WG.edges(data=True)
nx.draw(WG, pos, with_labels=True, node_color='black', font_color='white')
nx.draw_networkx_edge_labels(WG, pos, edge_labels={(u, v): d['weight'] for u, v, d in edges})
plt.show()
```

#### Grafo Completo
Um grafo completo é um grafo onde cada par de vértices está conectado por uma aresta. Vamos criar um exemplo de grafo completo.

**Definição Matemática:** Um grafo completo  K_n  é um grafo onde cada par de vértices  v_i, v_j n V  está conectado por uma aresta.

```python
# Criar um grafo completo
KG = nx.complete_graph(5)

# Plotar o grafo completo
pos = nx.spring_layout(KG)
nx.draw(KG, pos, with_labels=True, node_color='black', font_color='white')
plt.show()
```
