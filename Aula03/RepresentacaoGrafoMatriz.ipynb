As **matrizes de adjacência** e **matrizes de incidência** são duas formas de representar grafos de maneira estruturada. Cada uma tem suas vantagens e desvantagens, dependendo do tipo de grafo e da operação que se deseja realizar. Vamos detalhar cada uma delas:

---

### 1. **Matriz de Adjacência**

A **matriz de adjacência** é uma matriz quadrada que indica a existência de arestas (ou arcos) entre os vértices de um grafo.

#### Definição:
Dado um grafo $G = (V, E)$ com $n$ vértices, a matriz de adjacência $A$ é uma matriz $n \times n$, onde:
$
A[i][j] = 
\begin{cases} 
1 & \text{se existe uma aresta entre os vértices } i \text{ e } j, \\
0 & \text{caso contrário}.
\end{cases}
$

#### Propriedades:
- Para grafos **não direcionados**, a matriz de adjacência é **simétrica** ($A[i][j] = A[j][i]$).
- Para grafos **direcionados** (dígrafos), a matriz de adjacência não é necessariamente simétrica.
- Para grafos **ponderados**, $A[i][j]$ pode armazenar o peso da aresta em vez de apenas $1$.

#### Exemplo:
Considere o grafo $G$ com:
- **Vértices**: $V = \{1, 2, 3\}$.
- **Arestas**: $E = \{\{1, 2\}, \{2, 3\}\}$.

A matriz de adjacência $A$ é:
$
A = \begin{bmatrix}
0 & 1 & 0 \\
1 & 0 & 1 \\
0 & 1 & 0
\end{bmatrix}
$

---

### 2. **Matriz de Incidência**

A **matriz de incidência** é uma matriz que relaciona vértices e arestas de um grafo.

#### Definição:
Dado um grafo $G = (V, E)$ com $n$ vértices e $m$ arestas, a matriz de incidência $B$ é uma matriz $n \times m$, onde:
$
B[i][j] = 
\begin{cases} 
1 & \text{se o vértice } i \text{ é incidente à aresta } j, \\
0 & \text{caso contrário}.
\end{cases}
$

#### Propriedades:
- Para grafos **não direcionados**, cada aresta $j$ é representada por duas entradas $1$ na coluna $j$, correspondentes aos dois vértices incidentes.
- Para grafos **direcionados**, a matriz de incidência pode usar $-1$ para indicar a direção da aresta (vértice de origem) e $1$ para o vértice de destino.
- A matriz de incidência é menos eficiente em termos de espaço para grafos densos (com muitas arestas).

#### Exemplo:
Considere o grafo $G$ com:
- **Vértices**: $V = \{1, 2, 3\}$.
- **Arestas**: $E = \{e_1 = \{1, 2\}, e_2 = \{2, 3\}\}$.

A matriz de incidência $B$ é:
$
B = \begin{bmatrix}
1 & 0 \\
1 & 1 \\
0 & 1
\end{bmatrix}
$

---

### Comparação entre Matriz de Adjacência e Matriz de Incidência

| Propriedade               | Matriz de Adjacência          | Matriz de Incidência          |
|---------------------------|-------------------------------|-------------------------------|
| **Tamanho**               | $n \times n$              | $n \times m$              |
| **Uso de Memória**        | Eficiente para grafos densos  | Eficiente para grafos esparsos|
| **Facilidade de Uso**     | Simples para verificar adjacência | Útil para análise de arestas |
| **Grafos Direcionados**   | Simétrica (não direcionado)   | Pode usar $-1$ e $1$  |

---

### Resumo:
- A **matriz de adjacência** é ideal para verificar a existência de arestas entre vértices e é mais eficiente para grafos densos.
- A **matriz de incidência** é útil para analisar a relação entre vértices e arestas, sendo mais eficiente para grafos esparsos.

### Forma preferida de representar grafos densos
- A matriz de adjacência é a forma preferida de representar grafos densos (|E| ≈ |V|^2).
- Indica rapidamente (O(1)) se existe uma aresta conectando dois vértices.



```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph_with_labels_outside(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    plt.title(title)
    plt.show()


# Função para desenhar o grafo com bordas pretas, nós brancos e rótulos pretos fora das bolinhas
def draw_graph_with_labels_outside_q(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=400, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=16, font_color='black', font_family='sans-serif', font_weight='bold', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    plt.title(title)
#

# Função para desenhar o grafo com bordas pretas, nós brancos e rótulos pretos fora das bolinhas
def draw_graph_with_labels_outside_quad(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=400, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=16, font_color='black', font_family='sans-serif', font_weight='bold', bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.3'))
    plt.title(title)

def draw_graph_with_labels_outside_peso(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    edge_labels = nx.get_edge_attributes(G, 'weight')
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(title)
    plt.show()

# Função para desenhar o grafo com rótulos nas arestas e pesos
def draw_graph_with_labels_outside_arestas(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15, arrows=True)

    edge_labels = {(u, v): f"{u}->{v}" for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(titulo)
    plt.show()


# Função para desenhar o grafo com rótulos nas arestas e pesos
def draw_graph_with_labels_outside_arestas_pesos(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15, arrows=True)
    # Incluindo os pesos das arestas no rótulo
    edge_labels = {(u, v): f"{u}->{v} ({d['weight']})" for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(titulo)
    plt.show()

    
```

```python
# Definindo o dígrafo
DG = nx.DiGraph()
DG.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

# Plotando o dígrafo usando a função personalizada
draw_graph_with_labels_outside(DG, "Dígrafo")

# Obtendo a matriz de adjacência
adj_matrix = nx.adjacency_matrix(DG).todense()
print("Matriz de Adjacência:")
print(adj_matrix)

```

```python

# Definindo o grafo ponderado
G = nx.Graph()
G.add_weighted_edges_from([(1, 2, 2.5), (2, 3, 1.2), (3, 4, 3.8), (4, 1, 4.1), (1, 3, 2.0)])

# Plotando o grafo ponderado usando a função personalizada
draw_graph_with_labels_outside_arestas(G, "Grafo Ponderado")


# Obtendo a matriz de adjacência
adj_matrix = nx.adjacency_matrix(G).todense()
print("Matriz de Adjacência:")
print(adj_matrix)
```

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Definindo o grafo não ponderado
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)])

# Plotando o grafo usando a função personalizada
draw_graph_with_labels_outside_arestas(G, "Grafo")

# Obtendo a matriz de incidência
inc_matrix = nx.incidence_matrix(G, oriented=False).todense()
print("Matriz de Incidência:")
print(inc_matrix)
```

```python
# Definindo o grafo direcionado
G = nx.DiGraph()

edgelist = [(1, 2), (2, 3), (3, 4), (4, 1), (1, 3)]

G.add_edges_from(edgelist)

# Plotando o grafo usando a função personalizada
draw_graph_with_labels_outside_arestas(G, "Grafo Direcionado")

# Obtendo a matriz de incidência
inc_matrix = nx.incidence_matrix(G, oriented=True).todense()

# Exibindo a matriz de incidência e a ordem das arestas
edges = list(G.edges())
print("Ordem das Arestas:", edges)
print("Matriz de Incidência:")
print(inc_matrix)
```
