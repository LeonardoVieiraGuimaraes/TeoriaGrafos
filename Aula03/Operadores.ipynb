# Operadores de Teoria dos Grafos

## Definições
A teoria dos grafos é um campo da matemática que estuda as propriedades dos grafos. Um grafo é uma estrutura composta por vértices (ou nós) e arestas (ou linhas) que conectam pares de vértices. Aqui estão algumas definições importantes:

1. **Grafo**: Um grafo G é um par ordenado (V, E), onde V é um conjunto de vértices e E é um conjunto de arestas.
2. **Vértice**: Um vértice é um ponto ou nó em um grafo. Representa uma entidade ou objeto.
3. **Aresta**: Uma aresta é uma conexão entre dois vértices em um grafo. Representa a relação entre as entidades.
4. **Grafo Dirigido**: Um grafo onde as arestas têm uma direção associada, indicando uma relação assimétrica.
5. **Grafo Não Dirigido**: Um grafo onde as arestas não têm direção, indicando uma relação simétrica.
6. **Grau de um Vértice**: O número de arestas incidentes a um vértice. Em grafos dirigidos, distinguimos entre grau de entrada e grau de saída.
7. **Caminho**: Uma sequência de vértices onde cada par consecutivo é conectado por uma aresta. Representa uma rota ou sequência de passos.
8. **Ciclo**: Um caminho que começa e termina no mesmo vértice. Indica uma repetição ou loop.
9. **Conectividade**: Um grafo é conectado se há um caminho entre qualquer par de vértices. Indica que todas as partes do grafo estão acessíveis umas às outras.
10. **Subgrafo**: Um grafo formado a partir de um subconjunto dos vértices e arestas de um grafo maior. Representa uma parte ou seção do grafo original.

```python
# Operadores de Teoria dos Grafos
import networkx as nx
import matplotlib.pyplot as plt

# Criar um grafo não dirigido
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4)])

# Exibir o grafo
nx.draw(G, with_labels=True)
plt.show()

# Verificar o grau de um vértice
grau_vertice = G.degree(1)
print(f"Grau do vértice 1: {grau_vertice}")

# Verificar se o grafo é conectado
conectado = nx.is_connected(G)
print(f"O grafo é conectado? {conectado}")

# Encontrar o caminho mais curto entre dois vértices
caminho_curto = nx.shortest_path(G, source=1, target=4)
print(f"Caminho mais curto entre 1 e 4: {caminho_curto}")

# Verificar se há um ciclo no grafo
ciclo = nx.find_cycle(G, source=1)
print(f"Ciclo encontrado: {ciclo}")
```

## Operadores de Teoria dos Grafos

### União
A união de dois grafos G1 e G2 é um grafo G que contém todos os vértices e arestas de G1 e G2. Este operador combina os elementos de ambos os grafos, preservando todas as conexões.

### Interseção
A interseção de dois grafos G1 e G2 é um grafo G que contém apenas os vértices e arestas que estão presentes em ambos os grafos G1 e G2. Este operador identifica os elementos comuns entre os grafos.

### Junção
A junção de dois grafos G1 e G2 é um grafo G que contém todos os vértices de G1 e G2 e todas as arestas de G1 e G2, além de arestas adicionais que conectam cada vértice de G1 a cada vértice de G2. Este operador cria um grafo completo a partir da combinação dos dois grafos.

### Produto
O produto de dois grafos G1 e G2 é um grafo G onde os vértices são pares ordenados de vértices de G1 e G2, e há uma aresta entre dois vértices se e somente se as arestas correspondentes existirem em G1 e G2.

### Complemento
O complemento de um grafo G é um grafo G' com os mesmos vértices de G, mas onde duas vértices são adjacentes em G' se e somente se não são adjacentes em G.

### Contração
A contração de uma aresta em um grafo G é a operação de fundir os dois vértices extremos dessa aresta em um único vértice, removendo a aresta.

### Supressão e Subdivisão
A supressão de um vértice em um grafo G é a remoção desse vértice e todas as arestas incidentes. A subdivisão de uma aresta é a inserção de um novo vértice no meio da aresta, dividindo-a em duas arestas.

### Decomposição
A decomposição de um grafo G é a divisão do grafo em subgrafos menores que, juntos, representam o grafo original.

### Operador Linha L
O operador linha L de um grafo G é um grafo onde cada vértice representa uma aresta de G, e duas vértices são adjacentes se e somente se as arestas correspondentes em G compartilham um vértice comum.

```python
# Operadores de Teoria dos Grafos Avançados
import networkx as nx
import matplotlib.pyplot as plt

# Criar dois grafos
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3)])

G2 = nx.Graph()
G2.add_edges_from([(3, 4), (4, 5)])

# União
G_union = nx.compose(G1, G2)
print("União de G1 e G2:")
nx.draw(G_union, with_labels=True)
plt.show()

# Interseção
G_intersection = nx.intersection(G1, G2)
print("Interseção de G1 e G2:")
nx.draw(G_intersection, with_labels=True)
plt.show()

# Junção
G1.add_nodes_from(G2.nodes())
G2.add_nodes_from(G1.nodes())
G_join = nx.complete_graph(len(G1.nodes) + len(G2.nodes))
print("Junção de G1 e G2:")
nx.draw(G_join, with_labels=True)
plt.show()

# Produto
G_product = nx.tensor_product(G1, G2)
print("Produto de G1 e G2:")
nx.draw(G_product, with_labels=True)
plt.show()

# Complemento
G_complement = nx.complement(G1)
print("Complemento de G1:")
nx.draw(G_complement, with_labels=True)
plt.show()

# Contração
G_contraction = nx.contracted_edge(G1, (1, 2), self_loops=False)
print("Contração da aresta (1, 2) em G1:")
nx.draw(G_contraction, with_labels=True)
plt.show()


# Supondo que G1 já esteja definido
# ...

# Supressão e Subdivisão
G_suppression = G1.copy()
G_suppression.remove_node(2)
print("Supressão do vértice 2 em G1:")
nx.draw(G_suppression, with_labels=True)
plt.show()

# Subdivisão manual da aresta (1, 2)
G_subdivision = G1.copy()
u, v = 1, 2
if G_subdivision.has_edge(u, v):
    G_subdivision.remove_edge(u, v)
    new_node = max(G_subdivision.nodes) + 1
    G_subdivision.add_node(new_node)
    G_subdivision.add_edge(u, new_node)
    G_subdivision.add_edge(new_node, v)
print("Subdivisão da aresta (1, 2) em G1:")
nx.draw(G_subdivision, with_labels=True)
plt.show()

# Decomposição
components = list(nx.connected_components(G1))
print("Decomposição de G1 em componentes conectados:")
for i, component in enumerate(components):
    subgraph = G1.subgraph(component)
    print(f"Componente {i + 1}:")
    nx.draw(subgraph, with_labels=True)
    plt.show()

# Operador Linha L
G_line = nx.line_graph(G1)
print("Operador Linha L de G1:")
nx.draw(G_line, with_labels=True)
plt.show()

# Decomposição
components = list(nx.connected_components(G1))
print("Decomposição de G1 em componentes conectados:")
for i, component in enumerate(components):
    subgraph = G1.subgraph(component)
    print(f"Componente {i + 1}:")
    nx.draw(subgraph, with_labels=True)
    plt.show()

# Operador Linha L
G_line = nx.line_graph(G1)
print("Operador Linha L de G1:")
nx.draw(G_line, with_labels=True)
plt.show()
```
