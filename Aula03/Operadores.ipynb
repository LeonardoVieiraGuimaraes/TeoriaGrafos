## Operador União em Grafos
O operador união em grafos combina dois grafos em um único grafo que contém todos os vértices e arestas dos grafos originais. 
Formalmente, a união de dois grafos G1 = (V1, E1) e G2 = (V2, E2) é um grafo G = (V, E), onde V = V1 ∪ V2 e E = E1 ∪ E2.

```python
# Criando o grafo G1
G1 = nx.Graph()
G1.add_edges_from([(0, 1), (0, 3), (2, 3)])

# Criando o grafo G2
G2 = nx.Graph()
G2.add_edges_from([(1, 2), (1, 4), (2, 4), (1, 5)])

# Unindo os grafos G1 e G2
G = nx.compose(G1, G2)

# Plotar os grafos um embaixo do outro
# Plotar os grafos um embaixo do outro
plt.figure(figsize=(8, 12))
plt.subplot(311)
draw_graph_with_labels_outside(G1, 'Grafo G1')
plt.figure(figsize=(8, 12))
plt.subplot(312)
draw_graph_with_labels_outside(G2, 'Grafo G2')
plt.show()
plt.figure(figsize=(8, 12))
plt.subplot(313)
draw_graph_with_labels_outside(G, 'Grafo G')
plt.show()
```

## Operador Interseção em Grafos
O operador interseção em grafos cria um novo grafo que contém apenas os vértices e arestas que estão presentes em ambos os grafos originais. 
Formalmente, a interseção de dois grafos G1 = (V1, E1) e G2 = (V2, E2) é um grafo G = (V, E), onde V = V1 ∩ V2 e E = E1 ∩ E2.

```python
# Criando o grafo G1
G1 = nx.Graph()
G1.add_edges_from([(0, 1), (0, 3), (1, 5), (2, 3), (2, 4), (3, 4), (4, 5)])

# Criando o grafo G2
G2 = nx.Graph()
G2.add_edges_from([(1, 4), (1, 6), (3, 6), (3, 4), (3, 5), (3, 7), (4, 5), (5, 7), (6, 7)])

# Interseção dos grafos
G_intersection = nx.intersection(G1, G2)

# Plotar os grafos um embaixo do outro
plt.figure(figsize=(8, 18))
plt.subplot(311)
draw_graph_with_labels_outside(G1, 'Grafo G1')
plt.figure(figsize=(8, 18))
plt.subplot(312)
draw_graph_with_labels_outside(G2, 'Grafo G2')
plt.figure(figsize=(8, 18))
plt.subplot(313)
draw_graph_with_labels_outside(G_intersection, 'Interseção dos Grafos G1 e G2')

plt.show()

# Vértices e arestas da interseção
vertices = G_intersection.nodes()
arestas = G_intersection.edges()
print(f"Vértices da interseção: {list(vertices)}")
print(f"Arestas da interseção: {list(arestas)}")
```

## Operador Junção em Grafos
O operador junção em grafos cria um novo grafo que contém todos os vértices dos grafos originais e adiciona arestas entre todos os pares de vértices que pertencem a diferentes grafos originais. 
Formalmente, a junção de dois grafos G1 = (V1, E1) e G2 = (V2, E2) é um grafo G = (V, E), onde V = V1 ∪ V2 e E = E1 ∪ E2 ∪ {(u, v) | u ∈ V1, v ∈ V2}.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo G1
G1 = nx.Graph()
G1.add_edges_from([(0, 1), (0, 2), (1, 2)])

# Criando o grafo G2
G2 = nx.Graph()
G2.add_edges_from([('a', 'b'), ('b', 'c'), ('b', 'd')])

# Criando a junção dos grafos G1 e G2
G = nx.Graph()
G.add_edges_from(G1.edges())
G.add_edges_from(G2.edges())

# Adicionando arestas entre todos os vértices de G1 e todos os vértices de G2
for v1 in G1.nodes():
    for v2 in G2.nodes():
        G.add_edge(v1, v2)

# Plotar os grafos um embaixo do outro

plt.figure(figsize=(8, 18))

plt.subplot(311)
draw_graph_with_labels_outside(G1, 'Grafo G1')
plt.figure(figsize=(8, 18))
plt.subplot(312)
draw_graph_with_labels_outside(G2, 'Grafo G2')
plt.figure(figsize=(8, 18))
plt.subplot(313)
draw_graph_with_labels_outside(G, 'Junção dos Grafos G1 e G2')

plt.show()
```

## Operador Junção Direta (ou Soma Direta) em Grafos

Formalmente, a junção direta de dois grafos G1 = (V1, E1) e G2 = (V2, E2) é um grafo G = (V, E), onde V = V1 ∪ V2 e E = E1 ∪ E2.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando o grafo G1
G1 = nx.Graph()
G1.add_edges_from([(0, 1), (0, 2), (1, 2)])

# Criando o grafo G2
G2 = nx.Graph()
G2.add_edges_from([('a', 'b'), ('b', 'c'), ('b', 'd')])

# Criando a junção direta dos grafos G1 e G2
G = nx.compose(G1, G2)

# Plotar os grafos um embaixo do outro
plt.figure(figsize=(8, 18))

plt.subplot(311)
draw_graph_with_labels_outside(G1, 'Grafo G1')
plt.figure(figsize=(8, 18))
plt.subplot(312)
draw_graph_with_labels_outside(G2, 'Grafo G2')
plt.figure(figsize=(8, 18))
plt.subplot(313)
draw_graph_with_labels_outside(G, 'Junção Direta dos Grafos G1 e G2')

plt.show()
```

## Operador Produto em Grafos
O operador produto em grafos cria um novo grafo a partir de dois grafos originais, onde os vértices do novo grafo são pares ordenados de vértices dos grafos originais e as arestas são definidas de acordo com um tipo específico de produto (cartesiano, tensorial, forte, etc.). 
Por exemplo, no produto cartesiano de dois grafos G1 = (V1, E1) e G2 = (V2, E2), o novo grafo G = (V, E) tem V = V1 × V2 e E = {((u1, v1), (u2, v2)) | (u1 = u2 e (v1, v2) ∈ E2) ou ((u1, u2) ∈ E1 e v1 = v2)}.

```python
import networkx as nx
import matplotlib.pyplot as plt


# Criando o grafo G1
G1 = nx.Graph()
G1.add_edges_from([(0, 1)])

# Criando o grafo G2
G2 = nx.Graph()
G2.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'a')])

# Criando o produto cartesiano dos grafos G1 e G2
G_product = nx.cartesian_product(G1, G2)

# Plotar os grafos um embaixo do outro
plt.figure(figsize=(8, 18))

plt.subplot(311)
draw_graph_with_labels_outside(G1, 'Grafo G1')
plt.figure(figsize=(8, 18))
plt.subplot(312)
draw_graph_with_labels_outside(G2, 'Grafo G2')
plt.figure(figsize=(8, 18))
plt.subplot(313)
draw_graph_with_labels_outside_q(G_product, 'Produto Cartesiano dos Grafos G1 e G2')

plt.show()
```

## Operador Complemento em Grafos
O operador complemento em grafos cria um novo grafo a partir de um grafo original, onde os vértices permanecem os mesmos, mas as arestas são invertidas. 
Ou seja, se dois vértices estão conectados no grafo original, eles não estarão conectados no complemento, e vice-versa. 
Formalmente, o complemento de um grafo G = (V, E) é um grafo G' = (V, E'), onde E' = {(u, v) | u, v ∈ V e (u, v) ∉ E}.

## Operador Remoção em Grafos

### Remoção de um Vértice

Quando um vértice é removido de um grafo, ele e todas as arestas incidentes a ele são eliminados.

**Notação**: Dado um grafo $G = (V, E)$ e um vértice $v \in V$, o grafo resultante da remoção de $v$ é denotado por $G - v$.

**Definição formal**:

$$ G - v = (V \setminus \{v\}, E \setminus \{\text{arestas incidentes a } v\}) $$


### Remoção de uma Aresta

Quando uma aresta é removida de um grafo, apenas a aresta é eliminada, os vértices permanecem.

**Notação**: Dado um grafo $G = (V, E)$ e uma aresta $e \in E$, o grafo resultante da remoção de $e$ é denotado por $G - e$.

**Definição formal**:

$$ G - e = (V, E \setminus \{e\}) $$



```python
# Criar o grafo original G
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (1, 3)])

# Desenhar o grafo original
draw_graph_with_labels_outside(G, 'Grafo Original G')

# Remover o vértice 2
G_v_removed = G.copy()
G_v_removed.remove_node(2)

# Desenhar o grafo resultante após a remoção do vértice 2
draw_graph_with_labels_outside(G_v_removed, 'Grafo G - 2')

# Remover a aresta (1, 2)
G_e_removed = G.copy()
G_e_removed.remove_edge(1, 2)

# Desenhar o grafo resultante após a remoção da aresta (1, 2)
draw_graph_with_labels_outside(G_e_removed, 'Grafo G - (1, 2)')

# Imprimir os vértices e arestas do grafo resultante após remoção do vértice
print("Vértices do grafo G:", G.nodes())
print("Arestas do grafo G:", G.edges())


# Imprimir os vértices e arestas do grafo resultante após remoção do vértice
print("Vértices do grafo resultante após remoção do vértice 2:", G_v_removed.nodes())
print("Arestas do grafo resultante após remoção do vértice 2:", G_v_removed.edges())

# Imprimir os vértices e arestas do grafo resultante após remoção da aresta
print("Vértices do grafo resultante após remoção da aresta (1, 2):", G_e_removed.nodes())
print("Arestas do grafo resultante após remoção da aresta (1, 2):", G_e_removed.edges())
```

## Contração de uma Aresta

Quando uma aresta é contraída em um grafo, os dois vértices que ela conecta são fundidos em um único vértice, e todas as arestas incidentes a esses vértices são ajustadas para refletir essa fusão.

**Notação**: Dado um grafo $G = (V, E)$ e uma aresta $e = \{u, v\} \in E$, o grafo resultante da contração de $e$ é denotado por $G / e$.

**Definição formal**:

$$ G / e = (V', E') $$

onde:
- $V' = (V \setminus \{u, v\}) \cup \{w\}$, onde $w$ é o novo vértice resultante da fusão de $u$ e $v$.
- $E' = \{\{x, y\} \in E \mid x, y \notin \{u, v\}\} \cup \{\{x, w\} \mid \{x, u\} \in E \text{ ou } \{x, v\} \in E\}$

```python
# Função para contrair uma aresta
def contract_edge(G, edge):
    u, v = edge
    G_contracted = nx.contracted_nodes(G, u, v, self_loops=False)
    return G_contracted

# Criar o grafo original G
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (1, 3)])

# Desenhar o grafo original
draw_graph_with_labels_outside(G, 'Grafo Original G')

# Contrair a aresta (1, 2)
G_contracted = contract_edge(G, (1, 2))

# Desenhar o grafo resultante após a contração da aresta (1, 2)
draw_graph_with_labels_outside(G_contracted, 'Grafo G / (1, 2)')


print("Vértices do grafo G:", G.nodes())
print("Arestas do grafo G:", G.edges())
# Imprimir os vértices e arestas do grafo resultante após contração da aresta

print("Vértices do grafo resultante após contração da aresta (1, 2):", G_contracted.nodes())
print("Arestas do grafo resultante após contração da aresta (1, 2):", G_contracted.edges())
```

## Operador de Supressão e Subdivisão de Grafos

### Supressão de um Vértice

A operação de supressão de um vértice é realizada apenas em vértices de grau 2 e consiste no seguinte processo. Seja $G = (V, A)$ um grafo qualquer e $v$ um vértice adjacente a dois vértices $u$ e $w$. A operação de supressão de $v$ remove $v$ do conjunto $V$ e cria uma aresta entre $u$ e $w$, produzindo um novo grafo $G' = (V', A')$. O grafo $G'$ é chamado grafo reduzido de $G$ ou grafo homeomorficamente reduzido de $G$. Ele possui $V' = V \setminus \{v\}$ e $A' = A \setminus A_1 \cup \{(u, w)\}$, com $A_1 = \{(v, u), (v, w)\}$.

### Subdivisão de uma Aresta

A operação inversa é chamada subdivisão de arestas. Ela consiste em inserir vértices de grau 2 nas arestas do grafo, criando subdivisões de arestas. Esta operação funciona da seguinte maneira. Seja $G = (V, A)$ um grafo qualquer, $v$ um vértice novo e uma aresta $a = \{u, w\}$. A operação de subdivisão de $a$ remove $a$ do conjunto $A$ e adiciona duas novas arestas $\{u, v\}$ e $\{v, w\}$, produzindo um novo grafo $G' = (V', A')$. Ele possui $V' = V \cup \{v\}$ e $A' = A \setminus \{a\} \cup \{(u, v), (v, w)\}$.

**Exemplo de Supressão de Vértice**:

Grafo original $G$:

- **Vértices**: $V = \{1, 2, 3\}$
- **Arestas**: $A = \{(1, 2), (2, 3)\}$

Supressão do vértice $2$:

- **Vértices restantes**: $V' = \{1, 3\}$
- **Arestas restantes**: $A' = \{(1, 3)\}$

Grafo resultante: $G' = (\{1, 3\}, \{(1, 3)\})$

**Exemplo de Subdivisão de Aresta**:

Grafo original $G$:

- **Vértices**: $V = \{1, 2\}$
- **Arestas**: $A = \{(1, 2)\}$

Subdivisão da aresta $(1, 2)$:

- **Vértices restantes**: $V' = \{1, 2, 3\}$, onde $3$ é o novo vértice inserido.
- **Arestas restantes**: $A' = \{(1, 3), (3, 2)\}$

Grafo resultante: $G' = (\{1, 2, 3\}, \{(1, 3), (3, 2)\})$

```python
# Função para suprimir um vértice de grau 2
def suppress_vertex(G, v):
    neighbors = list(G.neighbors(v))
    if len(neighbors) == 2:
        u, w = neighbors
        G.remove_node(v)
        G.add_edge(u, w)
    return G

# Função para subdividir uma aresta
def subdivide_edge(G, edge):
    u, v = edge
    w = max(G.nodes) + 1  # Novo vértice
    G.remove_edge(u, v)
    G.add_node(w)
    G.add_edges_from([(u, w), (w, v)])
    return G

# Criar o grafo original G para supressão
G_suppress = nx.Graph()
G_suppress.add_edges_from([(1, 2), (2, 3)])

# Desenhar o grafo original para supressão
draw_graph_with_labels_outside(G_suppress, 'Grafo Original G')

# Suprimir o vértice 2
G_suppress_result = suppress_vertex(G_suppress.copy(), 2)

# Desenhar o grafo resultante após a supressão do vértice 2
draw_graph_with_labels_outside(G_suppress_result, 'Grafo G - 2 (Supressão)')

# Criar o grafo original G para subdivisão
G_subdivide = nx.Graph()
G_subdivide.add_edges_from([(1, 2)])

# Desenhar o grafo original para subdivisão
draw_graph_with_labels_outside(G_subdivide, 'Grafo Original G')

# Subdividir a aresta (1, 2)
G_subdivide_result = subdivide_edge(G_subdivide.copy(), (1, 2))

# Desenhar o grafo resultante após a subdivisão da aresta (1, 2)
draw_graph_with_labels_outside(G_subdivide_result, 'Grafo G + e\' (Subdivisão)')


# Imprimir os vértices e arestas do grafo resultante após supressão do vértice
print("Vértices do grafo resultante após supressão do vértice 2:", G_suppress_result.nodes())
print("Arestas do grafo resultante após supressão do vértice 2:", G_suppress_result.edges())

# Imprimir os vértices e arestas do grafo resultante após subdivisão da aresta
print("Vértices do grafo resultante após subdivisão da aresta (1, 2):", G_subdivide_result.nodes())
print("Arestas do grafo resultante após subdivisão da aresta (1, 2):", G_subdivide_result.edges())
```

## Decomposição de um Grafo

A operação de decomposição de um grafo $G$ gera um conjunto de subgrafos $G_1, G_2, \ldots$ tal que cada aresta de $G$ aparece exatamente uma única vez em um único subgrafo. Ou seja, considerando que o conjunto de arestas do grafo $G_i$ é $A_i$, temos $A_i \cap A_j = \emptyset$, $\forall i \neq j$. Os subgrafos gerados podem compartilhar vértices, entretanto não podem compartilhar arestas.

**Notação**: Dado um grafo $G = (V, E)$, a decomposição de $G$ é um conjunto de subgrafos $\{G_1, G_2, \ldots, G_k\}$ tal que:

$$ E = \bigcup_{i=1}^{k} A_i \quad \text{e} \quad A_i \cap A_j = \emptyset \quad \forall i \neq j $$

**Exemplo**:

Grafo original $G$:

- **Vértices**: $V = \{1, 2, 3, 4\}$
- **Arestas**: $E = \{\{1, 2\}, \{2, 3\}, \{3, 4\}, \{4, 1\}\}$

Decomposição do grafo $G$ em dois subgrafos $G_1$ e $G_2$:

- **Subgrafo $G_1$**:
  - **Vértices**: $V_1 = \{1, 2, 3\}$
  - **Arestas**: $A_1 = \{\{1, 2\}, \{2, 3\}\}$

- **Subgrafo $G_2$**:
  - **Vértices**: $V_2 = \{3, 4, 1\}$
  - **Arestas**: $A_2 = \{\{3, 4\}, \{4, 1\}\}$

Os subgrafos $G_1$ e $G_2$ compartilham o vértice $3$ e $3$, mas não compartilham arestas.

```python
# Criar o grafo original G
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Desenhar o grafo original

draw_graph_with_labels_outside(G, 'Grafo Original G')

# Decompor o grafo G em dois subgrafos
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3)])

G2 = nx.Graph()
G2.add_edges_from([(3, 4), (4, 1)])

# Desenhar os subgrafos resultantes

draw_graph_with_labels_outside(G1, 'Subgrafo G1')


draw_graph_with_labels_outside(G2, 'Subgrafo G2')


# Imprimir os vértices e arestas dos subgrafos resultantes
print("Vértices do subgrafo G1:", G1.nodes())
print("Arestas do subgrafo G1:", G1.edges())
print("Vértices do subgrafo G2:", G2.nodes())
print("Arestas do subgrafo G2:", G2.edges())
```

## Operador Linha \( L \)

Para facilitar a solução de certos problemas, convém realizar o mapeamento das linhas do grafo para vértices e vice-versa. Este mapeamento é realizado pelo operador $L$, o qual recebe como operando um grafo (ou um dígrafo) e retorna um novo grafo (ou dígrafo), chamado grafo linha (ou dígrafo linha).

### Grafo Linha

Dado um grafo $G = (V, A)$, o grafo linha $L(G) = (V_L, A_L)$ correspondente é construído da seguinte maneira:
- Cada aresta $\{v, u\} \in A$ dá origem a um vértice $vu \in L(G)$, ou seja, um vértice $vu \in V_L$ se e somente se $\{v, u\} \in A$.
- A adjacência entre os vértices em $L(G)$ é definida da seguinte maneira: dois vértices distintos $uv, xw \in V_L$ são adjacentes se as arestas correspondentes em $G$ são adjacentes, ou seja, $\{u, v\} \cap \{x, w\} \neq \emptyset$.

### Exemplo

#### Grafo Original $G$

- **Vértices**: $V = \{1, 2, 3, 4\}$
- **Arestas**: $A = \{\{1, 2\}, \{2, 3\}, \{3, 4\}, \{4, 1\}\}$

#### Grafo Linha $L(G)$

- **Vértices**: $V_L = \{12, 23, 34, 41\}$
- **Arestas**: $A_L = \{\{12, 23\}, \{23, 34\}, \{34, 41\}, \{41, 12\}, \{12, 41\}, \{23, 34\}\}$

Os vértices $12$ e $23$ em $L(G)$ são adjacentes porque as arestas $\{1, 2\}$ e $\{2, 3\}$ em $G$ compartilham o vértice $2$.

```python
# Função para criar o grafo linha
def line_graph(G):
    L = nx.Graph()
    for edge in G.edges():
        L.add_node(edge)
    for edge1 in G.edges():
        for edge2 in G.edges():
            if edge1 != edge2 and (edge1[0] in edge2 or edge1[1] in edge2):
                L.add_edge(edge1, edge2)
    return L

# Criar o grafo original G
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])

# Desenhar o grafo original

draw_graph_with_labels_outside(G, 'Grafo Original G')

# Criar o grafo linha L(G)
L_G = line_graph(G)

# Desenhar o grafo linha

draw_graph_with_labels_outside_quad(L_G, 'Grafo Linha L(G)')

plt.show()

# Imprimir os vértices e arestas do grafo linha
print("Vértices do grafo linha L(G):", L_G.nodes())
print("Arestas do grafo linha L(G):", L_G.edges())
```
