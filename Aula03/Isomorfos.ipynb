# Teoria dos Grafos
## Aula sobre Grafos Isomorfos

Nesta aula, vamos explorar o conceito de grafos isomorfos, incluindo a definição matemática e como verificar isomorfismo entre grafos usando a biblioteca NetworkX do Python.

```python
# Importar a biblioteca NetworkX e Matplotlib
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Função para desenhar o grafo com bordas pretas, nós brancos e rótulos pretos fora das bolinhas
def draw_graph_with_labels_outside(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    plt.title(title)
    plt.show()
# # Função para desenhar o grafo com bordas pretas, nós brancos e rótulos pretos fora das bolinhas
# def draw_graph_with_border(G, title):
#     pos = nx.spring_layout(G)
#     nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=700, linewidths=2)
#     nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=500, linewidths=2)
#     nx.draw_networkx_labels(G, pos, font_size=10, font_color='black')
#     plt.title(title)
#     plt.show()
```

### Conceito de Grafos Isomorfos
Dois grafos são isomorfos se existe uma correspondência entre seus vértices e arestas que preserva a estrutura do grafo.

**Definição Matemática:** Dois grafos $G_1 = (V_1, E_1)$ e $G_2 = (V_2, E_2)$ são isomorfos se existe uma bijeção $f: V_1 \rightarrow V_2$ tal que qualquer aresta $(u, v) \in E_1$ se e somente se $(f(u), f(v)) \in E_2$.

Dois grafos $G_1 = (V_1, A_1)$ e $G_2 = (V_2, A_2)$ são isomorfos, ou seja, $G_1 \cong G_2$, se existe uma função bijetora $f: V_1 \rightarrow V_2$, tal que $(u, v) \in A_1$ se $(f(u), f(v)) \in A_2$.

```python
import networkx as nx
import matplotlib.pyplot as plt


# Criar dois grafos não isomorfos
G1 = nx.Graph()
G1.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'), ('a', 'c')])

G2 = nx.Graph()
G2.add_edges_from([('w', 'x'), ('x', 'y'), ('y', 'z'), ('z', 'w'), ('w', 'y')])

# Verificar se os grafos são isomorfos
is_isomorphic = nx.is_isomorphic(G1, G2)
print(f'Os grafos G1 e G2 são isomorfos? {is_isomorphic}')

# Verificar as condições necessárias para isomorfismo
# Condição 1: Mesmo número de vértices
num_vertices_G1 = G1.number_of_nodes()
num_vertices_G2 = G2.number_of_nodes()
print(f'Número de vértices em G1: {num_vertices_G1}')
print(f'Número de vértices em G2: {num_vertices_G2}')

# Condição 2: Mesmo número de arestas
num_arestas_G1 = G1.number_of_edges()
num_arestas_G2 = G2.number_of_edges()
print(f'Número de arestas em G1: {num_arestas_G1}')
print(f'Número de arestas em G2: {num_arestas_G2}')

# Condição 3: Mesmo número de vértices com um determinado grau
graus_G1 = sorted([G1.degree(n) for n in G1.nodes()])
graus_G2 = sorted([G2.degree(n) for n in G2.nodes()])
print(f'Graus dos vértices em G1: {graus_G1}')
print(f'Graus dos vértices em G2: {graus_G2}')

# Plotar os grafos um embaixo do outro
plt.figure(figsize=(8, 12))
plt.subplot(211)
draw_graph_with_labels_outside(G1, 'Grafo G1')
plt.figure(figsize=(8, 12))
plt.subplot(212)
draw_graph_with_labels_outside(G2, 'Grafo G2')
plt.show()
```

### Exemplo de Grafos não Isomorfos
Vamos criar dois grafos que são isomorfos e verificar essa propriedade usando a biblioteca NetworkX.

* Condições necessárias para isomorfismo entre dois grafos G e H:
* G e H devem possuir o mesmo número de vértices;
* G e H devem possuir o mesmo número de arestas;
* G e H devem possuir o mesmo número de vértices com um determinado grau.

As condições 1, 2 e 3 acima são suficientes?

```python
import networkx as nx
import matplotlib.pyplot as plt


# Criar dois grafos não isomorfos
G1 = nx.Graph()
G1.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'e'), ('d', 'f')])

G2 = nx.Graph()
G2.add_edges_from([('g', 'h'), ('h', 'i'), ('i', 'j'), ('i', 'l'), ('l', 'm')])

# Verificar se os grafos são isomorfos
is_isomorphic = nx.is_isomorphic(G1, G2)
print(f'Os grafos G1 e G2 são isomorfos? {is_isomorphic}')

# Verificar as condições necessárias para isomorfismo
# Condição 1: Mesmo número de vértices
num_vertices_G1 = G1.number_of_nodes()
num_vertices_G2 = G2.number_of_nodes()
print(f'Número de vértices em G1: {num_vertices_G1}')
print(f'Número de vértices em G2: {num_vertices_G2}')

# Condição 2: Mesmo número de arestas
num_arestas_G1 = G1.number_of_edges()
num_arestas_G2 = G2.number_of_edges()
print(f'Número de arestas em G1: {num_arestas_G1}')
print(f'Número de arestas em G2: {num_arestas_G2}')

# Condição 3: Mesmo número de vértices com um determinado grau
graus_G1 = sorted([G1.degree(n) for n in G1.nodes()])
graus_G2 = sorted([G2.degree(n) for n in G2.nodes()])
print(f'Graus dos vértices em G1: {graus_G1}')
print(f'Graus dos vértices em G2: {graus_G2}')

# Plotar os grafos um embaixo do outro

plt.figure(figsize=(8, 12))
plt.subplot(211)
draw_graph_with_labels_outside(G1, 'Grafo G')
plt.figure(figsize=(8, 12))
plt.subplot(212)
draw_graph_with_labels_outside(G2,'Grafo H')
plt.show()
```

Embora os grafos G3 e G4 pareçam semelhantes, eles não são isomorfos porque a estrutura de suas arestas não pode ser mapeada uma para a outra preservando a adjacência dos vértices.
