# Aula sobre Subgrafos

## Conceitos e Definições

Um **subgrafo** é um grafo cujos vértices e arestas são subconjuntos de um grafo maior. Em outras palavras, se tivermos um grafo G = (V, E), um subgrafo G' = (V', E') é tal que V' ⊆ V e E' ⊆ E.


```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph_with_labels_outside(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    plt.title(title)
    plt.show()
```

```python
# Criando um grafo G
G = nx.Graph()
G.add_edges_from([('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a')])

# Criando um subgrafo H a partir dos vértices ['a', 'b', 'c']
H = G.subgraph(['a', 'b', 'c'])

# Plotar os grafos um embaixo do outro
plt.figure(figsize=(8, 12))
plt.subplot(211)
draw_graph_with_labels_outside(G, 'Grafo G1')
plt.figure(figsize=(8, 12))
plt.subplot(212)
draw_graph_with_labels_outside(H, 'Grafo G2')
plt.show()
```

Existem diferentes tipos de subgrafos:
- **Subgrafo Próprio**: Um subgrafo que não é igual ao grafo original.
- **Subgrafo Induzido**: Um subgrafo formado por um subconjunto de vértices e todas as arestas entre esses vértices no grafo original.
- **Subgrafo Gerador**: Um subgrafo que contém todos os vértices do grafo original.


# Tipos de Subgrafos
## Subgrafo Gerado (Induzido por Vértices)
Mantém todos os vértices selecionados e todas as arestas de G entre eles.
Exemplo: Se G é um quadrado (C4), um subgrafo induzido por {A,B,C} é um triângulo.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Grafo original
G = nx.cycle_graph(4)
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()

# Subgrafo induzido por {0, 1, 2}
H = G.subgraph([0, 1, 2])
nx.draw(H, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()
```

## Subgrafo Aresta-induzido
Mantém um subconjunto de arestas e os vértices incidentes a elas.

```python
# Subgrafo aresta-induzido
edges = [(0, 1), (1, 2)]
H = nx.edge_subgraph(G, edges)
nx.draw(H, pos, with_labels=True, node_color='lightgreen', edge_color='black')
plt.show()
```

## Subgrafo Espalhado (Spanning)
Contém todos os vértices de G mas não todas as arestas.
Exemplo: Uma árvore geradora de G.

```python
# Subgrafo espalhado (árvore geradora mínima)
T = nx.minimum_spanning_tree(G)
nx.draw(T, pos, with_labels=True, node_color='lightcoral', edge_color='blue')
plt.show()
```

## Exemplo de Subgrafo Induzido vs. Não Induzido
### Grafo Original G:
Vértices: {1,2,3,4}.
Arestas: 1−2,1−3,2−3,3−4.
### Subgrafo Induzido por {1,2,3}:
Arestas: 1−2,1−3,2−3 (preserva todas as conexões originais).
### Subgrafo Não Induzido:
Vértices: {1,2,3}.
Arestas: 1−2,2−3 (falta 1−3, não é induzido).

```python
# Grafo original
G = nx.Graph()
G.add_edges_from([(1, 2), (1, 3), (2, 3), (3, 4)])
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
plt.show()

# Subgrafo induzido por {1, 2, 3}
H = G.subgraph([1, 2, 3])
nx.draw(H, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.show()

# Subgrafo não induzido
H = nx.Graph()
H.add_edges_from([(1, 2), (2, 3)])
nx.draw(H, pos, with_labels=True, node_color='lightgreen', edge_color='black')
plt.show()
```

## Propriedades de Subgrafos
### Hereditariedade:
Se um grafo G tem uma propriedade (ex: bipartido), nem todo subgrafo herdará essa propriedade.
Exemplo: K3 (triângulo) é um subgrafo de K4, mas K3 não é bipartido.

### Operações:
Remoção de Vértices: G−v.
Remoção de Arestas: G−e.

## Aplicações de Subgrafos
### Redes Sociais:
Subgrafos representam comunidades ou grupos específicos.
### Circuitos Elétricos:
Subcircuitos são subgrafos do circuito completo.
### Análise de Conectividade:
Subgrafos conexos (ex: componentes fortemente conexos).
