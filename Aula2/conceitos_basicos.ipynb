# Criação do Grafo Representando a Ponte de Königsberg

A Ponte de Königsberg é um famoso problema histórico de teoria dos grafos. O problema consiste em determinar se é possível atravessar todas as pontes da cidade de Königsberg exatamente uma vez e retornar ao ponto de partida. Este problema foi resolvido por Leonhard Euler em 1736, que demonstrou que não é possível fazer tal travessia.



```python
# Criação do grafo
G = nx.MultiGraph()  # Usando MultiGraph para permitir arestas múltiplas

# Adicionando nós (representando as áreas da cidade)
G.add_nodes_from(["A", "B", "C", "D"])

# Adicionando arestas (representando as pontes)
G.add_edges_from([
    ("A", "B"),
    ("A", "B"),
    ("A", "C"),
    ("A", "D"),
    ("B", "C"),
    ("B", "D"),
    ("C", "D")
])

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="lightblue", font_size=12, font_weight="bold")

# Destacando arestas duplicadas
edges = G.edges(keys=True)
edge_colors = []
for u, v, k in edges:
    if G.number_of_edges(u, v) > 1:
        edge_colors.append('red')
    else:
        edge_colors.append('black')

nx.draw_networkx_edges(G, pos, edgelist=edges, edge_color=edge_colors)
plt.show()
```

## Verificação da Solução
Para verificar se é possível atravessar todas as pontes exatamente uma vez e retornar ao ponto de partida, precisamos verificar se o grafo possui um circuito Euleriano. Um grafo possui um circuito Euleriano se e somente se todos os vértices têm grau par.

```python
# Verificação dos graus dos vértices
degrees = dict(G.degree())
print("Graus dos vértices:", degrees)

# Verificação se todos os vértices têm grau par
eulerian = all(degree % 2 == 0 for degree in degrees.values())
if eulerian:
    print("É possível atravessar todas as pontes exatamente uma vez e retornar ao ponto de partida.")
else:
    print("Não é possível atravessar todas as pontes exatamente uma vez e retornar ao ponto de partida.")
```

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
    plt.title(title)
    plt.show()

# Conceito de Grafos e Dígrafos
G = nx.Graph()
G.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
draw_graph(G, "Grafo")

DG = nx.DiGraph()
DG.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
draw_graph(DG, "Dígrafo")

# Laços e Arestas Paralelas
G_loops = nx.MultiGraph()
G_loops.add_edges_from([(1, 1), (1, 2), (1, 2), (2, 3)])
draw_graph(G_loops, "Laços e Arestas Paralelas")

# Parâmetros Quantitativos
print("Ordem do grafo:", G.number_of_nodes())
print("Tamanho do grafo:", G.number_of_edges())
print("Grau do vértice 1:", G.degree[1])

# Grafos Rotulados e Não-Rotulados
G_rotulado = nx.Graph()
G_rotulado.add_edges_from([(1, 2), (2, 3), (3, 4)])
nx.set_node_attributes(G_rotulado, {1: 'A', 2: 'B', 3: 'C', 4: 'D'}, 'label')
draw_graph(G_rotulado, "Grafo Rotulado")

# Isomorfismo
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 1)])
G2 = nx.Graph()
G2.add_edges_from([(5, 6), (6, 7), (7, 8), (8, 5)])
print("G1 e G2 são isomorfos?", nx.is_isomorphic(G1, G2))

# Subgrafos
subgraph = G.subgraph([1, 2, 3])
draw_graph(subgraph, "Subgrafo")

# Classes de Grafos
# Grafo Completo
G_completo = nx.complete_graph(5)
draw_graph(G_completo, "Grafo Completo")

# Grafo Bipartido
G_bipartido = nx.complete_bipartite_graph(3, 2)
draw_graph(G_bipartido, "Grafo Bipartido")

# Grafo Cíclico
G_ciclico = nx.cycle_graph(5)
draw_graph(G_ciclico, "Grafo Cíclico")

# Grafo Acíclico (Árvore)
G_arvore = nx.balanced_tree(2, 3)
draw_graph(G_arvore, "Árvore")

# Percursos, Trilhas e Caminhos
# Percurso
percurso = list(nx.dfs_edges(G, source=1))
print("Percurso:", percurso)

# Trilha
trilha = list(nx.edge_dfs(G, source=1))
print("Trilha:", trilha)

# Caminho
caminho = nx.shortest_path(G, source=1, target=3)
print("Caminho:", caminho)
```
