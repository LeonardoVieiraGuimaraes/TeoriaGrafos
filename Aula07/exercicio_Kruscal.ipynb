# Definições de Árvore em Grafos

1. **Árvore é um grafo conexo e acíclico**  
Um grafo não direcionado é uma árvore se for conexo (existe caminho entre quaisquer dois vértices) e não possuir ciclos.

2. **Árvore é um grafo conexo com n vértices e n-1 arestas**  
Se um grafo não direcionado possui $n$ vértices, é conexo e tem exatamente $n-1 arestas, então ele é uma árvore.

3. **Árvore é um grafo acíclico maximamente**  
Uma árvore é um grafo acíclico ao qual, se adicionarmos qualquer aresta, obrigatoriamente formará um ciclo.

4. **Árvore é um grafo conexo minimamente**  
Uma árvore é um grafo conexo tal que, se removermos qualquer aresta, ele se torna desconexo.

5. **Entre quaisquer dois vértices de uma árvore existe exatamente um caminho simples**  
Em uma árvore, para cada par de vértices distintos, existe um único caminho simples ligando-os.

# Prova: Toda árvore com n ≥ 2 vértices tem pelo menos dois vértices de grau 1 (Lema do Aperto de Mão)

O lema do aperto de mão afirma que a soma dos graus de todos os vértices de um grafo é igual ao dobro do número de arestas:

Soma dos graus:  
  grau(v₁) + grau(v₂) + ... + grau(vₙ) = 2 × número de arestas

Em uma árvore com n vértices, o número de arestas é n-1:  
  grau(v₁) + grau(v₂) + ... + grau(vₙ) = 2(n-1)

Seja k o número de vértices de grau 1 (folhas) e n-k o número de vértices de grau pelo menos 2.

A soma dos graus pode ser escrita como:  
  k × 1 + soma dos graus dos demais vértices = 2(n-1)

Como cada vértice não folha tem grau pelo menos 2:  
  k + 2(n-k) ≤ 2(n-1)

Desenvolvendo:  
  k + 2n - 2k ≤ 2n - 2  
  2n - k ≤ 2n - 2  
  -k ≤ -2  
  k ≥ 2

Portanto, toda árvore com n ≥ 2 vértices tem pelo menos dois vértices de grau 1.

# Exercício sobre Florestas

## a) Seja G uma floresta com n vértices e k componentes conexas. Quantas arestas existem em G?

Cada componente conexa de uma floresta é uma árvore.  
Uma árvore com $n_i$ vértices tem $n_i - 1$ arestas.  
Se a floresta tem $k$ componentes com $n_1, n_2, ..., n_k$ vértices, então:

- Número total de vértices: $n = n_1 + n_2 + ... + n_k$
- Número total de arestas: $(n_1 - 1) + (n_2 - 1) + ... + (n_k - 1) = (n_1 + n_2 + ... + n_k) - k = n - k$

**Portanto, uma floresta com $n$ vértices e $k$ componentes conexas tem $n - k$ arestas.**

---

## b) Construa (e desenhe) uma floresta com 12 vértices e 9 arestas


````python
import networkx as nx
import matplotlib.pyplot as plt

# Floresta com 12 vértices e 9 arestas (exemplo: 3 árvores)
Gf = nx.Graph()
# Árvore 1: 5 vértices (4 arestas)
Gf.add_edges_from([('A', 'B'), ('B', 'C'), ('C', 'D'), ('D', 'E')])
# Árvore 2: 4 vértices (3 arestas)
Gf.add_edges_from([('F', 'G'), ('G', 'H'), ('H', 'I')])
# Árvore 3: 3 vértices (2 arestas)
Gf.add_edges_from([('J', 'K'), ('K', 'L')])

# Layout manual para separar as componentes
pos = {
    'A': (0, 0), 'B': (1, 0), 'C': (2, 0), 'D': (3, 0), 'E': (4, 0),
    'F': (0, -2), 'G': (1, -2), 'H': (2, -2), 'I': (3, -2),
    'J': (0, -4), 'K': (1, -4), 'L': (2, -4)
}

plt.figure(figsize=(10, 3))
nx.draw(Gf, pos, with_labels=True, node_color='lightblue', edge_color='green')
plt.title("Floresta com 12 vértices e 9 arestas")
plt.show()
````

Observe que uma floresta é, por definição, um **grafo** não conexo, cujas componentes conexas são árvores.  
Ou seja, uma floresta é um único grafo que pode ser formado por várias árvores (cada árvore é uma componente conexa desse grafo).

No exemplo acima, temos uma floresta composta por três árvores (três componentes conexas), mas todas pertencem ao mesmo grafo (objeto NetworkX).  
Cada componente é uma árvore, e o conjunto delas, sem conexões entre si, forma a floresta.

## c) É verdade que qualquer floresta com k componentes tem pelo menos 2k vértices de grau 1?

Sim, é verdade.

Cada componente conexa de uma floresta é uma árvore, e já foi provado que toda árvore com pelo menos 2 vértices tem pelo menos 2 vértices de grau 1.  
Portanto, uma floresta com $k$ componentes tem pelo menos $2k$ vértices de grau 1 (soma das folhas de cada árvore).

Se alguma componente for um único vértice isolado, ele tem grau 0, mas para componentes com pelo menos 2 vértices, sempre há pelo menos 2 folhas.

# Exercício: Possibilidade de grafos com as características dadas

## a) Um grafo simples com 6 vértices, 2 componentes conexas e 11 arestas.

**Não existe.**  
O número máximo de arestas em um grafo simples com 6 vértices é 15 (grafo completo).  
Se houver 2 componentes, cada componente deve ter pelo menos 1 vértice.  
O máximo de arestas ocorre quando as componentes são completas:  
- 1 componente com 5 vértices: até 10 arestas  
- 1 componente com 1 vértice: 0 arestas  
Ou 3 e 3 vértices: cada $K_3$ tem 3 arestas, total 6.  
O máximo possível é 10 arestas (5+1) ou 9 (4+2), nunca 11.  
**Portanto, não existe tal grafo.**

---

## b) Um grafo simples com 7 vértices, 3 componentes conexas e 10 arestas.

**Não existe.**  
O número máximo de arestas ocorre quando as componentes são completas:  
- 5,1,1: $K_5$ (10 arestas), dois vértices isolados (0 arestas): total 10 arestas.  
Isso é possível:  
- Componente 1: $K_5$ (vértices A,B,C,D,E)  
- Componente 2: F  
- Componente 3: G  
**Sim, é possível.**

---

## c) Um grafo simples com 8 vértices, 2 componentes conexas, 9 arestas e exatamente 3 ciclos.

**Não existe.**  
Para 2 componentes, a soma dos ciclos das componentes deve ser 3.  
O número mínimo de arestas para ter 3 ciclos em 8 vértices e 2 componentes é maior que 9.  
Exemplo:  
- Se uma componente é um ciclo $C_4$ (4 vértices, 4 arestas, 1 ciclo), a outra teria 4 vértices e 5 arestas, mas com 4 vértices e 5 arestas, o máximo de ciclos é 2.  
Total: 1+2=3 ciclos, 4+5=9 arestas.  
**Sim, é possível.**

---

## d) Um grafo simples com 8 vértices, 2 componentes conexas, 10 arestas e exatamente 3 ciclos.

**Sim, é possível.**  
Exemplo:  
- Componente 1: $C_4$ (4 vértices, 4 arestas, 1 ciclo)  
- Componente 2: 4 vértices, 6 arestas (um $K_4$ menos uma aresta, que tem 2 ciclos)  
Total: 4+6=10 arestas, 1+2=3 ciclos.

---

## e) Um grafo simples com 9 vértices, 2 componentes conexas, 10 arestas e exatamente 2 ciclos.

**Sim, é possível.**  
Exemplo:  
- Componente 1: $C_4$ (4 vértices, 4 arestas, 1 ciclo)  
- Componente 2: árvore com 5 vértices (4 arestas, 0 ciclos) + mais 2 arestas formando 1 ciclo (total 6 arestas, 1 ciclo)  
Total: 4+6=10 arestas, 1+1=2 ciclos.

---

## f) Um grafo simples conexo com 9 vértices, 12 arestas e exatamente 2 ciclos.

**Sim, é possível.**  
Em um grafo conexo com 9 vértices, uma árvore tem 8 arestas.  
Cada aresta extra adiciona um ciclo.  
12-8=4 arestas extras, mas cada ciclo novo pode compartilhar arestas.  
Para ter exatamente 2 ciclos, é necessário que os ciclos não compartilhem arestas (ciclos disjuntos), ou que as arestas extras formem apenas 2 ciclos.  
Exemplo: Uma árvore + 2 arestas extras formando 2 ciclos (sem sobreposição).  
**Sim, é possível.**

---

## g) Um grafo simples conexo com 9 vértices, 12 arestas e exatamente 3 ciclos.

**Sim, é possível.**  
Árvore: 8 arestas, 3 arestas extras formam 3 ciclos (se não sobreporem).  
**Sim, é possível.**

---

## h) Uma floresta com 10 vértices, 2 componentes conexas e 9 arestas.

**Sim, é possível.**  
Uma floresta com 2 componentes e 10 vértices tem $10-2=8$ arestas se todas forem árvores.  
Com 9 arestas, pelo menos uma componente não é árvore (tem ciclo).  
**Portanto, não é possível uma floresta (todas componentes árvores) com 9 arestas, 10 vértices e 2 componentes.**

---

## i) Uma floresta com 10 vértices, 3 componentes conexas e 9 arestas.

**Sim, é possível.**  
Uma floresta com 3 componentes e 10 vértices tem $10-3=7$ arestas.  
Com 9 arestas, não é floresta (alguma componente tem ciclo).  
**Portanto, não é possível.**

---

## j) Um grafo simples conexo com 11 vértices, 14 arestas e 5 ciclos disjuntos (ciclos que não compartilham arestas).

**Não existe.**  
Uma árvore com 11 vértices tem 10 arestas.  
Cada ciclo disjunto exige pelo menos 3 arestas, então 5 ciclos disjuntos precisam de pelo menos $5 \times 3 = 15$ arestas, além das 10 da árvore, mas algumas arestas podem ser compartilhadas.  
Com apenas 14 arestas, não é possível ter 5 ciclos disjuntos.  
**Portanto, não existe tal grafo.**

````python
def prim_algoritmo_passos(G, raiz):
    """
    Algoritmo de Prim mostrando passo a passo os valores de chave e predecessor (fl)
    para todos os vértices a cada iteração do laço principal.
    """
    chave = {v: float('inf') for v in G.nodes()}
    predecessor = {v: None for v in G.nodes()}
    chave[raiz] = 0
    Q = set(G.nodes())
    passo = 1

    print(f"{'Passo':<5} {'Q':<20} " + "  ".join([f"{v:^10}" for v in sorted(G.nodes())]))
    print(f"{'':<5} {'':<20} " + "  ".join([f"{'chave':^5} {'fl':^4}" for _ in sorted(G.nodes())]))

    while Q:
        # REMOVE-MINIMO(Q) com desempate alfabético
        u = min(Q, key=lambda v: (chave[v], v))
        Q.remove(u)

        # Mostra o estado antes da atualização dos vizinhos
        print(f"{passo:<5} {str(sorted(Q)):<20} " + "  ".join(
            [f"{chave[v]:>5.0f} {predecessor[v] if predecessor[v] is not None else 'raiz':>4}" for v in sorted(G.nodes())]
        ))

        for v in sorted(G.neighbors(u)):  # ordem alfabética dos vizinhos
            if v in Q and G[u][v]['weight'] < chave[v]:
                predecessor[v] = u
                chave[v] = G[u][v]['weight']
        passo += 1

    custo_total = sum(chave[v] for v in G.nodes())
    print("\nResultado final:")
    for v in sorted(G.nodes()):
        fl = predecessor[v] if predecessor[v] is not None else 'raiz'
        print(f"{v}: chave = {chave[v]}, fl = {fl}")
    print(f"Custo total da árvore geradora mínima (Prim): {custo_total}")
    return predecessor, chave, custo_total

# Grafo das localidades e estradas (mesmo exemplo anterior)
G_estradas = nx.Graph()
G_estradas.add_weighted_edges_from([
    ('A', 'B', 30),
    ('A', 'C', 20),
    ('A', 'D', 40),
    ('A', 'E', 10),
    ('B', 'C', 25),
    ('B', 'D', 40),
    ('B', 'E', 30),
    ('C', 'D', 45),
    ('C', 'E', 20),
    ('D', 'E', 35)
])

pos_estradas = {
    'A': (0, 2),
    'B': (-2, 0),
    'C': (0, -2),
    'D': (2, 0),
    'E': (0, 0)
}

# Execução passo a passo do algoritmo de Prim
predecessor, chave, custo_total = prim_algoritmo_passos(G_estradas, raiz='A')

# Visualização da AGM encontrada por Prim
edges_agm = [(predecessor[v], v) for v in G_estradas.nodes() if predecessor[v] is not None]
plt.figure(figsize=(7, 6))
nx.draw(G_estradas, pos_estradas, with_labels=True, node_color='lightblue', edge_color='gray')
nx.draw_networkx_edges(G_estradas, pos_estradas, edgelist=edges_agm, edge_color='red', width=2)
labels = nx.get_edge_attributes(G_estradas, 'weight')
nx.draw_networkx_edge_labels(G_estradas, pos_estradas, edge_labels=labels)
plt.title("Árvore Geradora Mínima encontrada pelo algoritmo de Prim (passo a passo)")
plt.show()
````

````python
def prim_algoritmo_print_passos(G, raiz, pos=None):
    """
    Algoritmo de Prim mostrando passo a passo os valores de chave e predecessor (fl)
    para todos os vértices a cada iteração do laço principal, e plotando o grafo a cada passo.
    """
    import matplotlib.pyplot as plt

    chave = {v: float('inf') for v in G.nodes()}
    predecessor = {v: None for v in G.nodes()}
    chave[raiz] = 0
    Q = set(G.nodes())
    passo = 1

    print(f"{'Passo':<5} {'Q':<30} " + "  ".join([f"{v:^10}" for v in sorted(G.nodes())]))
    print(f"{'':<5} {'':<30} " + "  ".join([f"{'chave':^5} {'fl':^4}" for _ in sorted(G.nodes())]))

    # Print do estado inicial
    print(f"{0:<5} {str(sorted(Q)):<30} " + "  ".join(
        [f"{chave[v]:>5.0f} {predecessor[v] if predecessor[v] is not None else 'raiz':>4}" for v in sorted(G.nodes())]
    ))

    while Q:
        u = min(Q, key=lambda v: (chave[v], v))
        Q.remove(u)

        # Atualiza os vizinhos ANTES do print
        for v in sorted(G.neighbors(u)):
            if v in Q and G[u][v]['weight'] < chave[v]:
                predecessor[v] = u
                chave[v] = G[u][v]['weight']

        print(f"{passo:<5} {str(sorted(Q)):<30} " + "  ".join(
            [f"{chave[v]:>5.0f} {predecessor[v] if predecessor[v] is not None else 'raiz':>4}" for v in sorted(G.nodes())]
        ))

        # Plot do grafo a cada passo
        if pos is not None:
            plt.figure(figsize=(6, 5))
            nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
            labels = nx.get_edge_attributes(G, 'weight')
            nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
            agm_edges = [(predecessor[v], v) for v in G.nodes() if predecessor[v] is not None and v not in Q]
            nx.draw_networkx_edges(G, pos, edgelist=agm_edges, edge_color='red', width=2)
            plt.title(f"Passo {passo} do algoritmo de Prim")
            plt.show()

        passo += 1

    custo_total = sum(chave[v] for v in G.nodes())
    print("\nResultado final:")
    for v in sorted(G.nodes()):
        fl = predecessor[v] if predecessor[v] is not None else 'raiz'
        print(f"{v}: chave = {chave[v]}, fl = {fl}")
    print(f"Custo total da árvore geradora mínima (Prim): {custo_total}")
    return predecessor, chave, custo_total
````

````python
def print_matriz_adjacencia(G):
    """
    Imprime a matriz de adjacência do grafo G, mostrando os pesos das arestas.
    """
    import pandas as pd
    nodes = sorted(G.nodes())
    matriz = []
    for u in nodes:
        linha = []
        for v in nodes:
            if G.has_edge(u, v):
                linha.append(G[u][v].get('weight', 1))
            else:
                linha.append(0)
        matriz.append(linha)
    df = pd.DataFrame(matriz, index=nodes, columns=nodes)
    print("Matriz de adjacência (pesos):")
    display(df)
````
