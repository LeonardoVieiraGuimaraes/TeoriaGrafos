# Representações de Grafo Não Direcionado

## Grafo Não direcionado

```python
import numpy as np
import networkx as nx

# Criando o grafo não direcionado
G_undirected = nx.Graph()
G_undirected.add_edge(0, 1, weight=2.5)
G_undirected.add_edge(0, 2, weight=1.5)
G_undirected.add_edge(1, 2, weight=1.0)
G_undirected.add_edge(2, 3, weight=2.0)

import matplotlib.pyplot as plt

# Plotando o grafo
pos = nx.spring_layout(G_undirected)
labels = nx.get_edge_attributes(G_undirected, 'weight')
nx.draw(G_undirected, pos, with_labels=True, node_color='lightblue', node_size=500, font_size=10)
nx.draw_networkx_edge_labels(G_undirected, pos, edge_labels=labels)
plt.title("Grafo Não Direcionado")
plt.show()
```

## Matriz de Adjacência Não Direcionada

```python
# Matriz de adjacência
adj_matrix_undirected = nx.adjacency_matrix(G_undirected).todense()
print("Matriz de Adjacência Não Direcionada:")
print(adj_matrix_undirected)
```

## Matriz de Incidência Não Direcionada

```python
# Matriz de incidência
inc_matrix_undirected = nx.incidence_matrix(G_undirected, oriented=False).todense()
print("Matriz de Incidência Não Direcionada:")
print(inc_matrix_undirected)
```

```python
# Matriz de Incidência com pesos
inc_matrix_weighted_directed = nx.incidence_matrix(G_undirected, oriented=False, weight='weight').todense()
print("Matriz de Incidência Direcionada (com pesos):")
print(inc_matrix_weighted_directed)
```

```python
# Matriz de incidência com pesos
inc_matrix_weighted = nx.incidence_matrix(G_undirected, oriented=False, weight='weight').todense()
print("Matriz de Incidência Não Direcionada (com pesos):")
print(inc_matrix_weighted)
```

## Lista de Adjacência Não Direcionada

```python
# Lista de adjacência
adj_list_undirected = {node: list(G_undirected.adj[node].items()) for node in G_undirected.nodes()}
print("Lista de Adjacência Não Direcionada:")
for node, neighbors in adj_list_undirected.items():
    print(f"{node}: {[f'(vizinho={neighbor}, peso={data["weight"]})' for neighbor, data in neighbors]}")
```

## Grafo Direcionado

```python
import networkx as nx

# Criando o grafo direcionado
G_directed = nx.DiGraph()
G_directed.add_edge(0, 1, weight=2.5)
G_directed.add_edge(0, 2, weight=1.5)
G_directed.add_edge(1, 2, weight=1.0)
G_directed.add_edge(2, 3, weight=2.0)

import matplotlib.pyplot as plt

# Plotando o grafo direcionado
pos = nx.spring_layout(G_directed)
labels = nx.get_edge_attributes(G_directed, 'weight')
nx.draw(G_directed, pos, with_labels=True, node_color='lightgreen', node_size=500, font_size=10, arrowsize=20)
nx.draw_networkx_edge_labels(G_directed, pos, edge_labels=labels)
plt.title("Grafo Direcionado")
plt.show()
```

## Matriz de Adjacência Direcionada

```python
# Matriz de adjacência
adj_matrix_directed = nx.adjacency_matrix(G_directed).todense()
print("Matriz de Adjacência Direcionada:")
print(adj_matrix_directed)
```

## Matriz de Incidência Direcionada

```python
# Matriz de incidência
inc_matrix_directed = nx.incidence_matrix(G_directed, oriented=True).todense()
print("Matriz de Incidência Direcionada:")
print(inc_matrix_directed)
```

```python
# Matriz de incidência com pesos
inc_matrix_weighted_directed = nx.incidence_matrix(G_directed, oriented=True, weight='weight').todense()
print("Matriz de Incidência Direcionada (com pesos):")
print(inc_matrix_weighted_directed)
```

## Lista de Adjacência Direcionada

```python
# Lista de adjacência ponderada (direcionada)
adj_list_directed = {node: list(G_directed.adj[node].items()) for node in G_directed.nodes()}
print("Lista de Adjacência Direcionada (com pesos):")
for node, neighbors in adj_list_directed.items():
    print(f"{node}: {[f'(vizinho={neighbor}, peso={data.get("weight", "N/A")})' for neighbor, data in neighbors]}")
```
