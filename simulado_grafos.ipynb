## Questão 01 — Árvore

Uma empresa de logística deseja construir uma rede de estradas para conectar 6 cidades de forma eficiente. A rede deve ser construída de forma que:
1. Todas as cidades estejam conectadas diretamente ou indiretamente.
2. Não existam ciclos na rede.
3. A remoção de qualquer estrada desconecte a rede.

Com base nas condições acima e nas definições de árvores, qual das opções abaixo representa corretamente o número de estradas necessárias para conectar as 6 cidades?

Alternativas

a) 4  
b) 5  
c) 6  
d) 7  
e) 9  

**Gabarito:** b

**Feedback:**  
A resposta correta é b) 5 porque, de acordo com o Teorema 1, uma árvore com n vértices possui exatamente n - 1 arestas. Para conectar 6 cidades n = 6, são necessárias 6 - 1 = 5 estradas.
