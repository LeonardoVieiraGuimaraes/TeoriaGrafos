
 ## 1. Dado o grafo G1 abaixo, que representa conexões entre amigos em uma rede social, desenhe-o:

* G1 = ({A, B, C, D, E, F}, {(A, B), (A, C), (B, D), (C, E), (D, E), (E, F)})

```python
import networkx as nx
import matplotlib.pyplot as plt
# Criação do grafo
G = nx.MultiGraph()  # Usando MultiGraph para permitir arestas múltiplas

# Adicionando nós (representando as áreas da cidade)
G.add_nodes_from(["A", "B", "C", "D", "E", "F"])

# Adicionando arestas (representando as pontes)
G.add_edges_from([
    ("A", "B"),
    ("A", "C"),
    ("B", "D"),
    ("C", "E"),
    ("D", "E"),
    ("E", "F")
])
```

```python
# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_size=700, node_color="green", font_size=12, font_weight="bold")
```

```python
# Criando Grafo Digrafo
G1 = nx.DiGraph()

# Adicionando nós (representando as áreas da cidade)
G1.add_nodes_from(["CEO", "Gerente1", "Gerente2", "Func1", "Func2"])

# Adicionando arestas (representando as pontes)
G1.add_edges_from([
    ("CEO", "Gerente1"), ("CEO",
"Gerente2"), ("Gerente1", "Func1"), ("Gerente2", "Func2")
])
```

```python
# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G1)
nx.draw(G1, pos, with_labels=True, node_size=700, node_color="green", font_size=12, font_weight="bold")
```

```python
# Criando Grafo Digrafo
G1 = nx.DiGraph()


# Adicionando arestas (representando as pontes)
G1.add_edges_from([
    ('A', 'B'), ('B', 'C'), ('C', 'D')
])

G2 = nx.DiGraph()

G2.add_edges_from([
    ('B', 'A'), ('C', 'B'), ('D', 'C')
])

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G1)
nx.draw(G1, pos, with_labels=True, node_size=700, node_color="green", font_size=12, font_weight="bold")

# Desenhando o grafo
plt.figure(figsize=(8, 6))
pos = nx.spring_layout(G1)
nx.draw(G2, pos, with_labels=True, node_size=700, node_color="green", font_size=12, font_weight="bold")
```
