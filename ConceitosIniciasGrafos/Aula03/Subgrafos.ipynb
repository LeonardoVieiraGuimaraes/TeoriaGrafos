## Subgrafos
Um subgrafo é um grafo cujos vértices e arestas são subconjuntos dos vértices e arestas de um grafo maior. 

### Definição
Dado um grafo $G = (V, E)$, um subgrafo $H = (V_H, E_H)$ é tal que $V_H \subseteq V$ e $E_H \subseteq E$. Ou seja, os vértices $V_H$ e as arestas $E_H$ de $H$ são subconjuntos dos vértices $V$ e das arestas $E$ de $G$, respectivamente.



```python
# Função para desenhar o grafo
import matplotlib.pyplot as plt
import networkx as nx

def draw_graph_with_labels_outside(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    plt.title(title)
    plt.show()

    # Função para desenhar o grafo com bordas pretas, nós brancos e rótulos pretos fora das bolinhas
def draw_graph(G, title, node_color='lightblue'):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color=node_color, edge_color='gray', node_size=500, font_size=15)
    plt.title(title)
    plt.show()
```

```python
# Criar o grafo original G
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4)])

# Definir os vértices e arestas do subgrafo H
V_H = [0, 1, 2]
E_H = [(0, 1), (1, 2)]

# Criar o subgrafo H
H = nx.Graph()
H.add_nodes_from(V_H)
H.add_edges_from(E_H)


# Desenhar os grafos
plt.figure(figsize=(10, 5))

plt.subplot(121)
draw_graph_with_labels_outside(G, 'Grafo Original G')

plt.subplot(122)
draw_graph_with_labels_outside(H, 'Subgrafo H')

```

### Tipos de Subgrafos

- **Subgrafo Próprio**: Um subgrafo que não é igual ao grafo original. 
  - Um subgrafo $H = (V_H, E_H)$ de um grafo $G = (V, E)$ que não é igual ao grafo original, ou seja, $H \subset G$.

- **Subgrafo Induzido**: Um subgrafo formado por um subconjunto de vértices do grafo original e todas as arestas entre esses vértices no grafo original.
  - Um subgrafo $H = (V_H, E_H)$ de um grafo $G = (V, E)$ onde $V_H \subseteq V$ e $E_H$ consiste de todas as arestas $(u, v) \in E$ tais que $u, v \in V_H$. Formalmente, $E_H = \{(u, v) \in E \mid u, v \in V_H\}$.

- **Subgrafo Espanning**: Um subgrafo que inclui todos os vértices do grafo original.
  - Um subgrafo $H = (V_H, E_H)$ de um grafo $G = (V, E)$ que inclui todos os vértices do grafo original, ou seja, $V_H = V$.



### Exemplo de Subgrafo Próprio

```python
# Criar o grafo original
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (2, 3)])

# Criar um subgrafo próprio
subgraph = G.subgraph([0, 1, 2])

# Desenhar o subgrafo
draw_graph_with_labels_outside(subgraph, 'Subgrafo Próprio')
```

### Exemplo de Subgrafo Induzido

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph_with_labels_outside(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    plt.title(title)
    plt.show()

# Criar o grafo original
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (2, 3)])

# Criar um subgrafo induzido
induced_subgraph = G.subgraph([1, 2, 3])

# Desenhar o subgrafo induzido
draw_graph_with_labels_outside(induced_subgraph, 'Subgrafo Induzido')
```

### Exemplo de Subgrafo Espanning

```python
import networkx as nx
import matplotlib.pyplot as plt

def draw_graph_with_labels_outside(G, title):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=False, node_color='black', edge_color='black', node_size=800, linewidths=2)
    nx.draw(G, pos, with_labels=False, node_color='white', edge_color='black', node_size=600, linewidths=2)
    nx.draw_networkx_labels(G, pos, font_size=20, font_color='black')
    plt.title(title)
    plt.show()

# Criar o grafo original
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (2, 3)])

# Criar um subgrafo espanning (neste caso, é o próprio grafo original)
spanning_subgraph = G.subgraph(G.nodes())

# Desenhar o subgrafo espanning
draw_graph_with_labels_outside(spanning_subgraph, 'Subgrafo Espanning')
```

### Vértices e Arestas Disjuntas

Dois subgrafos de um grafo $G$, $G_1 = (V_1, E_1)$ e $G_2 = (V_2, E_2)$, são **aresta-disjuntos** se eles não possuem arestas em comum, ou seja, $E_1 \cap E_2 = \emptyset$. Se $V_1 \cap V_2 = \emptyset$, os dois subgrafos são chamados de **vértice-disjuntos**.

### Exemplo de Vértices e Arestas Disjuntas

```python
import networkx as nx
import matplotlib.pyplot as plt


# Criar o grafo original
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4), (4, 5)])

# Criar dois subgrafos aresta-disjuntos
G1 = G.edge_subgraph([(0, 1), (1, 2)])
G2 = G.edge_subgraph([(3, 4), (4, 5)])

# Criar dois subgrafos vértice-disjuntos
G3 = G.subgraph([0, 1, 2])
G4 = G.subgraph([3, 4, 5])

# Desenhar os grafos


draw_graph_with_labels_outside(G, 'Grafo Original G')

draw_graph_with_labels_outside(G1, 'Subgrafo Aresta-Disjunto G1')

draw_graph_with_labels_outside(G2, 'Subgrafo Aresta-Disjunto G2')

draw_graph_with_labels_outside(G3, 'Subgrafo Vértice-Disjunto G3')

draw_graph_with_labels_outside(G4, 'Subgrafo Vértice-Disjunto G4')
```

### Aplicações de Subgrafos

Subgrafos são usados em várias áreas, incluindo:
- **Redes de Computadores**: Para analisar sub-redes dentro de uma rede maior.
- **Biologia Computacional**: Para estudar subestruturas em redes biológicas.
- **Teoria dos Grafos**: Para provar propriedades de grafos maiores através de seus subgrafos.
