Exercício de Grafos Isomorfos
Verifique se os grafos abaixo são isomorfos. Se forem, encontre uma correspondência entre os vértices dos dois grafos.
Grafo G1:
Vértices: {A, B, C, D}
Arestas: {(A, B), (A, C), (B, D), (C, D)}
Grafo G2:
Vértices: {1, 2, 3, 4}
Arestas: {(1, 2), (1, 3), (2, 4), (3, 4)}

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar grafos com arestas paralelas
def desenhar_grafo_com_arestas_paralelas(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Definindo o grafo G1 como MultiGraph para permitir arestas paralelas
G1 = nx.MultiGraph()
G1.add_edges_from([(1, 2), (1, 5), (1, 3), (1, 3), (2, 3), (3, 3), (3, 4), (3, 5), (4, 5)])
desenhar_grafo_com_arestas_paralelas(G1, "Grafo G1")

# Definindo o grafo G2 como MultiGraph para permitir arestas paralelas
G2 = nx.MultiGraph()
G2.add_edges_from([('a', 'b'), ('a', 'e'), ('b', 'c'), ('b', 'e'), ('c', 'd'), ('c', 'e'), ('c', 'e'), ('d', 'e'), ('e', 'e')])
desenhar_grafo_com_arestas_paralelas(G2, "Grafo G2")

# Verificando se os grafos são isomorfos
GM = nx.isomorphism.GraphMatcher(G1, G2)
is_isomorphic = GM.is_isomorphic()
print(f"Os grafos G1 e G2 são isomorfos? {is_isomorphic}")

# Informações adicionais sobre os grafos
print(f"Número de vértices em G1: {G1.number_of_nodes()}")
print(f"Número de vértices em G2: {G2.number_of_nodes()}")
print(f"Número de arestas em G1: {G1.number_of_edges()}")
print(f"Número de arestas em G2: {G2.number_of_edges()}")

# Calculando o grau de cada vértice e a sequência de graus dos vértices
graus_G1 = {n: d for n, d in G1.degree()}
graus_G2 = {n: d for n, d in G2.degree()}

print("Grau de cada vértice em G1:")
for vertice, grau in graus_G1.items():
    print(f"Vértice {vertice}: {grau}")

print("Grau de cada vértice em G2:")
for vertice, grau in graus_G2.items():
    print(f"Vértice {vertice}: {grau}")

sequencia_graus_G1 = sorted(graus_G1.values())
sequencia_graus_G2 = sorted(graus_G2.values())

print(f"Sequência de graus dos vértices em G1: {sequencia_graus_G1}")
print(f"Sequência de graus dos vértices em G2: {sequencia_graus_G2}")

# Encontrando a correspondência entre os vértices
if is_isomorphic:
    mapping = GM.mapping
    print("Correspondência entre os vértices:")
    for k, v in mapping.items():
        print(f"{k} -> {v}")
```

```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.drawing.nx_pydot import pydot_layout

# Função para desenhar grafos
def desenhar_grafo(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Definindo o grafo G
G = nx.MultiGraph()
G.add_edges_from([('u', 'v'), ('u', 'v'), ('u', 'x'), ('v', 'w'), ('w', 'w'), ('w', 'x')])
desenhar_grafo(G, "Grafo G")

# Definindo o subgrafo S1
S1 = nx.MultiGraph()
S1.add_edges_from([('u', 'v'), ('u', 'v')])
desenhar_grafo(S1, "Subgrafo S1")

# Definindo o subgrafo S2
S2 = nx.MultiGraph()
S2.add_edges_from([('u', 'v'), ('v', 'w')])
desenhar_grafo(S2, "Subgrafo S2")

# Definindo o subgrafo S3
S3 = nx.MultiGraph()
S3.add_edges_from([('u', 'v'), ('u', 'v'), ('u', 'x'), ('v', 'w'), ('w', 'w'), ('w', 'x')])
desenhar_grafo(S3, "Subgrafo S3")

# Verificando se os subgrafos são subgrafos de G
is_subgraph_S1 = nx.subgraph(G, S1)
is_subgraph_S2 = nx.subgraph(G, S2)
is_subgraph_S3 = nx.subgraph(G, S3)

print(f"O subgrafo S1 {is_subgraph_S1}")
print(f"O subgrafo S2 {is_subgraph_S2}")
print(f"O subgrafo S3 {is_subgraph_S3}")

# Função para verificar se um grafo é subgrafo de outro
def verificar_subgrafo(G, S):
    return all(edge in G.edges() for edge in S.edges()) and all(node in G.nodes() for node in S.nodes())

is_subgraph_S1 = verificar_subgrafo(G, S1)
is_subgraph_S2 = verificar_subgrafo(G, S2)
is_subgraph_S3 = verificar_subgrafo(G, S3)

print(f"O subgrafo S1 é subgrafo de G? {is_subgraph_S1}")
print(f"O subgrafo S2 é subgrafo de G? {is_subgraph_S2}")
print(f"O subgrafo S3 é subgrafo de G? {is_subgraph_S3}")
```

# Operadores em Grafos

## Funções

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar grafos
def desenhar_grafo(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Função para subdividir arestas
def subdividir_arestas(G, edges):
    G_subdividido = G.copy()
    for u, v in edges:
        if G_subdividido.has_edge(u, v):
            G_subdividido.remove_edge(u, v)
            novo_no = max(G_subdividido.nodes) + 1
            G_subdividido.add_edge(u, novo_no)
            G_subdividido.add_edge(novo_no, v)
    return G_subdividido
```

## Grafos G1 e G2

```python
# Definindo os grafos G1 e G2
G1 = nx.Graph()
G1.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 4)])
print(f"Grafo G1: Vértices = {list(G1.nodes)}, Arestas = {list(G1.edges)}")
desenhar_grafo(G1, "Grafo G1")

G2 = nx.Graph()
G2.add_edges_from([(3, 4), (4, 5), (5, 6)])
print(f"Grafo G2: Vértices = {list(G2.nodes)}, Arestas = {list(G2.edges)}")
desenhar_grafo(G2, "Grafo G2")
```

## União

```python
# União
G_union = nx.compose(G1, G2)
print(f"União: Vértices = {list(G_union.nodes)}, Arestas = {list(G_union.edges)}")
desenhar_grafo(G_union, "União de G1 e G2")
```

```python
# Interseção
G_intersection = nx.intersection(G1, G2)
print(f"Interseção: Vértices = {list(G_intersection.nodes)}, Arestas = {list(G_intersection.edges)}")
desenhar_grafo(G_intersection, "Interseção de G1 e G2")
```

## Junção

```python
#  Junção (Soma)
G_join = nx.Graph()
G_join.add_nodes_from(G1.nodes)
G_join.add_nodes_from(G2.nodes)
G_join.add_edges_from(G1.edges)
G_join.add_edges_from(G2.edges)

# Adicionando arestas entre todos os pares de vértices que pertencem a diferentes grafos originais
for u in G1.nodes:
    for v in G2.nodes:
        G_join.add_edge(u, v)

print(f"Junção: Vértices = {list(G_join.nodes)}, Arestas = {list(G_join.edges)}")
desenhar_grafo(G_join, "Junção de G1 e G2")
```

## Decomposição

```python
# Decomposição
def decompor_grafo(G):
    decomposicao = []
    for edge in G.edges:
        subgrafo = nx.Graph()
        subgrafo.add_edge(*edge)
        decomposicao.append(subgrafo)
    return decomposicao
# Decompondo o grafo G2
decomposicao_G2 = decompor_grafo(G2)
for i, subgrafo in enumerate(decomposicao_G2):
    print(f"Subgrafo {i+1} de G2: Vértices = {list(subgrafo.nodes)}, Arestas = {list(subgrafo.edges)}")
    desenhar_grafo(subgrafo, f"Subgrafo {i+1} de G2")
```

## Matriz de Incidência não direcionada

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Função para desenhar grafos
def desenhar_grafo(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    edge_labels = {(u, v): f"{u}-{v}" for u, v in G.edges()}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(titulo)
    plt.show()

# Definindo o grafo não direcionado
G = nx.Graph()
G.add_edges_from([(0, 1), (0, 2), (1, 2), (2, 3)])
print(f"Grafo: Vértices = {list(G.nodes)}, Arestas = {list(G.edges)}")
desenhar_grafo(G, "Grafo Não Direcionado Exemplo")

# Obtendo a matriz de incidência
matriz_incidencia = nx.incidence_matrix(G, oriented=False).todense()
print("Matriz de Incidência:")
print(matriz_incidencia)

# Exibindo a matriz de incidência de forma mais legível
def exibir_matriz_incidencia(G, matriz):
    arestas = list(G.edges())
    print("Matriz de Incidência:")
    print("   ", end="")
    for i, aresta in enumerate(arestas):
        print(f"{aresta} ", end="")
    print()
    for i, linha in enumerate(matriz):
        print(f"{i}: ", end="")
        for valor in linha:
            print(f"  {valor}  ", end="")
        print()

exibir_matriz_incidencia(G, matriz_incidencia)

```

## Matriz de Incidência Ponderada

```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np

# Função para desenhar grafos
def desenhar_grafo(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15, arrows=True)
    edge_labels = {(u, v): f"{u}->{v} ({d['weight']})" for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(titulo)
    plt.show()

# Definindo o grafo direcionado e ponderado
G = nx.DiGraph()
G.add_weighted_edges_from([(0, 1, 2.5), (0, 2, 1.5), (1, 2, 1.0), (2, 3, 2.0)])
print(f"Grafo: Vértices = {list(G.nodes)}, Arestas = {list(G.edges(data=True))}")
desenhar_grafo(G, "Grafo Direcionado e Ponderado Exemplo")

# Obtendo a matriz de incidência ponderada
def matriz_incidencia_ponderada(G):
    n = len(G.nodes)
    m = len(G.edges)
    matriz = np.zeros((n, m))
    for j, (u, v, d) in enumerate(G.edges(data=True)):
        matriz[u, j] = d['weight']
        matriz[v, j] = -d['weight']
    return matriz

matriz_incidencia = matriz_incidencia_ponderada(G)
print("Matriz de Incidência Ponderada:")
print(matriz_incidencia)

# Exibindo a matriz de incidência de forma mais legível
def exibir_matriz_incidencia(G, matriz):
    arestas = list(G.edges(data=True))
    print("Matriz de Incidência Ponderada:")
    print("   ", end="")
    for i, aresta in enumerate(arestas):
        print(f"{aresta[:2]} ", end="")
    print()
    for i, linha in enumerate(matriz):
        print(f"{i}: ", end="")
        for valor in linha:
            print(f"  {valor:.1f}  ", end="")
        print()

exibir_matriz_incidencia(G, matriz_incidencia)
```

## Lista de Incidência Direncionada Ponderada

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar grafos
def desenhar_grafo(G, titulo):
    pos = nx.spring_layout(G)
    nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15, arrows=True)
    edge_labels = {(u, v): f"{u}->{v} ({d['weight']})" for u, v, d in G.edges(data=True)}
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels)
    plt.title(titulo)
    plt.show()

# Definindo o grafo direcionado e ponderado
G = nx.DiGraph()
G.add_weighted_edges_from([(0, 1, 2.5), (0, 2, 1.5), (1, 2, 1.0), (2, 3, 2.0)])
print(f"Grafo: Vértices = {list(G.nodes)}, Arestas = {list(G.edges(data=True))}")
desenhar_grafo(G, "Grafo Direcionado e Ponderado Exemplo")

# Obtendo a lista de adjacência
lista_adjacencia = {n: [(v, G[n][v]['weight']) for v in G.successors(n)] for n in G.nodes()}
print("Lista de Adjacência Ponderada:")
for vertice, adjacentes in lista_adjacencia.items():
    print(f"{vertice}: {adjacentes}")

# Obtendo a lista de incidência
lista_incidencia = {n: [(u, v, G[u][v]['weight']) for u, v in G.in_edges(n)] for n in G.nodes()}
print("Lista de Incidência Ponderada:")
for vertice, arestas in lista_incidencia.items():
    print(f"{vertice}: {arestas}")
```
