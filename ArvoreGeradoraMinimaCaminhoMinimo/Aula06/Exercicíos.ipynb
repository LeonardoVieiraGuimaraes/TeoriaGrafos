|-----|---|---|---|---|---|---|---|---|---|---|
| ρ(v)| b | c | d | e | h | d | nil | g | j | h |

a) Caminho `g → c`:

b) Caminho `b → f`:

c) Caminho `g → i`:

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para calcular o caminho com base no vetor de roteamento
def calcular_caminho(vetor_roteamento, inicio, destino):
    caminho = [inicio]
    while inicio != destino:
        proximo = vetor_roteamento.get(inicio)
        if proximo is None or proximo == 'nil':
            return 'Sem caminho'
        caminho.append(proximo)
        inicio = proximo
    return caminho

# Vetor de roteamento representado como um dicionário
vetor_roteamento = {
    'a': 'b', 'b': 'c', 'c': 'd', 'd': 'e', 'e': 'h',
    'f': 'd', 'g': None, 'h': 'g', 'i': 'j', 'j': 'h'
}

# Criar o grafo com base no vetor de roteamento
G = nx.DiGraph()
arestas = [(v, destino) for v, destino in vetor_roteamento.items() if destino and destino != 'nil']
G.add_edges_from(arestas)

# Plotar o grafo
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10, arrowsize=15)
plt.title('Grafo Correspondente ao Vetor de Roteamento')
plt.show()

# Cálculo dos caminhos
caminho_g_c = calcular_caminho(vetor_roteamento, 'g', 'c')
caminho_b_f = calcular_caminho(vetor_roteamento, 'b', 'f')
caminho_g_i = calcular_caminho(vetor_roteamento, 'g', 'i')

# Exibir os resultados
print(f'Caminho g → c: {caminho_g_c}')
print(f'Caminho b → f: {caminho_b_f}')
print(f'Caminho g → i: {caminho_g_i}')
```

6. Elabore um algoritmo para verificar se um grafo possui ciclos.

```python
def possui_ciclo_puro(grafo):
    visitados = set()
    ciclo = []

    def dfs(no, pai):
        visitados.add(no)
        for vizinho in grafo[no]:
            if vizinho not in visitados:
                if dfs(vizinho, no):
                    ciclo.append(vizinho)
                    return True
            elif vizinho != pai:
                ciclo.append(vizinho)
                return True
        return False

    for vertice in grafo:
        if vertice not in visitados:
            if dfs(vertice, None):
                ciclo.append(vertice)
                return ciclo[::-1]  # Retorna o ciclo encontrado
    return None

# Exemplo de grafo com ciclos
grafo = {
    'A': ['B', 'D'],
    'B': ['A', 'C'],
    'C': ['B', 'D', 'E'],
    'D': ['A', 'C'],
    'E': ['C']
}

# Verificar se o grafo possui ciclos e retornar o ciclo
ciclo = possui_ciclo_puro(grafo)
if ciclo:
    print(f'Python puro: O grafo possui ciclos: {ciclo}')
else:
    print('Python puro: O grafo não possui ciclos.')
```

7. Implemente o algoritmo de busca em largura (algoritmo 3.1).

```python
### Resolução da Questão 7: Implementação do Algoritmo de Busca em Largura (BFS)
from collections import deque

def busca_em_largura(grafo, inicio):
    visitados = set()
    fila = deque([inicio])
    ordem_visita = []

    while fila:
        vertice = fila.popleft()
        if vertice not in visitados:
            visitados.add(vertice)
            ordem_visita.append(vertice)
            fila.extend(v for v in grafo[vertice] if v not in visitados)

    return ordem_visita

# Exemplo de grafo
grafo = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

# Executar a busca em largura a partir do vértice 'A'
resultado_bfs = busca_em_largura(grafo, 'A')
print(f'Ordem de visita (BFS) a partir do vértice A: {resultado_bfs}')
```

### Implementação do Algoritmo de Busca em Largura (BFS) com Estados, Predecessores e Distâncias

```python
from collections import deque

def busca_em_largura(grafo, raiz):
    estado = {v: "NAO_VISITADO" for v in grafo}
    fl = {v: None for v in grafo}
    d = {v: float("inf") for v in grafo}
    
    d[raiz] = 0
    estado[raiz] = "VISITADO"
    
    fila = deque([raiz])
    
    while fila:
        vi = fila.popleft()

        for vj in grafo[vi]:
            if estado[vj] == "NAO_VISITADO":
                fila.append(vj)
                estado[vj] = "VISITADO"
                fl[vj] = vi
                d[vj] = d[vi] + 1
        
        estado[vi] = "ENCERRADO"

    return d, fl

# Exemplo de uso Lista de Adjacência
grafo_exemplo = {
    'A': ['B', 'C'],
    'B': ['A', 'D', 'E'],
    'C': ['A', 'F'],
    'D': ['B'],
    'E': ['B', 'F'],
    'F': ['C', 'E']
}

G.exemplo = nx.Graph(grafo_exemplo)
# Desenhar o grafo
pos = nx.spring_layout(G.exemplo)
nx.draw(G.exemplo, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=500, font_size=10)
plt.title('Grafo Exemplo')
plt.show()

raiz = 'A'
distancias, predecessores = busca_em_largura(grafo_exemplo, raiz)

print("Distâncias:", distancias)
print("Predecessores:", predecessores)
```

```python
# Matrins de Adjacencia

def bfs_matriz_adjacencia(matriz, raiz):
    n = len(matriz)
    estado = ["NAO_VISITADO"] * n
    distancia = [float("inf")] * n
    predecessores = [None] * n

    estado[raiz] = "VISITADO"
    distancia[raiz] = 0
    fila = deque([raiz])

    while fila:
        vi = fila.popleft()
        for vj in range(n):
            if matriz[vi][vj] == 1 and estado[vj] == "NAO_VISITADO":
                fila.append(vj)
                estado[vj] = "VISITADO"
                distancia[vj] = distancia[vi] + 1
                predecessores[vj] = vi
        estado[vi] = "ENCERRADO"

    return distancia, predecessores
```

```python
def busca_em_profundidade(grafo, raiz):
    estado = {v: "NAO_VISITADO" for v in grafo}
    fl = {v: None for v in grafo}
    tempo = 0
    ta = {}
    te = {}

    def visita_vertice(vi):
        nonlocal tempo
        estado[vi] = "VISITADO"
        tempo += 1
        ta[vi] = tempo
        
        for vj in grafo[vi]:
            if estado[vj] == "NAO_VISITADO":
                fl[vj] = vi
                visita_vertice(vj)

        estado[vi] = "ENCERRADO"
        tempo += 1
        te[vi] = tempo

    visita_vertice(raiz)
    
    return ta, te, fl

# Exemplo de uso com lista de adjacência
grafo_exemplo = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}

raiz = 'A'
tempos_inicio, tempos_fim, predecessores = busca_em_profundidade(grafo_exemplo, raiz)

print("Tempos de início:", tempos_inicio)
print("Tempos de término:", tempos_fim)
print("Predecessores:", predecessores)
```

1. **Busca em Largura (BFS)**:  
   - Prioriza explorar primeiro os vizinhos mais próximos do vértice inicial.  
   - O número atribuído na BFS, **l(x)**, está diretamente ligado à distância do vértice inicial **v** até **x**.

2. **Busca em Profundidade (DFS)**:  
   - Segue um caminho profundo antes de retroceder.  
   - O número atribuído na DFS, **p(x)**, pode não estar diretamente relacionado à distância ao vértice inicial.

- **Alternativa (a)**:

- **Alternativa (b)**:  
  Se **l(u) < l(w)** e **p(u) > p(w)**, então **d(v, u) = d(v, w)**.

  → **Falsa**, pois a BFS sempre visita primeiro os vértices mais próximos.

**Letra (b)** → A ordem de visitação na BFS garante que **d(v, u) = d(v, w)**.
