# Árvores em Teoria de Grafos
Uma **árvore** é um tipo especial de grafo que possui as seguintes propriedades:
- É um grafo conexo.
- Não contém ciclos.
- Para um grafo com  n  vértices, uma árvore possui exatamente  n-1  arestas.

As árvores são amplamente utilizadas em ciência da computação, como em algoritmos de busca, estruturas de dados (árvores binárias, AVL, etc.), e redes de comunicação.

## Propriedades de Árvores
1. Uma árvore com  n  vértices possui exatamente  n-1  arestas.
2. Qualquer grafo conexo com  n-1  arestas é uma árvore.
3. Uma árvore é um grafo minimamente conexo, ou seja, a remoção de qualquer aresta desconecta o grafo.
4. Entre dois vértices de uma árvore, existe exatamente um caminho simples.

## Exemplo de Implementação de Árvores em Python

```python
import networkx as nx
import matplotlib.pyplot as plt

# Criando uma árvore
T = nx.Graph()
T.add_edges_from([(1, 2), (1, 3), (2, 4), (2, 5), (3, 6), (3, 7)])

# Desenhando a árvore
nx.draw(T, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=2000, font_size=15)
plt.title("Árvore Exemplo")
plt.show()

# Verificando se o grafo é uma árvore
is_tree = nx.is_tree(T)
print(f"O grafo é uma árvore? {is_tree}")
```

## Busca em Árvores
Árvores são frequentemente usadas em algoritmos de busca, como:
- **Busca em Profundidade (DFS)**
- **Busca em Largura (BFS)**

```python
# Busca em Profundidade (DFS)
dfs = list(nx.dfs_edges(T, source=1))
print("Busca em Profundidade (DFS):", dfs)

# Busca em Largura (BFS)
bfs = list(nx.bfs_edges(T, source=1))
print("Busca em Largura (BFS):", bfs)
```

## Árvores Geradoras Mínimas
Uma **árvore geradora mínima** (MST - Minimum Spanning Tree) é uma subárvore de um grafo ponderado que conecta todos os vértices com o menor peso total possível.
### Algoritmos para MST:
- **Kruskal**
- **Prim**

```python
# Criando um grafo ponderado
G = nx.Graph()
G.add_weighted_edges_from([
    (1, 2, 4), (1, 3, 3), (2, 3, 2), (2, 4, 5), (3, 4, 1)
])

# Desenhando o grafo ponderado
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
labels = nx.get_edge_attributes(G, 'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)
plt.title("Grafo Ponderado")
plt.show()

# Calculando a Árvore Geradora Mínima (MST) usando Kruskal
mst = nx.minimum_spanning_tree(G, algorithm='kruskal')
nx.draw(mst, with_labels=True, node_color='lightgreen', edge_color='gray', node_size=2000, font_size=15)
plt.title("Árvore Geradora Mínima (Kruskal)")
plt.show()
```

# Caracterização de árvores


```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar uma árvore
def desenhar_arvore(grafo, titulo):
    pos = nx.spring_layout(grafo)  # Layout para melhor visualização
    nx.draw(grafo, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Criando árvores não rotuladas com até 5 vértices
# Árvore 1: Apenas 1 vértice
arvore1 = nx.Graph()
arvore1.add_node(1)
desenhar_arvore(arvore1, "Árvore com 1 vértice")

# Árvore 2: 2 vértices conectados
arvore2 = nx.Graph()
arvore2.add_edges_from([(1, 2)])
desenhar_arvore(arvore2, "Árvore com 2 vértices")

# Árvore 3: 3 vértices em linha
arvore3 = nx.Graph()
arvore3.add_edges_from([(1, 2), (2, 3)])
desenhar_arvore(arvore3, "Árvore com 3 vértices")

# Árvore 4: 4 vértices em forma de estrela
arvore4 = nx.Graph()
arvore4.add_edges_from([(1, 2), (1, 3), (1, 4)])
desenhar_arvore(arvore4, "Árvore com 4 vértices (estrela)")

# Árvore 5: 5 vértices em forma de linha
arvore5 = nx.Graph()
arvore5.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])
desenhar_arvore(arvore5, "Árvore com 5 vértices (linha)")

# Árvore 6: 5 vértices com um nó central e dois ramos
arvore6 = nx.Graph()
arvore6.add_edges_from([(1, 2), (1, 3), (3, 4), (3, 5)])
desenhar_arvore(arvore6, "Árvore com 5 vértices (nó central com ramos)")
```

## Proposição 2: qualquer par de vértices em uma árvore está conectado por exatamente um caminho.


```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar uma árvore
def desenhar_arvore(grafo, titulo):
    pos = nx.spring_layout(grafo)  # Layout para melhor visualização
    nx.draw(grafo, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Criando uma árvore inicial com 3 vértices
arvore = nx.Graph()
arvore.add_edges_from([(1, 2), (2, 3)])
desenhar_arvore(arvore, "Árvore inicial com 3 vértices")

# Adicionando um novo vértice (4) conectado ao vértice 2
arvore.add_edge(2, 4)
desenhar_arvore(arvore, "Árvore com 4 vértices (novo vértice conectado ao vértice 2)")

# Verificando os caminhos entre todos os pares de vértices
for u in arvore.nodes:
    for v in arvore.nodes:
        if u != v:
            caminho = nx.shortest_path(arvore, source=u, target=v)
            print(f"Caminho entre {u} e {v}: {caminho}")
```

## Proposição 3: a remoção de qualquer aresta de uma árvore desconecta a árvore.

```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar uma árvore
def desenhar_arvore(grafo, titulo):
    pos = nx.spring_layout(grafo)  # Layout para melhor visualização
    nx.draw(grafo, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Criando uma árvore inicial
arvore = nx.Graph()
arvore.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])
desenhar_arvore(arvore, "Árvore inicial com 5 vértices")

# Removendo uma aresta
arvore.remove_edge(3, 4)
desenhar_arvore(arvore, "Árvore após remoção da aresta (3, 4)")

# Verificando se o grafo ainda é conexo
is_connected = nx.is_connected(arvore)
print(f"O grafo é conexo após a remoção da aresta? {is_connected}")
```

## Proposição 4: a adição de uma aresta em uma árvore cria um ciclo.


```python
import networkx as nx
import matplotlib.pyplot as plt

# Função para desenhar uma árvore
def desenhar_arvore(grafo, titulo):
    pos = nx.spring_layout(grafo)  # Layout para melhor visualização
    nx.draw(grafo, pos, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=15)
    plt.title(titulo)
    plt.show()

# Criando uma árvore inicial
arvore = nx.Graph()
arvore.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])
desenhar_arvore(arvore, "Árvore inicial com 5 vértices")

# Adicionando uma aresta que cria um ciclo
arvore.add_edge(5, 2)
desenhar_arvore(arvore, "Grafo após adicionar a aresta (5, 2)")

# Verificando se o grafo contém ciclos
tem_ciclo = not nx.is_tree(arvore)
print(f"O grafo contém um ciclo após a adição da aresta? {tem_ciclo}")
```

## Teorema 1 (Definições equivalentes de árvores). Seja T um grafo com n vértices. Então as seguintes afirmações são equivalentes:
* T é uma árvore.
* T não tem ciclos e tem n - 1 arestas.
* T é conexo e tem n - 1 arestas.
* T é conexo e a remoção de qualquer uma de suas arestas o desconecta.
* Quaisquer dois vértices de T estão conectados por exatamente um caminho.
* T não contém ciclos. E para qualquer nova aresta e, o grafo (T + e) tem exatamente um ciclo.


```python
import networkx as nx

# Criando uma árvore com 5 vértices
arvore = nx.Graph()
arvore.add_edges_from([(1, 2), (2, 3), (3, 4), (4, 5)])

# Verificando as propriedades do Teorema 1
n = len(arvore.nodes)
m = len(arvore.edges)

# Propriedades
print(f"T é uma árvore? {nx.is_tree(arvore)}")
print(f"T não tem ciclos e tem n-1 arestas? {not nx.cycle_basis(arvore) and m == n - 1}")
print(f"T é conexo e tem n-1 arestas? {nx.is_connected(arvore) and m == n - 1}")

# Verificando se a remoção de qualquer aresta desconecta o grafo
desconecta = True
for e in arvore.edges:
    copia = arvore.copy()  # Cria uma cópia do grafo
    copia.remove_edge(*e)  # Remove a aresta da cópia
    if nx.is_connected(copia):  # Verifica se o grafo ainda é conexo
        desconecta = False
        break

print(f"T é conexo e a remoção de qualquer aresta o desconecta? {desconecta}")

# Verificando se quaisquer dois vértices estão conectados por exatamente um caminho
caminho_unico = True
for u in arvore.nodes:
    for v in arvore.nodes:
        if u != v:
            caminhos = list(nx.all_simple_paths(arvore, source=u, target=v))
            if len(caminhos) != 1:  # Deve haver exatamente um caminho simples
                caminho_unico = False
                break

print(f"Quaisquer dois vértices estão conectados por exatamente um caminho? {caminho_unico}")
```
