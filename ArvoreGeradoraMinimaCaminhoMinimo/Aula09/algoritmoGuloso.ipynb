## üí° O que √© um Algoritmo Guloso?

Um algoritmo guloso √© uma estrat√©gia de resolu√ß√£o de problemas que consiste em tomar, a cada etapa, a decis√£o localmente √≥tima, ou seja, a melhor escolha poss√≠vel naquele momento, sem considerar as consequ√™ncias futuras.

**Defini√ß√£o matem√°tica:**  
Seja um problema de otimiza√ß√£o definido por um conjunto de solu√ß√µes vi√°veis $S$ e uma fun√ß√£o objetivo $f: S \rightarrow \mathbb{R}$.  
O algoritmo guloso constr√≥i uma solu√ß√£o $s^* \in S$ iterativamente, escolhendo em cada passo o elemento que proporciona o maior ganho imediato em rela√ß√£o a $f$, sem reavaliar escolhas anteriores.

**Caracter√≠sticas principais:**
- Sempre faz a escolha que parece ser a melhor no momento.
- N√£o volta atr√°s em decis√µes anteriores.
- Busca construir uma solu√ß√£o passo a passo, adicionando elementos que maximizam ou minimizam algum crit√©rio local.
- √â geralmente simples e eficiente, mas nem sempre garante a solu√ß√£o √≥tima global.

**Exemplo fora de grafos:**

## ‚öôÔ∏è Como Funciona um Algoritmo Guloso?

O funcionamento b√°sico segue estes passos:

1. **Inicializa√ß√£o:** Comece com uma solu√ß√£o vazia ou parcial.

3. **Valida√ß√£o:** Verifique se a escolha mant√©m a solu√ß√£o vi√°vel (por exemplo, n√£o forma ciclos, n√£o viola restri√ß√µes).
4. **Atualiza√ß√£o:** Adicione o elemento escolhido √† solu√ß√£o.
5. **Repeti√ß√£o:** Repita os passos at√© que a solu√ß√£o esteja completa.

**Importante:**  
A abordagem gulosa s√≥ garante a solu√ß√£o √≥tima se o problema possuir a propriedade de subestrutura √≥tima e a propriedade de escolha gulosa.

## üßë‚Äçüíª Exerc√≠cio Pr√°tico

Considere o seguinte grafo n√£o direcionado:

- **V√©rtices:** A, B, C, D, E  
- **Arestas:** (A, B), (A, C), (B, D), (C, D), (D, E)

O objetivo √© colorir os v√©rtices do grafo usando o menor n√∫mero poss√≠vel de cores, de modo que v√©rtices ligados por uma aresta n√£o tenham a mesma cor (problema de colora√ß√£o de v√©rtices).

**Tarefa:**  
Utilize uma abordagem gulosa para colorir os v√©rtices do grafo. O algoritmo guloso para colora√ß√£o funciona assim:
1. Escolha um v√©rtice (por exemplo, ordem alfab√©tica).
2. Atribua a menor cor poss√≠vel que n√£o foi usada por seus vizinhos.
3. Repita para todos os v√©rtices.

**Passos esperados:**
- Liste a ordem em que os v√©rtices foram coloridos.
- Para cada passo, indique quais cores j√° foram usadas pelos vizinhos e qual cor foi atribu√≠da ao v√©rtice.
- Ao final, mostre as cores atribu√≠das a cada v√©rtice e quantas cores diferentes foram necess√°rias.

**Reflex√£o:**  
A solu√ß√£o encontrada √© √≥tima? O resultado mudaria se a ordem de escolha dos v√©rtices fosse diferente? Explique.

````python
# Resolu√ß√£o do exerc√≠cio pr√°tico: Colora√ß√£o gulosa de v√©rtices com visualiza√ß√£o dos passos

import networkx as nx
import matplotlib.pyplot as plt

# Definindo o grafo
G = nx.Graph()
G.add_edges_from([('A', 'B'), ('A', 'C'), ('B', 'D'), ('C', 'D'), ('D', 'E')])

# Ordem de colora√ß√£o (pode ser qualquer ordem, aqui usaremos ordem alfab√©tica)
vertices = list(G.nodes())
cores = {}  # Dicion√°rio para armazenar a cor de cada v√©rtice
cores_usadas = ['red', 'blue', 'green', 'yellow', 'orange']

for idx, v in enumerate(vertices):
    # Descobrir as cores dos vizinhos
    vizinhos = list(G.neighbors(v))
    cores_vizinhos = set(cores.get(n) for n in vizinhos)
    # Atribuir a menor cor dispon√≠vel
    for cor in cores_usadas:
        if cor not in cores_vizinhos:
            cores[v] = cor
            break

    # Plotando o passo atual
    color_map = [cores.get(node, 'white') for node in G.nodes()]
    plt.figure(figsize=(4,3))
    nx.draw(G, with_labels=True, node_color=color_map, edge_color='gray')
    plt.title(f"Passo {idx+1}: V√©rtice {v} recebe cor {cores[v]}")
    plt.show()
    print(f"Passo {idx+1}: {v} recebe cor {cores[v]}")

print("Cores finais atribu√≠das:", cores)
````
