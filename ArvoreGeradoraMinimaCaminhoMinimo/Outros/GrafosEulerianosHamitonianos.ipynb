# 5.2 Grafos e Ciclos Eulerianos

- O Problema das Pontes de Königsberg: marco inicial da teoria dos grafos, estudado por Leonhard Euler no século XVIII.
- Grau dos vértices: conceito fundamental para determinar propriedades eulerianas.
- **Teorema 5.1 (Lema do Aperto de Mão):**  
  A soma dos graus de todos os vértices de um grafo é igual ao dobro do número de arestas.
- **Teorema 5.2:**  
  Todo grafo possui um número par de vértices de grau ímpar.
- **Teorema 5.3 (Grafos Dirigidos):**  
  Em grafos dirigidos, a soma dos graus de entrada e a soma dos graus de saída de todos os vértices são iguais ao número de arestas.
- **Definições:**  
  - *Ciclo Euleriano*: trilha fechada que percorre todas as arestas do grafo uma única vez.
  - *Grafo Euleriano*: grafo conexo que admite um ciclo Euleriano.
- **Critério de Euler:**  
  Um grafo conexo é Euleriano se, e somente se, todos os seus vértices têm grau par.
- **Teorema 5.4:**  
  Grafos com todos os vértices de grau par podem ser decompostos em ciclos disjuntos, sem arestas em comum.
- **Teorema 5.6:**  
  Todo grafo Euleriano pode ser decomposto em ciclos que não compartilham arestas.
- **Grafo Semi-euleriano:**  
  Um grafo conexo com exatamente dois vértices de grau ímpar admite um caminho semi-euleriano (trilha aberta que percorre todas as arestas).
- Exemplos práticos: decomposição em ciclos e identificação de grafos semi-eulerianos.

## Texto Explicativo

Esta seção aprofunda o estudo dos grafos e ciclos Eulerianos, tema originado no clássico Problema das Pontes de Königsberg, analisado por Leonhard Euler no século XVIII. Euler investigou se era possível atravessar todas as sete pontes da cidade apenas uma vez, retornando ao ponto de partida. Sua análise resultou em conceitos fundamentais para a teoria dos grafos, especialmente a importância do grau dos vértices.

**Teorema 5.1 (Lema do Aperto de Mão):**  
Em qualquer grafo G, a soma dos graus de todos os vértices é igual ao dobro do número de arestas.  
Formalmente, para um grafo com n vértices e m arestas:  
∑ grau(vi) = 2m

*Justificativa:* Cada aresta conecta dois vértices, sendo contada duas vezes na soma dos graus.

**Teorema 5.2:**  
Todo grafo possui um número par de vértices de grau ímpar.

*Justificativa:* Como a soma dos graus é par, a soma dos graus ímpares também deve ser par, o que só ocorre se houver um número par de vértices ímpares.

**Teorema 5.3 (Grafos Dirigidos):**  
Em grafos dirigidos, a soma dos graus de entrada e a soma dos graus de saída de todos os vértices são iguais ao número total de arestas.

Esses resultados permitem definir:

**Ciclo Euleriano:**  
Uma trilha fechada que percorre todas as arestas do grafo exatamente uma vez.

**Grafo Euleriano:**  
Um grafo conexo que admite um ciclo Euleriano.

**Critério de Euler:**  
Um grafo conexo é Euleriano se, e somente se, todos os seus vértices têm grau par. Isso garante que, ao entrar em um vértice por uma aresta, sempre será possível sair por outra, sem repetir arestas.

**Teorema 5.4 (Ciclos Disjuntos):**  
Se todos os vértices de um grafo têm grau par, é possível decompô-lo em ciclos disjuntos, ou seja, ciclos que não compartilham arestas.

O procedimento consiste em encontrar um ciclo, remover suas arestas e repetir o processo até que todas as arestas sejam utilizadas, formando assim ciclos disjuntos que cobrem o grafo.

**Exemplo:**  
Um grafo com todos os vértices de grau par pode ser decomposto em diferentes conjuntos de ciclos disjuntos, facilitando a visualização das propriedades eulerianas.

**Teorema de Euler (Teorema 5.5):**  
Um grafo conexo é Euleriano se, e somente se, todos os seus vértices têm grau par.

**Teorema 5.6:**  
Um grafo Euleriano pode ser decomposto em ciclos sem sobreposição de arestas.

Antes de avançar para algoritmos, é importante considerar:

**Teorema 5.7 (Grafo Semi-euleriano):**  
Um grafo conexo é semi-euleriano se, e somente se, exatamente dois de seus vértices têm grau ímpar.

**Caminho Semi-euleriano:**  
Uma trilha aberta que percorre todas as arestas do grafo, começando e terminando nos vértices de grau ímpar.

**Exemplo:**  
Em um grafo semi-euleriano, todos os vértices têm grau par, exceto dois. Uma trilha que inicia em um desses vértices ímpares e termina no outro percorre todas as arestas sem repeti-las.

Esses conceitos são essenciais para compreender, identificar e trabalhar com grafos e caminhos Eulerianos e semi-eulerianos, tanto na teoria quanto em aplicações práticas.

````python
# Exemplo prático: decomposição de um grafo Euleriano em ciclos disjuntos usando networkx

import networkx as nx
import matplotlib.pyplot as plt

# Criando um grafo Euleriano simples
G = nx.Graph()
edges = [
    ('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'),
    ('a', 'c'), ('b', 'd')
]
G.add_edges_from(edges)

# Verificando se o grafo é Euleriano
is_eulerian = nx.is_eulerian(G)
print("O grafo é Euleriano?", is_eulerian)

# Encontrando um ciclo Euleriano
if is_eulerian:
    ciclo = list(nx.eulerian_circuit(G))
    print("Ciclo Euleriano encontrado:")
    print(ciclo)

# Visualizando o grafo
pos = nx.circular_layout(G)
nx.draw(G, pos, with_labels=True, node_color='lightblue', edge_color='gray')
plt.title("Exemplo de Grafo Euleriano")
plt.show()
````

# 5.3 Grafos e Ciclos Hamiltonianos

- Origem do conceito: homenagem a William Rowan Hamilton e ao “Jogo Icosiano” (1857), que buscava um ciclo fechado passando por todos os vértices de um dodecaedro regular.
- **Definição:**  
  - *Ciclo Hamiltoniano*: ciclo fechado que passa por todos os vértices do grafo exatamente uma vez.
  - *Grafo Hamiltoniano*: grafo conexo que admite um ciclo Hamiltoniano.
- Exemplo:  
  - Dodecaedro com vértices de A a T; ciclo Hamiltoniano: A, B, C, D, E, H, G, F, O, N, M, L, K, S, T, P, Q, R, I, J, A.
- Diferença em relação aos grafos Eulerianos:  
  - Não existe um critério simples e direto para decidir se um grafo é Hamiltoniano.
- Estratégias para identificação:
  - Grafos completos (ex: K5) são sempre Hamiltonianos.
  - Adicionar arestas a um grafo Hamiltoniano mantém a propriedade.
  - Grafos mais densos tendem a ser Hamiltonianos.
- **Teorema de Ore (1960):**  
  Se G é simples, conexo, com n ≥ 3 vértices, e para todo par de vértices não adjacentes u e v, grau(u) + grau(v) ≥ n, então G é Hamiltoniano.
- Observação:  
  - O teorema de Ore fornece uma condição suficiente, mas não necessária.
  - Mesmo que um grafo não satisfaça o teorema, ele pode ser Hamiltoniano.
  - Para multigrafos, basta considerar o grafo simples correspondente.
- Exemplo:  
  - Grafo com n = 5 e pares não adjacentes: grau(a) + grau(c) = 5 ≥ n, logo é Hamiltoniano.
  - Outro grafo: grau(a) + grau(c) = 4 < n, mas ainda assim pode ser Hamiltoniano.
- Encontrar ciclos Hamiltonianos é fácil em grafos especiais (ciclos, completos), mas no caso geral é um problema de alta complexidade (não polinomial).

## Texto Explicativo

Os Grafos e Ciclos Hamiltonianos receberam esse nome em homenagem ao matemático e astrônomo irlandês William Rowan Hamilton. Em 1857, Hamilton criou o “Jogo Icosiano”, que consistia em encontrar um ciclo fechado passando por todos os vértices de um dodecaedro regular apenas uma vez. O objetivo era inserir pinos e barbantes nos vértices do dodecaedro de madeira, formando o que hoje chamamos de ciclo Hamiltoniano.

**Definição (Ciclo e Grafo Hamiltoniano):**  
Um grafo conexo é Hamiltoniano se contém um ciclo que passa por todos os seus vértices exatamente uma vez. Esse ciclo é chamado de Ciclo Hamiltoniano.

Por exemplo, considere um dodecaedro regular com vértices rotulados de A a T. Uma solução para o jogo icosiano, representada em uma planificação do dodecaedro, pode ser:  
A, B, C, D, E, H, G, F, O, N, M, L, K, S, T, P, Q, R, I, J, A.  
Essa sequência é um ciclo Hamiltoniano.

Diferentemente dos grafos Eulerianos, não existe um critério simples e direto para decidir se um grafo é Hamiltoniano. Uma abordagem inicial é verificar se o grafo pertence a uma classe especial, como os grafos completos (exemplo: K5), que são sempre Hamiltonianos, pois sempre há uma aresta ligando qualquer par de vértices não visitados.

Além disso, ao adicionar uma aresta a um grafo Hamiltoniano, ele permanece Hamiltoniano, pois o ciclo Hamiltoniano continua existindo. Isso sugere que grafos mais densos têm maior probabilidade de serem Hamiltonianos.

Em 1960, Oysten Ore estabeleceu um importante resultado:

**Teorema de Ore:**  
Seja G um grafo simples, conexo, com n ≥ 3 vértices. Se, para todo par de vértices não adjacentes u e v, grau(u) + grau(v) ≥ n, então G é Hamiltoniano.

É importante notar que o teorema de Ore fornece uma condição suficiente, mas não necessária. Ou seja, um grafo pode não satisfazer o teorema e ainda assim ser Hamiltoniano.

**Exemplo:**  
- Grafo com n = 5 vértices e pares não adjacentes: grau(a) + grau(c) = 5 ≥ n, logo é Hamiltoniano.
- Outro grafo: grau(a) + grau(c) = 4 < n, mas ainda assim pode ser Hamiltoniano, pois existe um ciclo Hamiltoniano: a, b, c, d, e, a.

Para multigrafos, basta considerar o grafo simples correspondente (removendo laços e arestas paralelas). Se o grafo simples for Hamiltoniano, o multigrafo também será.

Encontrar ciclos Hamiltonianos é simples em grafos especiais (como ciclos e completos), mas no caso geral é um problema de alta complexidade, sem algoritmo eficiente conhecido para todos os casos.

````python
# Exemplo prático: verificação de ciclo Hamiltoniano usando networkx

import networkx as nx

# Exemplo: grafo completo K5
G = nx.complete_graph(5)
print("Grafo completo K5 é Hamiltoniano?")

# Encontrando um ciclo Hamiltoniano (circuito de Hamilton)
try:
    ciclo = nx.algorithms.tournament.hamiltonian_path(G.to_directed())
    print("Ciclo Hamiltoniano encontrado:", ciclo + [ciclo[0]])
except Exception as e:
    print("Não foi possível encontrar um ciclo Hamiltoniano:", e)

# Observação: networkx não possui função nativa para ciclo Hamiltoniano em grafos não dirigidos,
# mas para grafos completos e pequenos, é possível encontrar manualmente ou por busca exaustiva.
````

# 5.2.1 Algoritmo de Fleury

- Proposto por Pierre-Henry Fleury em 1883 para encontrar ciclos eulerianos e semi-eulerianos.
- Procedimento:
  - Percorra as arestas do grafo, removendo cada aresta já utilizada.
  - Só atravesse uma aresta do tipo “ponte” se não houver outra opção disponível (exceto se for o último caminho possível).
- Regras:
  1. Remova a aresta percorrida; se algum vértice ficar isolado, remova-o também.
  2. Só atravesse uma ponte se não houver outra aresta disponível.
- Complexidade:  
  - Determinar se uma aresta é ponte leva O(|V| + |E|).
  - Como isso é feito para cada aresta, a complexidade total é O(|E|²).
- Observação:  
  - O algoritmo é intuitivo, mas ineficiente para grafos grandes.
  - O algoritmo de Hierholzer, apresentado a seguir, resolve o problema em tempo linear.

````python
# Exemplo prático: Algoritmo de Fleury usando networkx

import networkx as nx

def fleury_algorithm(G, start=None):
    if not nx.is_eulerian(G) and not nx.has_eulerian_path(G):
        return None

    G = G.copy()
    if start is None:
        start = list(G.nodes())[0]
    path = [start]
    u = start

    while G.edges(u):
        for v in list(G.neighbors(u)):
            # Verifica se a aresta é ponte
            G.remove_edge(u, v)
            is_bridge = not nx.is_connected(G)
            G.add_edge(u, v)
            if not is_bridge or len(G.edges(u)) == 1:
                path.append(v)
                G.remove_edge(u, v)
                u = v
                break
    return path

# Exemplo de uso
G = nx.Graph()
edges = [('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'), ('a', 'c'), ('b', 'd')]
G.add_edges_from(edges)
print("Caminho Euleriano pelo algoritmo de Fleury:", fleury_algorithm(G, start='a'))
````

# 5.2.2 Algoritmo de Hierholzer

- Inspirado no trabalho de Hierholzer e Wiener (1873).
- Procedimento para encontrar ciclos eulerianos em grafos eulerianos:
  1. Escolha um vértice inicial arbitrário e encontre um subciclo euleriano.
  2. Se algum vértice do subciclo tiver arestas não utilizadas, inicie um novo subciclo a partir dele.
  3. Una os subciclos formando um ciclo maior.
  4. Repita até que todas as arestas sejam utilizadas.
  5. O ciclo final é um ciclo euleriano do grafo.
- Eficiência:
  - Cada aresta é visitada uma única vez.
  - Complexidade linear: O(|E|).
- Exemplo:  
  - O algoritmo constrói subciclos e os une até formar o ciclo euleriano completo.

## Pseudocódigo do Algoritmo de Hierholzer

Algoritmo: CicloEuleriano_Hierholzer(G)

```plaintext
// inicialização
Vvisit[1] ← s
nVvisit ← 1
k ← 1
enquanto k ≤ nVvisit faça
    u ← Vvisit[k]
    enquanto Iu ≠ null faça
        e0 ← Iu[1] // recebe a primeira aresta incidente a u
        v ← o outro vértice da aresta e0
        Remova a aresta e0 do grafo

        e1 ← e0
        enquanto v ≠ u faça
            se v ∉ Vvisit então
                nVvisit ← nVvisit + 1
                Vvisit[nVvisit] ← v
            e2 ← Iv[1] // recebe a primeira aresta incidente a v
            nextEdge[e1] ← e2
            prevEdge[e2] ← e1
            se EulerEdge[v] = null então
                EulerEdge[v] ← e2
            e1 ← e2
            v ← o outro vértice da aresta e1
            Remova a aresta e1 do grafo
        // um subciclo em u acabou de ser completado
        prevEdge[e0] ← e1
        nextEdge[e1] ← e0
        se EulerEdge[u] = null então
            EulerEdge[u] ← e0
        senão
            // insere um novo subciclo no subciclo preexistente em u
            e1 ← EulerEdge[u]
            e2 ← prevEdge[e1]
            e3 ← prevEdge[e0]
            nextEdge[e2] ← e0
            nextEdge[e3] ← e1
    k ← k + 1
```

````python
# Exemplo prático: Algoritmo de Hierholzer usando networkx

import networkx as nx

def hierholzer_algorithm(G):
    if not nx.is_eulerian(G):
        return None

    G = G.copy()
    circuit = []
    stack = [next(iter(G.nodes()))]

    while stack:
        u = stack[-1]
        if G.degree(u) == 0:
            circuit.append(stack.pop())
        else:
            v = next(iter(G.neighbors(u)))
            stack.append(v)
            G.remove_edge(u, v)
    return circuit[::-1]

# Exemplo de uso
G = nx.Graph()
edges = [('a', 'b'), ('b', 'c'), ('c', 'd'), ('d', 'a'), ('a', 'c'), ('b', 'd')]
G.add_edges_from(edges)
print("Ciclo Euleriano pelo algoritmo de Hierholzer:", hierholzer_algorithm(G))
````

## Pseudocódigo detalhado do Algoritmo de Hierholzer

O algoritmo de Hierholzer é um método eficiente para encontrar ciclos eulerianos em grafos eulerianos. Ele parte do princípio de que, em um grafo onde todos os vértices têm grau par e o grafo é conexo, é possível construir um ciclo que percorra todas as arestas exatamente uma vez. O procedimento consiste em construir subciclos e unir esses subciclos até que todas as arestas tenham sido utilizadas.

O algoritmo funciona da seguinte forma:

1. Inicialize uma lista de vértices visitados, começando por um vértice arbitrário s.
2. Para cada vértice visitado, percorra as arestas incidentes, formando um subciclo euleriano.
3. Sempre que encontrar um vértice com arestas não utilizadas, inicie um novo subciclo a partir dele.
4. Una os subciclos ao ciclo principal, mantendo a ordem correta das arestas.
5. Repita o processo até que todas as arestas do grafo tenham sido percorridas.
6. O resultado final é um ciclo euleriano que cobre todas as arestas do grafo.

Abaixo está o pseudocódigo detalhado do algoritmo:

Algoritmo: CicloEuleriano_Hierholzer(G)

```plaintext
// inicialização
Vvisit[1] ← s
nVvisit ← 1
k ← 1
enquanto k ≤ nVvisit faça
    u ← Vvisit[k]
    enquanto Iu ≠ null faça
        e0 ← Iu[1] // recebe a primeira aresta incidente a u
        v ← o outro vértice da aresta e0
        Remova a aresta e0 do grafo

        e1 ← e0
        enquanto v ≠ u faça
            se v ∉ Vvisit então
                nVvisit ← nVvisit + 1
                Vvisit[nVvisit] ← v
            e2 ← Iv[1] // recebe a primeira aresta incidente a v
            nextEdge[e1] ← e2
            prevEdge[e2] ← e1
            se EulerEdge[v] = null então
                EulerEdge[v] ← e2
            e1 ← e2
            v ← o outro vértice da aresta e1
            Remova a aresta e1 do grafo
        // um subciclo em u acabou de ser completado
        prevEdge[e0] ← e1
        nextEdge[e1] ← e0
        se EulerEdge[u] = null então
            EulerEdge[u] ← e0
        senão
            // insere um novo subciclo no subciclo preexistente em u
            e1 ← EulerEdge[u]
            e2 ← prevEdge[e1]
            e3 ← prevEdge[e0]
            nextEdge[e2] ← e0
            nextEdge[e3] ← e1
    k ← k + 1
```

Esse algoritmo garante que todas as arestas do grafo sejam percorridas exatamente uma vez, formando o ciclo euleriano de maneira eficiente e com complexidade linear em relação ao número de arestas.
